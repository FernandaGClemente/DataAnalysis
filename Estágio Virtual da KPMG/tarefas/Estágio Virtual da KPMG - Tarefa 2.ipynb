### Resposta da KPMG em português: <br>
EXPLORAÇÃO DE DADOS 

<br>

Compreender as características dos campos fornecidos nos dados subjacentes, como distribuições de variáveis, se o conjunto de dados está inclinado para um determinado grupo demográfico e a validade dos dados nos campos. Por exemplo, um conjunto de dados de treinamento pode estar fortemente inclinado para a faixa etária mais jovem. Se for esse o caso, como isso afetará seus resultados ao usá-lo para prever o restante da base de clientes.

<br>

Identificar limitações em torno dos dados e coletar dados externos que possam ser úteis para fins de modelagem. Isso pode incluir a inclusão de dados do ABS em diferentes níveis geográficos e a criação de recursos adicionais para o modelo. Por exemplo, a distância geográfica de diferentes códigos postais pode ser usada como indicador de proximidade para determinar se um cliente precisa de uma bicicleta para ir ao trabalho.

<br>

Explorar interações entre diferentes variáveis por meio de análise de correlação e verificar a multicolinearidade criando variáveis de interação. Um exemplo dessa correlação pode ocorrer entre as variáveis independentes idade e tempo de serviço - ou seja, pessoas em faixas etárias mais avançadas terão um tempo de serviço mais longo.

<br>

Além disso, transformar os dados necessários para que estejam em um formato apropriado para análise. Isso pode incluir etapas como garantir que os tipos de dados sejam apropriados e agregar os dados para um nível agregado. Ou seja, incluir dados já agregados do ABS em um nível geográfico para criar variáveis adicionais.

<br>

Documentar suposições, limitações e exclusões dos dados; bem como como você melhoraria na próxima etapa se houvesse tempo adicional para abordar suposições e remover limitações.

<br>

DESENVOLVIMENTO DO MODELO

<br>

Determinar uma hipótese relacionada à pergunta de negócio que possa ser respondida com os dados. Realizar testes estatísticos para determinar se a hipótese é válida ou não.

<br>

Criar campos calculados com base nos dados existentes, por exemplo, converter a data de nascimento em faixa etária. Outros campos que podem ser projetados incluem "Produto de Margem Alta", que pode indicar se o produto comprado pelo cliente está em uma categoria de margem alta nos últimos três meses com base nos campos "preço de lista" e "custo padrão".

<br>

Outros exemplos incluem o cálculo da distância entre o escritório e o endereço residencial como um fator para determinar se os clientes podem comprar uma bicicleta para fins de transporte. Além disso, isso pode incluir reflexões sobre a definição da variável prevista. Por exemplo, os resultados são previstos em intervalos ordinais, nominais, binários ou contínuos. Teste o desempenho do modelo usando fatores relevantes para o modelo escolhido (por exemplo, desvio residual, AIC, curvas ROC, R ao quadrado). Documente adequadamente o desempenho do modelo, suposições e limitações.
