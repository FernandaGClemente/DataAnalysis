Problemas notáveis de qualidade de dados que foram encontrados e os métodos usados para mitigar os dados identificados as inconsistências são as seguintes. Além disso, foram fornecidas recomendações para evitar a recorrência de problemas de qualidade de dados e melhorar a precisão dos dados subjacentes usados para impulsionar os negócios decisões.

<br>

1 - CUSTOMER_IDS adicionais na “Tabela de transações” e “Tabela de endereço do cliente” mas não em “Mestre de Clientes (Demografia do Cliente)”
Mitigação: certifique-se de que todas as tabelas sejam do mesmo período. Somente os clientes da lista Master de Clientes serão usados como um conjunto de treinamento para nosso modelo.
Isso indica que os dados recebidos podem não estar sincronizados entre si, o que pode distorcer os resultados da análise se houver registros de dados ausentes. Consulte o arquivo Excel 'data_outliers.xlsx' para obter a lista de valores discrepantes entre as tabelas.

<br>

2 - Várias colunas, como a marca de uma compra ou cargo, têm valores vazios em certos registros
Mitigação: se apenas um pequeno número de linhas estiver vazio, filtre o registro inteiramente do conjunto de treinamento para previsão. Caso contrário, se for um campo principal, impute com base na distribuição no conjunto de dados de treinamento.
Para conjuntos de dados importantes, como transações, menos de 1% das transações (totalizando menos de 0,1% da receita) têm campos ausentes. Esses registros foram removidos do conjunto de dados de treinamento.

<br>

3 – Valores inconsistentes para o mesmo atributo (por exemplo, Victoria sendo representada como “V”, “Vic” e “Victoria”)
Mitigação: use expressões regulares para substituir valores estendidos por abreviações para garantir a consistência através de endereços.
Recomendação: imponha uma lista suspensa para o usuário inserir os dados em vez de um campo de texto livre.
Para construir variáveis significativas para o modelo, os dados foram limpos para evitar múltiplas representações do mesmo valor. Além disso, os registros de gênero onde 'U' foram substituídos com base na distribuição do conjunto de dados de treinamento.

<br>

4 - Tipo de dados inconsistente para o mesmo atributo (por exemplo, valores numéricos para alguns campos e strings para outros)
Mitigação: Converta os registros selecionados em caracteres para numéricos. Remova os caracteres não numéricos da string.
Recomendação: Certifique-se de que as tabelas de fatos no banco de dados fornecido tenham restrições nos tipos de dados. Ter diferentes tipos de dados para um determinado campo dificulta a interpretação dos resultados no estágio posterior. Portanto, transformações de dados apropriadas são feitas para garantir tipos de dados consistentes para um determinado campo.
