# KPI: Key Performance Indicators
[&#x1F53C;](#Sumário:) Clique para voltar para o sumário

Hoje em dia, a competitividade das empresas e, por conseguinte, o seu sucesso empresarial dependem em muito boa parte da forma como os gestores as gerem no dia-a-dia. A boa gestão está, como se sabe, dependente também do conhecimento que tem sobre o desempenho da empresa e do impacto das decisões que tomam. Ter disponível em tempo útil a informação crítica e verdadeira sobre as realizações da empresa é uma parte fundamental de um processo de gestão, para que o gestor possa tomar decisões mais rápidas e mais corretas.

<br>

Conhecer os resultados atingidos, confrontá-los com as metas previamente definidas, compará-los com os valores do setor, compreender a sua evolução ao longo do tempo e perceber a sua tendência gera uma informação de valor indiscutível para quem tem de tomar decisões. O “domínio do negócio” acontece quando o gestor conhece as realizações da empresa, das suas unidades de negócio e do andamento dos seus projetos.

<br>

Existem milhares de indicadores que podem ser utilizados na gestão de todas as áreas da empresa. O importante é saber escolher quais os mais importantes KPIS e aplicá-los, adaptando-os à realidade da empresa e às necessidades de informação. É verdade que nem sempre é fácil operacionalizar sistemas de monitorização. São muitos os casos de implementações de modelos de monitorização que não tiveram sucesso. As razões são também muitas. Mas também existem casos de sucesso e o valor obtido para a gestão é muito superior ao investimento efetuado. Para que isso ocorra é necessário começar e ir descobrindo quais os indicadores que podem de fato fazer a diferença na entrega de informação a quem tem de conhecer a performance e tomar decisões.

<br>

A urgência necessária das decisões obriga a que tenhamos de estar permanentemente ligados à informação de gestão. Assim, os níveis de exigência são cada vez mais elevados, necessitando de mais informação, mais rápida e com maior qualidade. Em resumo:

<br>

1 - Dados com maior qualidade, quantidade e rapidez > <br>
2 - Escolha dos indicadores e KPIS > <br>
3 - Monitorização (Medir) > <br>
4 - Conhecimento do desempenho (Resultado) e do impacto das decisões > <br>
5 - Visualização da informação > <br>
6 - Domínio do negócio > <br>
7 - Gestão empresarial > <br>
8 - Competitividade empresarial > <br>
9 - Sucesso empresarial <br>

# Conceitos
[&#x1F53C;](#Sumário:) Clique para voltar para o sumário

#### # Objetivo - A sua função é simplesmente apurar o nível das realizações da organização (resultados) para que estas possam ser comparadas com as metas pré-estabelecidas e apurado o desvio e o respectivo nível de performance. Devem ser capazes de expressar com clareza um determinado proposito estratégico e operacional, esclarecendo aos colaboradores sobre o que a empresa pretende conseguir num determinado aspecto e num determinado período de tempo. 
#### Exemplo: Qual é a intenção? > Aumentar em 4% as vendas
<br>    

#### # Métrica - São os valores brutos usados para mensurar e monitorar os processos de forma simples e objetivas, são o ponto de partida para a criação dos indicadores de desempenho, podem ser de vários tipos, como valor, quantidade, peso, volume, etc. 
#### Exemplo: Valor da venda, custo do produto, quantidade de produto vendido, quantidade de cliente atendido, quantidade de reclamação. 
<br>

#### # Indicador - São valores calculados que usam as métricas como materia prima, tem por objetivo verificar o cumprimento das metas (esperado x realizado). Os indicadorestem um significado, não são somente um número ou uma quantidade, podem ser apresentados por percentuais, comparações, faixa de preço, valor ou frequência. Mostra se a métrica é bom ou ruim.
#### Exemplo: Qual é o instrumento que será utilizado para medir o objetivo? > Taxa de variação das vendas, Taxa de conversão, Farol e índice de datisfação do cliente.
<br>

#### # KPI (Indicadores-chaves) - São indicadores mais importantes para a organização ou área de negocio e estão diretamente ligada as metas, tem por objetivo apoiar a tomada de decisão, a criação de plano de ação, projetos e mudanças em processos. Devem ser rastreáveis para que seja possível identificar problemas que estão afetando os KPIS.
#### Exemplo: Analisar a performace da organização ou área de negócio, analisar tendências dos clientes e do mercado, analisar a satisfação dos clientes.

<br>

#### # Algoritmo - É a fórmula de cálculo que permite apurar o resultado registado.
#### Exemplo: Como se calcula automaticamente o resultado? > (Vendas atuais / Vendas do período anterior – 1)
<br>

#### # Resultado - É o valor que se apura através do algoritmo. Representa um resultado alcançado pela empresa num determinado período de tempo.
#### Exemplo: Quanto é que atingimos? > 5%
<br>

#### # Meta - Tem por função eliminar a subjetividade, reforçar o compromisso, identificar a ambição, fomentar a melhoria contínua e promover a inovação.
#### Exemplo: Quanto pretendemos atingir? > 4%, Contratar mais 3 pessoas para a área de vendas até dezembro deste ano, Diminuir em 20% os custos com logística em 6 meses, Duplicar o número de publicações no instagram em 3 meses, Aumentar em 20% as vendas do 2° semestre em relação ao mesmo período do ano passado e Aumentar a avaliação média da empresa no google em 30% nos próximos 6 meses
* Como criar metas: 
    * Específica (O que se quer alcançar?, Quem será o responsável?, Existem pré-requisitos e/ou limitações?)
    * Mensurável (Qual percentual, quantidade, nível, prazo ou quantia monetária deve ser alcançada?)
    * Alcançável (O que torna a meta possível e realista dentro da empresa?)
    * Relevante (Por que a meta é importante neste momento e contexto da organização?)
    * Temporal (Qual o prazo para a realização da meta?)
    
<br>

#### # Desvio - Representa a diferença entre a meta previamente estabelecida e o resultado efetivamente realizado. Pode-se apresentar em número absoluto ou em %.
#### Exemplo: Qual foi a diferença do resultado face ao planejado (meta)? > 1% em valor absoluto e 25% em percentagem

<br>

#### # Performance - É uma forma de apresentar o resultado de modo a comparar indicadores que tenham diferentes unidades de medida. Regra geral, a performance é apresentada em %, sendo 100% quando atinge a meta, <100% quando não atinge e > 100% quando supera a meta.
#### Exemplo: Qual foi a performance? > 125%
<br>    

#### # Avaliação - Representa uma avaliação qualitativa obtida em função da performance/desvio. Por exemplo: Supera, atinge, em alerta e não atinge. Permite facilitar a comunicação do nível de performance. 
#### Exemplo: Qual é o ponto da situação? > Superamos a meta
    * [< 95%] = [Não atingimos]
    * [95%, 100%] = [Alerta]
    * [>=100%] = [Atinge] 
    * [>120%] = [Supera] 
    
#### # Dashboard - Dá acesso a métrica e KPIS de forma clara e objetiva, acompanha o desempenho da empresa ou área, apoia a tomada de decisão, permite a identificação de problemas rapidamente, incentiva a criação de novas ideias para o negócio e promove a transparência corporativa.

<br>

### <u>Criação:</u> <br>
#### a) Comece pelas metas >>> Saber onde quer chegar é o primeiro passo para definir quais indicadores precisam ser monitorados.
#### b) Identificar os pontos críticos >>> Precisa saber o que pode afetar o comprimento dos seus objetivos para que esses itens não fiquem de fora da lista de indicadores.
#### c) Garanta que as métricas estejam disponíveis >>> Avalie quais métricas são necessárias para a criação do indicador e se todas elas estão disponíveis.
#### d) Periodicidade >>> Defina o período de acompanhamento do indicador e a periodicidade de atualização dele.
#### Exemplos:
    * Meta = Aumentar em 20% as vendas do 2° semestre em relação ao mesmo período do ano passado.
    * Métrica = Volume de venda
    * Indicador = % de aumento nas vendas
    
    * Meta = Aumentar a avaliação média da empresa no google em 30% nos próximos 6 meses
    * Métrica = Avaliação dos clientes
    * Indicador = % de aumento na média da avaliação dos clientes
<br>

# Característica de um bom indicador
[&#x1F53C;](#Sumário:) Clique para voltar para o sumário

1. <u>Pertinência dos indicadores para a gestão:</u> O resultado que o indicador apresenta precisa acrescentar algum valor, ser importante para conhecer o desempenho da empresa; 
2. <u>Credibilidade do resultado:</u> Procure fontes de informação credíveis ou que tenham um processo de registro protegido de eventuais erros de introdução; Faça auditorias às fontes de dados dos seus indicadores;
3. <u>Esforço aceitável para o apuramento do resultado:</u> “valor da informação para a gestão” X “esforço para calcular o resultado”; Não podemos ter situações em que o custo de obtenção da informação é superior ao próprio valor da informação;
4. <u>Simplicidade de interpretação:</u> É muito importante que os destinatários compreendam aquilo que os indicadores se propõem a medir. Ter a correta e rápida interpretação dos resultado; Evite indicadores muito complexos que possam gerar confusão nos seus destinatários;
5. <u>Simplicidade do algoritmo de cálculo:</u> Quanto mais simples for o algoritmo do indicador, mais rápido e seguro será o processo de apuramento do resultado;     
6. <u>Fonte de dados dentro de “casa”:</u> Importa que os dados que alimentam o algoritmo estejam em suportes ou bases de dados de acesso fácil; É sempre mais complicado o acesso aos dados, se a sua localização for no exterior da unidade orgânica ou da organização;
7. <u>Cálculo automático:</u> A possibilidade de ter os indicadores a serem alimentados automaticamente, evitando a intervenção humana, dá credibilidade e torna mais ágil o processo de monitorização;
8. <u>Possibilidade de auditar as fontes de dados com eficácia:</u> A possibilidade de os dados poderem ser auditados e poderem ser identificados erros faz com que os responsáveis pela sua introdução sejam mais cautelosos no seu tratamento; 
9. <u>Alinhado com a frequência de monitorização:</u> Os indicadores devem ser capazes de apresentar os seus resultados com frequência igual ou superior à necessária para a monitorização estabelecida na empresa;
10. <u>Possibilidade de calcular em momentos extraordinários:</u> Por vezes, é necessária uma monitorização extraordinária; Importa que nos indicadores mais críticos seja possível calcular atempadamente o resultado do indicador, de forma a apresentar à gestão; 
11. <u>Protegido de efeitos externos:</u> É necessário identificar e compreender os efeitos externos que podem alterar/esconder a verdadeira dimensão do desempenho interno da empresa; Avalie sempre se os indicadores são influenciados por fatores externos; Tente eliminar ou reduzir o efeito desses fatores, através da implementação de condições ou da apresentação de outros indicadores com capacidade de explicar esses fatores;
12. <u>Não gera efeitos perversos:</u> Antes de validar o indicador, promova alguma reflexão para identificar os possíveis efeitos perversos e implementar medidas para os minimizar/eliminar; Por exemplo, verificar se a utilização de um determinado indicador não provoca na empresa um efeito negativo na eficácia, eficiência ou qualidade, na área em que se está a medir;
13. <u>Possibilidade de Benchmarking:</u> Benchmarking consiste no processo de busca das melhores práticas de gestão da entidade numa determinada indústria e que conduzem ao desempenho superior; A comparação de desempenhos entre atividades, projetos, unidades de negócio, organizações, etc., é sempre útil. Para além de induzir competição, promove substancialmente a melhoria contínua. Procure uniformizar os indicadores utilizados dentro da empresa, utilizando por exemplo escalas de medição comparáveis entre si.
14. <u>Atualizado:</u> Os indicadores devem ser rapidamente substituídos por outros, quando deixam de ser interessantes (obsoletos) ou quando surgem novas prioridades, atividades ou projetos na empresa;
15. <u>Possibilidade de ter uma meta:</u> A existência de uma meta é uma referência preciosa para que se possa perceber a distância que as realizações estão dos valores ideais. Os indicadores devem ter a possibilidade de ter metas anuais e intercalares, em função da frequência de monitorização estabelecida pela empresa.

# Níveis organizacionais
[&#x1F53C;](#Sumário:) Clique para voltar para o sumário

### Os níveis hierárquicos de uma organização podem variar dependendo do tamanho e da complexidade da empresa, bem como da sua estrutura organizacional. No entanto, em geral, existem:

<br>

* Nível Institucional - Alta Gerência (Tipo Estratégico): 
    * Alta gerência é formada pelos executivos de mais alto nível da empresa, como o CEO, o COO e o CFO.
    * São responsáveis pela tomada de decisões estratégicas da empresa e geralmente são os principais responsáveis pela gestão da organização.
    * Olha a crganização como um todo. 
    * Mostra os KPIS mais importantes de cada área e acompanha a evolução dos resultados.
    * Está ligado a missão e visão
    * Objetivo de médio e longo prazo
    * Exemplos: 
        - Market Share (Posição da empresa no Ranking Nacional - Comparar o faturamento e a diferença com o concorrente) 
        - Lucratividade (Lucro líquido / Receita Total x 100) 
        - Churn Rote (% Cancelamento por cliente = Total de cliente cancelados / Total de clientes ativos)
        - Turnover geral ((Total de admissão + Total de demissão / 2) / Total colaborador ativo) 
        - NPS (Net Promoter Score = % Promotores - % Detratores)
        - CSAT (Customer Satisfaction Score = Total Clientes satisfeitos / Total de clientes entrevistados
    
<br>    

* Nível Intermediário - Gerência Média (Tipo Tático):
    * A gerência média é formada pelos gerentes de nível médio da empresa, que estão abaixo da alta gerência e acima da gerência operacional. 
    * São responsáveis por gerenciar os departamentos e equipes específicas da empresa e garantir que as metas e objetivos sejam atingidos.
    * Cada área/departamento do negócio
    * Apoia a tomada de decisão com focos nos objetivos de médio prazo
    * Avaliar a eficácia ou erros na estratégia da área
    
<br>   

* Nível operacional - Gerência operacional e Funcionários operacionais (Tipo Operacional): 
    * A gerência operacional é formada pelos gerentes responsáveis por supervisionar as atividades diárias da empresa, incluindo a gestão de RH, finanças, produção, vendas, entre outras.
    * Os funcionários operacionais são a base da organização e são responsáveis por realizar as tarefas diárias da empresa. Eles trabalham em áreas específicas da empresa e realizam as atividades operacionais necessárias para que a empresa possa atingir seus objetivos.
    * Cada operação - Gerente e analistas
    * Apoia a otimização dos processos e acompanhamento de atividades e metas
    * Identificar erros e melhorias
    * Atualizados em tempo real
    * Medir a eficiência e a eficácia da operação ou colaborador

<br>

#### OBS: Existe ainda 3 tipos de indicadores: 
    a) Produtividade (Acompanhamento da gestão de processos - Execução de tarefas: Quanto tempo para se produzir x produtos)
    b) Capacidade (Medir a capacidade dos equipamentos em geral das industrias)  
    c) Qualidade (Usados na gestão de questões relativas ao nível de satisfação dos clientes ou exigências de normas de fiscalização ou certificações)

#### <u>Principais problema de negócio:</u>
1. Falta de fluxo de caixa (Resultar de um planejamento financeiro ruim, gerenciamento de caixa ineficaz ou despesas inesperadas)
2. Relatórios financeiros imprecisos (Resultar de erros no registro de transações financeiras, falha em seguir padrões contábeis ou manipulação intencional de demonstrações financeiras)

7. Falta de comunicação (É crucial para o planejamento financeiro e a tomada de decisões, pois pode levar a mal-entendidos, ineficiências e oportunidades perdidas)

## Tesouraria
[&#x1F53C;](#Sumário:) Clique para voltar para o sumário

#### <u>Indicadores:</u>
1. Ciclo de conversão de caixa (CCC): Este KPI mede a quantidade de tempo que leva para uma empresa converter seus investimentos em estoque e contas a receber em entradas de caixa. As empresas podem usar essa métrica para melhorar a gestão do fluxo de caixa e otimizar seu capital de giro.
    * Cálculo: Ciclo Operacional (Prazo Médio de Estoque (PME) + Prazo Médio de Recebimento (PMR)) – Prazo Médio Pagamentos a Fornecedores (PMP)
    * Origem da informação: 
    * Apuração: Mensalmente
    * Polaridade: Positiva (Quanto maior o valor, melhor)
    * Visualização: Gráfico de Coluna
    
<br>

2. Dias de vendas pendentes (DSO) ou Prazo médio de recebimento (PMR): Este KPI mede o número médio de dias que uma empresa leva para receber o pagamento dos clientes. Identifica quanto tempo demora a empresa, em média, a receber os créditos que concede aos seus clientes. Um valor alto é, em termos financeiros, desfavorável, mostrando por vezes ineficiência na área das cobranças ou falta de poder negocial junto dos clientes. (A minimização deste valor ajuda a tesouraria e a imagem junto das instituições financeiras. Quanto menor for o seu valor, menor serão as necessidades de fundo de maneio por parte da empresa. Muitas empresas necessitam de recorrer a financiamento bancário para acompanhar o crédito que dão aos seus clientes, o que origina custos de financiamento, reduzindo a rendibilidade do negócio.)
    * Cálculo: [Clientes / [(Rendimentos) × (1 + Taxa IVA)]] × 365 (Dias ou Meses)
    * Origem da informação: Balanço Patrimonial e DRE
    * Apuração: Mensalmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: Gráfico de colunas (PMR x Tempo em meses) com meta e média (2 linhas)

<br>

3. Prazo médio de pagamento (PMP): Identifica quanto tempo demora a empresa, em média, a pagar as suas dívidas aos fornecedores. Um valor muito elevado pode identificar dificuldades da empresa em satisfazer as suas obrigações. No entanto, um valor baixo pode revelar falta de poder negocial da empresa perante os seus fornecedores. (A minimização deste valor ajuda a manter uma boa imagem da empresa junto de fornecedores e das instituições financeiras.)
    * Cálculo: [Fornecedores / [(Compras + FSE) × (1 + Taxa IVA)]] × 365 (Dias ou Meses)
    * Origem da informação: Balanço Patrimonial e DRE
    * Apuração: Mensalmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: Gráfico de colunas (PMP x Tempos em meses) com meta e média (2 linhas)

<br>

4. Saldo de caixa (Disponibilidades à ordem): Este KPI mede a quantidade de caixa que uma empresa tem em mãos. Representa a disponibilidade financeira imediata para fazer face aos pagamentos correntes da empresa. Importa estar alinhado com as previsões do orçamento de tesouraria, de modo a poder honrar os compromissos que já foram e que iram ser assumidos e investir em oportunidades de crescimento. (Pode ser útil definir o valor do saldo mínimo que permite pagar as despesas mais críticas bem como um saldo ideal que permite dar algum conforto para os compromissos de um horizonte temporal de 1 a 2 meses.)
    * Cálculo: Disponibilidades financeiras nas contas à ordem 
    * Origem da informação: Contabilidade/Tesouraria
    * Apuração: Diariamente ou semanalmente
    * Polaridade: Positiva (Quanto maior o valor, melhor)
    * Visualização: Gráfico de Coluna (Real X Tempo em dias) com limites (linhas dos saldos mínimos e ideal)

<br>

5. Retorno sobre o investimento (ROI): Este KPI mede o retorno sobre os investimentos de uma empresa, incluindo seu caixa e capital de giro. As empresas podem usar essa métrica para avaliar a eficácia de suas decisões de investimento e identificar áreas onde podem melhorar seu ROI.
    * Cálculo: (Receita – Custo do Investimento) / Custo do Investimento.
    * Origem da informação: DRE
    * Apuração: Mensalmente
    * Polaridade: Positiva (Quanto maior o valor, melhor)
    * Visualização: Gráfico de Coluna

<br>

6. Taxa de cobertura de juros: Este KPI mede a capacidade de uma empresa de cobrir seus pagamentos de juros com seu lucro antes de juros e impostos (EBIT). As empresas podem usar essa métrica para avaliar sua saúde financeira e gerenciar suas obrigações de dívida. (Aponta a quantidade de tempo em que as despesas com juros podem ser pagas com o lucro de um determinado período.)
    * Cálculo: Divisão entre o Lucro antes dos Impostos e Tributos (EBIT) e as Despesas com Juros

(É relevante perceber quais os montantes que estão em dívida há mais de 3 meses, 6 meses e 1 ano, de modo a definir prioridades para o departamento de cobranças.)

    * Origem da informação: Contabilidade/Tesouraria
    * Apuração: Diariamente / semanalmente / mensalmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: Gráfico de coluna empilhada % (% Clientes em atraso x % Total de Clientes x Tempo em meses)

<br>

9. % de valores de fornecedores em atraso: Este indicador apresenta o valor das dívidas da empresa aos seus fornecedores que já venceram o prazo de pagamento que tinha sido estipulado. À medida que este valor aumenta, a imagem da empresa fica cada vez mais prejudicada e perde capacidade de poder adquirir matérias-primas e serviços com condições vantajosas. (Deve-se tentar minimizar este valor de forma a não prejudicar a imagem da empresa e as suas condições de pagamento. É relevante perceber quais os montantes que estão em dívida há mais de 3 meses, 6 meses e 1 ano.)
    * Cálculo: Valores de fornecedores que já passaram o prazo de pagamento/Valores em fornecedores (%)
    * Origem da informação: Contabilidade/Tesouraria
    * Apuração: Diariamente / semanalmente / mensalmente
    * Polaridade: Negativa (Quanto menor o valor, melhor) 
    * Visualização: Gráfico de coluna empilhada % (% Fornecedores em atraso x Total de Fornecedores x % Fornecedores em atraso mais de 3 meses x Tempo em meses)

10. Índice de dívida em relação ao patrimônio: Este KPI mede a quantidade de dívida que uma empresa tem em relação ao seu patrimônio. As empresas podem usar essa métrica para avaliar sua alavancagem financeira e gerenciar sua exposição ao risco. É bastante utilizado por investidores para avaliar o endividamento de empresas listadas na bolsa. (Alguns analistas de investimento sugerem buscar empresas que tenham menos de 50% de dívida em relação ao seu patrimônio.)
    * Cálculo: Divisão entre a soma do endividamento de uma empresa e o total de bens e direitos.
    * Origem da informação: Balanço Patrimonial 
    * Apuração: Anualmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: Gráfico de coluna
    
<br>

11. Liquidez reduzida (LR): Relaciona os elementos mais líquidos das aplicações com o passivo corrente. admite dificuldades conjunturais e de possível falência, pelo que considera que os inventários não poderão ser transformados de imediato em dinheiro ou, pelo menos, que essa conversão será sempre abaixo do custo de inventário. (O indicador de liquidez reduzida tem em vista destacar a influência dos inventários na liquidez da empresa. Quanto maior o valor do indicador, maior será a capacidade da empresa em solver os compromissos de curto prazo.)
    * Cálculo: (Ativo corrente – Inventários) / Passivo corrente (Índice)
    * Origem da informação: Balanço Patrimonial 
    * Apuração: Mensalmente
    * Polaridade: Positiva (Quanto maior o valor, melhor)
    * Visualização: Gráfico de linha (Índice X Tempo em meses)

12. Liquidez Imediata (LI): Dá-nos a conhecer o grau de cobertura do passivo circulante por disponibilidades. Quanto maior o valor do indicador, maior será a capacidade da empresa em solver os compromissos de curto prazo. (Note-se que este indicador não tem qualquer racionalidade do ponto de vista da gestão financeira, pois uma gestão financeira eficiente deve minimizar o volume de disponibilidades, diminuindo, dessa forma, as necessidades de financiamento.)
    * Cálculo: Disponibilidades / Passivo corrente (Índice)
    * Origem da informação: Balanço Patrimonial 
    * Apuração: Mensalmente
    * Polaridade: Positiva (Quanto maior o valor, melhor) 
    * Visualização: Gráfico de linha (Índice X Tempo em meses) 

<br>

13. Duração do ciclo de tesouraria de exploração: O ciclo de tesouraria de exploração inicia-se com a aquisição de matérias primas e conclui-se com os recebimentos efetivos dos clientes. Este ciclo é dependente de três variáveis: duração média de inventários em armazém, prazo médio de recebimento e prazo médio de pagamento. Quanto maior o ciclo de tesouraria, maior será o esforço pedido ao fundo de maneio. (A DMIA exprime o tempo que as existências em armazém permitem satisfazer os consumos da empresa. A optimização da duração do ciclo de tesouraria obriga à existência de um equilíbrio entre a redução dos recebimentos, a folga dos pagamentos, as condições de pagamento e as políticas de crédito da empresa.)
    * Cálculo: Duração Média de Inventários em Armazém (DMIA) + PMR – PMP DMIA = (Inventários/Vendas) × 365 (Dias ou Meses) 
    * Origem da informação: Balanço Patrimonial e DRE 
    * Apuração: Mensalmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: Gráfico de área (PMR x DMIA x Tempo em meses) com linha de duração do ciclo de tesouraria

<br>

14. Fundo de maneio (FM): Para colmatar a existência de rupturas de tesouraria, é necessário que a empresa disponha de uma margem de segurança. Esta margem de segurança toma a designação de Fundo de Maneio (FM). A determinação do FM tem uma importância crítica na gestão das empresas, já que é aqui onde as empresas investem uma grande parte da sua liquidez.
    * Cálculo: Capitais Permanentes – Ativo não-corrente (Real)
    * Origem da informação: Balanço Patrimonial
    * Apuração: Mensalmente
    * Polaridade: Positiva (Quanto maior o valor, melhor)
    * Visualização: Gráfico de coluna (Real FM x Tempo em meses) com linhas mostrando os capitais permanentes e ativos não correntes

<br>

#### <u>Principais problema de negócio:</u>
1. Gerenciamento do fluxo de caixa: o gerenciamento de entradas e saídas de caixa é crucial para qualquer organização, e o departamento de tesouraria é responsável por garantir que haja dinheiro suficiente disponível para atender às obrigações financeiras. No entanto, gerenciar o fluxo de caixa pode ser desafiador, especialmente durante períodos de volatilidade econômica ou eventos inesperados.
2. Gestão do risco de taxa de juro: A tesouraria deve gerir a exposição da organização ao risco de taxa de juro, que pode ter um impacto significativo no desempenho financeiro da organização. Isso pode envolver o gerenciamento do portfólio de dívidas da organização, proteção contra flutuações nas taxas de juros e monitoramento das tendências das taxas de juros.
3. Gerenciamento de risco cambial: Se uma organização opera em vários países ou realiza negócios em várias moedas, o departamento de tesouraria deve gerenciar o risco cambial. Isso pode envolver cobertura contra flutuações cambiais, monitoramento de taxas de câmbio e gerenciamento de exposições a moedas estrangeiras.
4. Gestão de investimentos: O departamento de tesouraria é responsável por investir o excesso de caixa da organização de forma a gerar retornos enquanto gerencia os riscos. Isso pode envolver o gerenciamento de investimentos em ações, títulos e outros instrumentos financeiros.
5. Conformidade e questões regulatórias: O departamento de tesouraria deve garantir que a organização cumpra os regulamentos e padrões financeiros relevantes. Isso pode envolver manter-se atualizado sobre as mudanças regulatórias, manter registros precisos e garantir que as transações financeiras sejam devidamente documentadas e relatadas.
6. Segurança cibernética e fraude: o departamento de tesouraria pode ser alvo de ataques cibernéticos e atividades fraudulentas, por isso deve tomar medidas para se proteger contra essas ameaças. Isso pode envolver a implementação de medidas de segurança robustas, monitoramento de transações financeiras em busca de atividades suspeitas e treinamento da equipe sobre como detectar e prevenir fraudes.

## RH

    * Cálculo: (Número de Funcionários que Deixaram a Empresa ÷ Número Médio de Funcionários) x 100
    * Origem da informação: registros de recursos humanos, como planilhas de controle de funcionários, registros de contratações e demissões, sistemas de gerenciamento de pessoal, etc.
    * Apuração: mensal, trimestral ou anualmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: gráficos de barras, gráficos de linha ou tabelas comparativas

<br>

2. Tempo de contratação: Este KPI mede o tempo médio necessário para preencher uma vaga. Um tempo de contratação mais longo pode indicar problemas de recrutamento, seleção de candidatos ou falta de urgência no processo de contratação.
    * Cálculo: Soma dos tempos de contratação de todas as vagas preenchidas ÷ Número total de vagas preenchidas
    * Origem da informação: obtidas por meio de sistemas de rastreamento de candidatos (ATS), registros de RH, planilhas de controle ou softwares específicos para gerenciamento de processos seletivos.
    * Apuração: mensalmente ou trimestralmente, para acompanhar o desempenho do processo de contratação ao longo do tempo.
    * Polaridade: Negativa (Quanto menor o valor, melhor) um tempo de contratação mais curto é desejável, pois indica eficiência no processo de seleção e contratação de novos funcionários
    * Visualização: gráficos de barras, gráficos de linha ou gráficos de dispersão que mostram a média de tempo de contratação ao longo do tempo

<br>

3. Custo por contratação: esse KPI mede o custo total de preenchimento de uma vaga, incluindo despesas com publicidade, recrutamento e integração. É importante acompanhar esse KPI para garantir que o custo de contratação não exceda os benefícios de adicionar novos funcionários à equipe.
    * Cálculo: Total de Custos de Contratação ÷ Número Total de Contratações
    * Origem da informação: obtidos através de registros financeiros, como notas fiscais, recibos e relatórios de despesas relacionados às atividades de contratação
    * Apuração: mensalmente ou trimestralmente, para acompanhar o desempenho do processo de contratação e identificar possíveis oportunidades de redução de custos ou eficiência
    * Polaridade: uma polaridade negativa (resultados mais baixos) é geralmente preferível.
    * Visualização: Gráficos de barras ou gráficos de linha podem ser usados para comparar o custo por contratação ao longo do tempo ou entre departamentos/posições específicas

<br>

4. Envolvimento dos funcionários: Este KPI mede o nível de satisfação e comprometimento que os funcionários têm com seu trabalho e com a empresa. Altos níveis de engajamento dos funcionários podem levar a maior produtividade, melhores taxas de retenção e um ambiente de trabalho positivo.
    * Cálculo: As perguntas devem ser projetadas para medir o grau de satisfação, o comprometimento, o senso de pertencimento e o engajamento dos funcionários com a empresa e seu trabalho. As respostas podem ser classificadas em uma escala, geralmente de 1 a 5 ou de 1 a 10, e o resultado pode ser obtido calculando a média das respostas.
    * Origem da informação: Requer a coleta de dados por meio de pesquisas de opinião, questionários ou avaliações de satisfação
    * Apuração: anuais ou semestrais (também podem ser realizadas pesquisas pontuais ou pulse surveys em momentos específicos para obter feedback atualizado)
    * Polaridade: Quanto maior for o resultado, maior será o nível de envolvimento dos funcionários.
    * Visualização: gráficos de barras, gráficos de linha ou gráficos de pizza (pode ser útil segmentar os resultados por departamento, níveis hierárquicos ou outras categorias relevantes para identificar variações entre grupos e áreas de melhoria específicas.)

<br>

5. Treinamento e desenvolvimento: Este KPI mede o investimento feito pela empresa em programas de treinamento e desenvolvimento de funcionários. É importante acompanhar esse KPI para garantir que os funcionários adquiram as habilidades e conhecimentos necessários para realizar seu trabalho de maneira eficaz e eficiente.
    * Cálculo: Total de Investimentos em Treinamento e Desenvolvimento (despesas com a contratação de instrutores, compra de materiais didáticos, custos de viagens para treinamentos, taxas de participação em cursos externos, aquisição de plataformas de e-learning, entre outros custos) / Número Total de Funcionários Beneficiados 
    * Origem da informação: obtida a partir de registros financeiros, como orçamentos e despesas específicas relacionadas a programas de treinamento e desenvolvimento. Também é importante registrar quais funcionários participaram dos programas, pois esses dados são necessários para calcular a média de investimento por funcionário.
    * Apuração: mensalmente, trimestralmente ou anualmente
    * Polaridade:Quanto maior for o resultado, maior será o investimento mais alto indica que a empresa está comprometida em capacitar e desenvolver seus funcionários
    * Visualização: gráficos de barras, gráficos de linha ou gráficos de pizza

<br>

6. Taxa de absenteísmo: Este KPI mede o número de dias que os funcionários estão ausentes do trabalho, seja por doença, férias ou motivos pessoais. Uma alta taxa de absenteísmo pode indicar problemas com o moral dos funcionários, carga de trabalho ou satisfação no trabalho. As principais causas do absentismo são: doenças, acidentes de trabalho, faltas justificadas, atrasos, faltas injustificadas, suspensão disciplinar, greves internas, situações familiares (assistência inadiável, luto, casamento, etc.), serviço militar, formação, atividade de organismos representativos de trabalhadores (sindical, comissão de trabalhadores).
    * Cálculo: Número de horas de ausência ao trabalho num período/ de horas potenciais de trabalho no período (%)
    * Origem da informação: Departamento de pessoal
    * Apuração: Mensalmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: Gráfico de Barra (% Taxa de Absentismo x Tempo em meses) com média em linha (meta)
    
<br>

7. Diversidade e inclusão: Este KPI mede o nível de diversidade dentro da empresa, incluindo gênero, etnia e idade. É importante acompanhar esse KPI para garantir que a empresa esteja criando uma cultura de local de trabalho inclusiva e diversificada.
    * Cálculo: Para calcular a diversidade de gênero, é necessário comparar o número de funcionários de cada gênero em relação ao total. Para medir a diversidade étnica, deve-se considerar a proporção de funcionários pertencentes a diferentes grupos étnicos. Além disso, a diversidade de idade pode ser calculada observando a distribuição etária dos colaboradores.
    * Origem da informação:  obtidos por meio de registros de RH, como sistemas de gerenciamento de recursos humanos, formulários de admissão e pesquisas internas. (É importante garantir que os dados sejam coletados de maneira ética, respeitando a privacidade dos funcionários e em conformidade com as políticas de proteção de dados.)
    * Apuração: anualmente
    * Polaridade: O objetivo principal é identificar e acompanhar a diversidade dentro da empresa, garantindo uma representação justa e inclusiva de diferentes grupos. O resultado é avaliado com base na comparação entre a distribuição real dos funcionários em diferentes categorias demográficas e as metas de diversidade estabelecidas pela organização.
    * Visualização: gráficos de barras, gráficos de pizza ou gráficos de dispersão
    
<br>  

8. Gerenciamento de desempenho: Este KPI mede a eficácia do processo de gerenciamento de desempenho, incluindo definição de metas, análises de desempenho e feedback dos funcionários. É importante acompanhar esse KPI para garantir que os funcionários sejam avaliados de forma justa e consistente.
    * Cálculo: Os resultados podem ser agregados e analisados para obter uma visão geral do processo de gerenciamento de desempenho.
    * Origem da informação: Uma abordagem comum é coletar feedback dos funcionários por meio de pesquisas, avaliações de desempenho e revisões individuais, utilizando escalas ou pontuações predefinidas e os registros de metas definidas e alcançadas podem indicar a eficácia do processo de definição de metas e a capacidade de acompanhamento e mensuração do desempenho.
    * Apuração: anualmente 
    * Polaridade: Um resultado positivo indica que o processo de gerenciamento de desempenho está sendo eficaz na definição de metas claras, fornecimento de análises de desempenho adequadas e feedback construtivo para os funcionários. 
    * Visualização: gráficos de barras, gráficos de radar ou gráficos de linha.

<br>

9. Remuneração média por trabalhador: Indica o nível de salários praticados pela empresa. Corresponde à remuneração média que cada trabalhador aufere por um ano de trabalho na empresa. Acaba por ser um indicador extremamente importante, na medida em que identifica um dos custos principais da empresa e permite compará-lo com a média do sector. Um valor baixo tanto pode indicar competitividade nos gastos com pessoal, mas também pode indicar que a empresa tem colaboradores menos qualificados e com baixa motivação. Importa analisar este valor relativamente à média do sector e enquadrado na estratégia da empresa.
    * Cálculo: Remunerações anuais dos colaboradores / de colaboradores (Real)
    * Origem da informação: Departamento de pessoal
    * Apuração: Anualmente
    * Polaridade: Depende da estratégia da empresa
    * Visualização: Gráfico de Barra com linha de meta (Real Remuneração média x Média do setor x Tempo em anos)

<br>

10. Leque salarial ilíquido: O índice do leque salarial indica o de vezes que a remuneração mais elevada cobre a remuneração mais baixa. Quanto maior for este valor, maior será a distância entre a remuneração mais elevada e a remuneração mais baixa. Um aumento do leque salarial pode indicar a existência de desequilíbrio salarial na empresa. Importa analisar este índice, tendo em conta o género e por categorias. É importante também perceber o posicionamento da remuneração base ilíquida média da empresa. O valor óptimo deste indicador depende da estratégia da empresa.
    * Cálculo: Maior remuneração base ilíquida/Menor remuneração base ilíquida (índice)
    * Origem da informação: Departamento de pessoal
    * Apuração: Anualmente
    * Polaridade: Depende da estratégia da empresa
    * Visualização: Gráfico de coluna com eixo secundário (Real Salário x Tempo em anos) com linha (Leque salarial)

<br>

11. Taxa de trabalho extraordinário: Refere-se ao peso do número de horas de trabalho extraordinário realizado em relação ao total de horas trabalháveis (considerando o total de trabalhadores que exerceram funções no período em referência). Um elevado peso das horas de trabalho extraordinário é sinónimo de um acréscimo de gastos por parte da empresa com as despesas com pessoal. Indica também que existiram situações não previstas/excepcionais na atividade da empresa.
    * Cálculo: Número de horas de trabalho extraordinário/Total de horas trabalháveis (%)
    * Origem da informação: Departamento de pessoal
    * Apuração: Mensalmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: Gráfico de coluna empilhada com eixo secundário (% Taxa de trabalho extraordinário x Horas Extraordinárias x Horas trabalháveis x Tempo em meses)

<br>

12. Índice de tecnicidade: Identifica o peso dos colaboradores que são considerados técnicos superiores. Em algumas empresas é crítico ter um índice de tecnicidade elevado, já que pode ser sinónimo de uma empresa que aposta/investe forte na contratação de quadros qualificados. Existem empresas onde este índice é muito baixo devido à forte componente de colaboradores operacionais (ex.: empresas de construção civil, empresas agrícolas, etc.). Existem outras empresas em que o peso de técnicos superiores deve ser muito elevado (ex.: empresas de consultoria, de informática, de investigação, farmacêuticas, etc.)
    * Cálculo: Número de técnicos superiores/Total de colaboradores (%)
    * Origem da informação: Departamento de pessoal
    * Apuração: Quadrimestralmente/Anualmente
    * Polaridade: Depende do tipo de empresa
    * Visualização: Gráfico de coluna empilhada (% Técnicos superiores x % Total de colaboradores x Tempo em trimestre)

<br>

13. Idade média: Indica a idade média dos colaboradores da empresa. Este indicador pode ser calculado por categorias, por género, por habilitações, por departamentos, etc. Permite perceber o nível de envelhecimento dos colaboradores da empresa.
    * Cálculo: Σ das idades/ de pessoas ao serviço (anos)
    * Origem da informação: Departamento de pessoal
    * Apuração: Anualmente
    * Polaridade: Depende da estratégia da empresa
    * Visualização: Gráfico de linhas (Idade x Tempo em anos)
    * Resultado: A análise da idade média deve ser analisada com algum cuidado. Em algumas empresas, um nível elevado pode ser sinónimo de recursos humanos com muita experiência. Noutras empresas, poderia ser indício de baixa capacidade operacional devido ao envelhecimento dos seus quadros.

<br>

14. Antiguidade média: Indica o número médio de anos que os colaboradores estão ao serviço da empresa. Não existindo alterações no quadro de colaboradores, o resultado deste indicador incrementa todos os anos um ano.    
    * Cálculo: Σ das antiguidades/ de pessoas ao serviço (anos)
    * Origem da informação: Departamento de pessoal
    * Apuração: Anualmente
    * Polaridade: Depende da estratégia da empresa
    * Visualização: Gráfico de linhas (Idade x Tempo em anos)
    * Resultado: Um valor elevado da antiguidade média pode indicar estabilidade da empresa, mas também pode reduzir a capacidade da empresa em inovar. Quanto maior a antiguidade média, maiores serão os custos associados à indemnização dos colaboradores
    
<br>

15. Taxa de contratação a termo: Apresenta a relação entre os colaboradores que estão a termo na empresa relativamente aos efetivos. Se o valor for igual a 100%, significa que existem tantos colaboradores a termo como colaboradores efetivos.
    * Cálculo: Número de contratados a termo/Total de efetivos do quadro (%)
    * Origem da informação: Departamento de pessoal
    * Apuração: Mensalmente/Trimestralmente
    * Polaridade: Depende da estratégia da empresa
    * Visualização: Gráfico de coluna empilhada com eixo secundário (Contratados a termos x efetivos x % Taxa de contratação a termo x Tempo em meses)
    * Resultado: Um valor elevado da taxa de contratação a termo pode indicar a existência de alguma flexibilidade na política laboral, mas também pode indiciar alguma instabilidade derivada à provável rotação de colaboradores a termo.

<br>

16. Taxa de incidência de acidentes no local de trabalho: Indica o nível de segurança no trabalho e a eficácia das medidas preventivas adaptadas pela empresa. Considera-se acidente de trabalho os “decorrentes das características da atividade profissional desempenhada” (acidentes de trabalho típicos) e os “ocorrido no percurso entre a residência e o local de trabalho e vice-versa” (acidentes de trabalho de trajeto). 
    * Cálculo: Número de acidentes no local de trabalho / Total de colaboradores (%)
    * Origem da informação: Departamento de pessoal
    * Apuração: Semanalmente/Mensalmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: Gráfico de coluna com linha de média (% Taxa de incidência de acidentes x Tempo em meses)

<br>

17. Médias de horas de formação por trabalhador: Identifica o número de horas médio de formação que os trabalhadores de uma empresa tiveram durante um ano. A legislação do trabalho obriga a que as entidades empregadoras assegurem um número mínimo de horas de formação aos seus colaboradores, de modo a garantir a permanente atualização dos conhecimentos e competências.
    * Cálculo: Total de horas de formação/ de trabalhadores (%)
    * Origem da informação: Departamento de pessoal
    * Apuração: Mensalmente/Trimestralmente
    * Polaridade: Positiva (Quanto maior o valor, melhor)
    * Visualização: Gráfico de coluna com linha de meta (N° de horas de formação x Tempo em trimestre)
    
<br>

#### <u>Principais problema de negócio:</u>
1. Aquisição e retenção de talentos: Encontrar e atrair os melhores talentos é um desafio para os departamentos de RH. Uma vez contratados, reter funcionários também é um desafio, pois os funcionários podem sair por melhores oportunidades ou outros motivos.
2. Desenvolvimento de funcionários: Oferecer oportunidades para crescimento e desenvolvimento de funcionários é essencial para reter talentos, mas pode ser difícil encontrar os recursos e programas certos para atender às diversas necessidades dos funcionários.
3. Diversidade e inclusão: Garantir um local de trabalho diverso e inclusivo é crucial para promover a inovação, a criatividade e a satisfação dos funcionários. No entanto, construir e manter uma cultura inclusiva pode ser um desafio, principalmente em organizações com normas culturais de longa data.
4. Cumprimento das leis e regulamentos trabalhistas: os departamentos de RH devem garantir a conformidade com várias leis e regulamentos trabalhistas, como leis de oportunidades iguais de emprego, leis de salários e horários e padrões de segurança no local de trabalho. A não conformidade pode resultar em responsabilidades legais e danos à reputação da organização.
5. Relações com funcionários: os departamentos de RH devem lidar com questões de relações com funcionários, como conflitos entre funcionários, reclamações de assédio e outras questões no local de trabalho que podem afetar o moral e a produtividade dos funcionários.
6. Tecnologia: O uso crescente de tecnologia em processos de RH, como recrutamento, avaliação de desempenho e treinamento, criou desafios relacionados à privacidade de dados, segurança cibernética e confiança dos funcionários no uso da tecnologia.
7. Restrições orçamentárias: os departamentos de RH devem equilibrar a necessidade de investir no desenvolvimento dos funcionários, benefícios e outras iniciativas com orçamentos e recursos limitados.

#### <u>Principais problema de negócio:</u>
1. Ameaças de cibersegurança: As ameaças de cibersegurança, como violações de dados e ataques de ransomware, são uma preocupação constante do departamento de sistemas de informação. O departamento deve garantir que medidas de segurança estejam em vigor para proteção contra essas ameaças e que a equipe seja treinada para detectar e responder a incidentes de segurança.
2. Sistemas legados: Os sistemas legados podem ser um desafio para o departamento de sistemas de informação, pois podem ser difíceis de manter, integrar com sistemas mais novos ou seguros. O departamento deve encontrar maneiras de modernizar ou substituir os sistemas legados para melhorar a eficiência e a segurança.
3. Gerenciamento de dados: O departamento de sistema de informação deve gerenciar os dados da organização, incluindo armazenamento, processamento e análise. Isso pode envolver garantir a qualidade dos dados, proteger dados confidenciais e garantir a conformidade com os regulamentos de privacidade de dados.
4. Migração para a nuvem: muitas organizações estão migrando sua infraestrutura de tecnologia para a nuvem, o que pode ser um processo complexo e demorado. O departamento de sistema de informação deve planejar e executar estratégias de migração para a nuvem, garantindo que os dados estejam seguros e acessíveis.
5. Gestão de talentos: Atrair e reter profissionais de tecnologia qualificados pode ser um desafio para o departamento de sistemas de informação. O departamento deve encontrar maneiras de atrair e reter os melhores talentos, ao mesmo tempo em que oferece oportunidades de crescimento e desenvolvimento profissional.
6. Integração: À medida que as organizações adotam novas tecnologias, o departamento de sistemas de informação deve garantir que essas tecnologias se integrem perfeitamente aos sistemas existentes. Isso pode envolver o desenvolvimento de integrações personalizadas, usando middleware ou alavancando APIs.
7. Restrições orçamentárias: O departamento de sistema de informação deve gerenciar os investimentos em tecnologia dentro das restrições orçamentárias da organização. Isso pode envolver priorizar projetos de tecnologia com base em seu impacto potencial na organização e encontrar maneiras de reduzir custos enquanto mantém os níveis de serviço.

## Operações / Produção
[&#x1F53C;](#Sumário:) Clique para voltar para o sumário

#### <u>Indicadores:</u>
1. Eficácia geral do equipamento (OEE): Este KPI mede a eficiência com que o equipamento está sendo usado para produzir mercadorias. Ele leva em consideração fatores como disponibilidade, desempenho e qualidade.
    * Cálculo: multiplicação de três índices: disponibilidade, desempenho e qualidade
        * Disponibilidade mede a porcentagem de tempo em que a máquina está disponível para produção. 
        * Desempenho mede a eficiência do equipamento em relação à sua capacidade máxima. 
        * Qualidade mede a porcentagem de produtos produzidos sem defeito em relação ao total de produtos produzidos.
    * Origem da informação: coletados a partir de indicadores de desempenho dos equipamentos, como tempo de parada, tempo de ciclo, tempo de setup, n° de peças produzidas com defeito etc
    * Apuração: mensal ou semanal.
    * Polaridade: Positiva (Quanto maior o valor, melhor) = quanto maior o valor do índice, melhor é a eficácia geral do equipamento.
    * Visualização: um gráfico de barras com três cores diferentes, representando cada um dos índices: disponibilidade, desempenho e qualidade. Assim, é possível identificar facilmente onde estão os maiores problemas a serem resolvidos.
    
<br>    
    
2. Tempo de ciclo: Este KPI mede o tempo que leva para completar um ciclo de produção, desde o início da produção até a entrega dos produtos acabados.
    * Cálculo: mede o tempo necessário para concluir um ciclo de produção, desde o início até a entrega do produto final.
    * Origem da informação: podem ser obtidos por meio de sistemas de gestão de produção, registros manuais do tempo de produção e tempo de espera e observação direta.
    * Apuração: dias, semanas ou meses
    * Polaridade: Negativa (Quanto menor o valor, melhor) = uma vez que um tempo de ciclo de produção menor indica maior eficiência e produtividade da equipe de produção.
    * Visualização: um gráfico de linha, onde é possível identificar facilmente variações e tendências ao longo do tempo.
    
<br>    
    
3. Taxa de defeitos ou Porcentagem de produtos com inconformidades: Este KPI mede o número de defeitos em uma execução de produção, expresso como uma porcentagem do número total de unidades produzidas. Este indicador identifica a percentagem de produtos com inconformidades (defeitos) detectados após o processo de produção. É um indicador que pretende avaliar a qualidade da produção final da empresa.
    * Cálculo: Número de produtos com inconformidades / total de produtos produzidos (%)
    * Origem da informação: Direção de produção / Direção qualidade
    * Apuração: Semanalmente/Mensalmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: 2 gráficos, sendo um de coluna (N° de produtos com inconformidades x Tempo em meses) e um de linha (% de produtos com inconformidades x Tempo em meses)
    * Resultado: Importa identificar o tipo de inconformidades que podem pôr em causa a qualidade do produto. Também poderão existir níveis diferentes no tipo de inconformidades, de modo a distinguir as mais críticas. A análise das inconformidades permitirá reavaliar o processo de produção, de modo a eliminar os aspectos que possam comprometer a qualidade dos produtos.
    
<br>    
    
4. Taxa de rendimento: Este KPI mede a porcentagem de unidades boas produzidas em comparação com o número total de unidades produzidas.
    * Cálculo: divisão do número de produtos produzidos pelo número total de produtos que poderiam ser produzidos em um determinado período de tempo.
    * Origem da informação: podem ser encontrados em sistemas de gerenciamento de produção
    * Apuração: mensal ou trimestral.
    * Polaridade: Positiva (Quanto maior o valor, melhor) = quanto maior a taxa de rendimento, melhor é o desempenho do departamento.
    * Visualização: um gráfico de barras mostrando o número de produtos produzidos versus o número total de produtos que poderiam ser produzidos no período de tempo em questão.
    
<br>    
    
5. Tempo de inatividade: Este KPI mede a quantidade de tempo que o equipamento não está disponível para produção devido a manutenção, reparos ou outros fatores.
    * a) Porcentagem de tempo de paragem da produção para reparações: Este indicador permite avaliar o peso na paragem da produção devido às reparações de máquinas e equipamentos. Valores elevados neste indicador podem indiciar problemas nos equipamentos produtivos, nomeadamente aspectos relacionados com a sua antiguidade. Podem também indiciar más práticas na utilização dos mesmos por parte dos funcionários da empresa. Importa relacionar este indicador com a % da capacidade instalada utilizada, uma vez que as reparações têm maior probabilidade de serem necessárias quando os equipamentos são utilizados com maior intensidade, nomeadamente nos picos de produção.
        * Cálculo: Tempo de paragem para reparações / Tempo total de paragem da produção (%)
        * Origem da informação: Direção de produção
        * Apuração: Mensalmente
        * Polaridade: Negativa (Quanto menor o valor, melhor)
        * Visualização: Gráfico de coluna empilhada com eixo secundário (Hora de paragem x Hora totais de paragem x % da capacidade utilização x tempo em meses)
    * b) Porcentagem de tempo de paragem da produção por quebras de stock: Este indicador permite avaliar o peso na paragem da produção devido às quebras de stock. Permite assim evidenciar as falhas por consequência de um mau planeamento das necessidades de matéria-prima ou por incapacidade de manter uma contabilização eficaz do inventário. As quebras de stock também podem ter origem em furtos, obsolescência, etc. Quando a empresa opta por ter um stock elevado, corre menor riscos de ter quebra de stocks, mas acaba por ter um custo maior com a sua gestão. Importa assim encontrar um ponto de equilíbrio que permita manter um stock de segurança com o menor custo possível e, ao mesmo tempo, minimizar o risco de quebras de stock.
        * Cálculo: Tempo de paragem por quebra de stock/Tempo total de paragem da produção (%)
        * Origem da informação: Direção de produção
        * Apuração: Mensalmente/Trimestralmente
        * Polaridade: Negativa (Quanto menor o valor, melhor)
        * Visualização: Gráfico de coluna empilhada (Horas de paragem por quebra x Horas totais de paragem x Tempo em meses)
    * c) Número de quebras de produção devido a ocorrências inesperadas: Identifica o de vezes que a produção parou devido a ocorrências inesperadas, tais como quebras de energia e avarias. Conhecer o tipo de causa e a frequência com que surgem estas situações é fundamental na fase de programação da produção, caso contrário a empresa arrisca-se a ser demasiado ambiciosa na produção que deseja obter.
        * Cálculo: Σ do de vezes que a produção parou por tipo de causa (N°)
        * Origem da informação: Direção de produção
        * Apuração: Mensalmente/Trimestralmente
        * Polaridade: Negativa (Quanto menor o valor, melhor)
        * Visualização: Gráfico de barras deitadas (invertidas) com a quantidade das causas
    
<br>    
    
6. Utilização da capacidade ou Porcentagem da capacidade instalada usada: Este KPI mede a porcentagem da capacidade de produção que está sendo usada em um determinado momento. O conceito de capacidade instalada pode ser traduzido como o limite máximo de produção (plena capacidade de produção). É a quantidade de unidades de produto/serviço que as máquinas e equipamentos/recursos humanos instalados são capazes de produzir. É fundamental que a empresa conheça este valor de modo a não aceitar encomendas que não possa cumprir dentro dos prazos acordados.
    * Cálculo: Capacidade utilizada (produção realizada) / Capacidade instalada (limite máximo de produção) (%)
    * Origem da informação: Direção de produção
    * Apuração: Semanalmente/Mensalmente
    * Polaridade: Positiva (Quanto maior o valor, melhor)
    * Visualização: Gráfico de área com linha (% de produção realizada e capacidade instalada x área de produção planejada x tempo em meses)
    
<br>    
    
7. Taxa de sucata ou Porcentagem do desperdício da matéria-prima: Este KPI mede a porcentagem de matérias-primas que são desperdiçadas durante o processo de produção. Indica a percentagem de desperdício de matérias-primas num determinado período de tempo, por linha de produção ou por produto. Tem por objetivo final a identificação de medidas de melhorias que previnam a reincidência das perdas no processo produtivo da empresa. A redução dos desperdícios de matéria-prima permite aumentar a eficiência produtiva da empresa através da redução dos gastos provenientes de menores necessidades de matéria-prima para obter o mesmo nível de produção. Existem ainda ganhos decorrentes da diminuição das necessidades de espaços de armazenagem.
    * Cálculo: Matéria-prima desperdiçada / Matéria-prima utilizada (%)
    * Origem da informação: Direção de produção
    * Apuração: Semanalmente/Mensalmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: Gráfico de coluna com eixo secundário (N° de mp desperdiçada x n° de mp utilizada x % de mp desperdiçada) 

<br>

8. Produção: É claramente um dos principais indicadores para as empresas. Indica aquilo que se está a produzir num determinado período de tempo. A produção pode ser calculada não apenas no seu total, mas também por linha de produção e por produto. O controle da produção torna-se mais eficaz quando a empresa define previamente um plano de produção e vai monitorizando esse plano (produção planeada) com as realizações que vai efetuando.
    * Cálculo: Σ dos produtos/serviços produzidos (n°, m², m³ etc.)
    * Origem da informação: Direção de produção
    * Apuração: Diariamente/Semanalmente
    * Polaridade: Positiva (Quanto maior o valor, melhor)
    * Visualização: Gráfico de área com linha (n° de produção realizada e produção média x área de produção planejada x n° de metros)

<br>

9. Custo por unidade produzida: Este indicador identifica o custo médio de produção de uma unidade de produto. É essencialmente um indicador de medição de eficiência. Este indicador pode ser calculado por tipo de produtos. Importa relacionar com as quantidades produzidas, uma vez que elas podem ter impacto na diluição dos custos fixos da produção.
    * Cálculo: Custos de produção/ total de produtos produzidos (Real)
    * Origem da informação: Direção de produção / Direção financeira
    * Apuração: Semanalmente/Mensalmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: Gráfico de coluna com eixo secundário (N° de unidades produzidas x Real custo por unidades x Tempo em meses)

<br>

10. Custo de manutenção por unidade produzida: É fundamental perceber a dimensão e a origem dos custos da produção, de forma a compreender a margem económica dos produtos. Os custos de manutenção podem ter um peso relevante, principalmente nas empresas industriais. Os custos de manutenção envolvem os custos da manutenção preventiva e corretiva. A antiguidade dos equipamentos pode começar a gerar custos excessivos ao nível da manutenção. As manutenções preventivas efetuadas em excesso podem retirar margem à produção, pelo que é sempre importante encontrar um equilíbrio.
    * Cálculo: Custos com manutenção/ total de unidades produzidas (Real)
    * Origem da informação: Direção de produção / Direção financeira
    * Apuração: Trimestralmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: Gráfica de coluna com linha de tendência dos custos de manutenção (Custo de manutenção por unidade x Tempo em meses)

<br>

11. % Manutenção preventiva relativa à manutenção total: Este indicador mede o peso dos custos da manutenção preventiva relativamente aos custos totais de manutenção. Procura perceber a importância da manutenção preventiva nas políticas globais de manutenção da empresa.
    * Cálculo: Custos da manutenção preventiva / Custos totais da manutenção preventiva e corretiva (%)
    * Origem da informação: Direção de produção / Direção financeira
    * Apuração: Mensalmente/Trimestralmente
    * Polaridade: Positiva (Quanto maior o valor, melhor)
    * Visualização: 2 gráficos, sendo um de coluna empilhada (Real custo de manutenção preventiva x Total de custo de manutenção x Tempo em meses) e de linha (% manutenção preventiva relativa a total x Tempo em meses)
    * Resultado: Quanto mais próximo o resultado do indicador estiver de 100%, melhor será para a empresa, uma vez que identifica uma situação de inexistência de manutenção corretiva, ou seja, a manutenção preventiva/planeada é eficaz e permite evitar as situações de surpresa das intervenções corretivas. De qualquer forma, importa salvaguardar que a manutenção preventiva não seja excessiva.
    
<br>

#### <u>Principais problema de negócio:</u>
1. Restrições de capacidade: o departamento de produção deve garantir que tenha capacidade suficiente para atender à demanda do cliente, mas as restrições de capacidade podem limitar a quantidade de bens ou serviços que podem ser produzidos. Isso pode incluir restrições de equipamentos, pessoal ou espaço.
2. Controle de qualidade: O departamento de produção deve garantir que seus produtos ou serviços atendam aos padrões de qualidade e que a qualidade seja mantida durante todo o processo de produção. Isso pode envolver o estabelecimento de procedimentos de controle de qualidade, treinamento de pessoal sobre padrões de qualidade e monitoramento da qualidade do produto ou serviço.
3. Interrupções na cadeia de suprimentos: o departamento de produção pode contar com uma rede complexa de fornecedores e fornecedores para fornecer matérias-primas, componentes ou serviços. Interrupções na cadeia de suprimentos, como atrasos ou problemas de qualidade, podem afetar os cronogramas de produção e a satisfação do cliente.
4. Gerenciamento de estoque: o departamento de produção deve gerenciar os níveis de estoque para garantir que haja estoque suficiente disponível para atender à demanda do cliente, evitando o excesso de estoque, que pode imobilizar o capital e aumentar os custos de armazenamento.
5. Manutenção de equipamentos: O departamento de produção deve manter seus equipamentos para garantir que eles operem de forma eficiente e eficaz. Isso pode envolver o agendamento de manutenção de rotina, reparo de equipamentos conforme necessário e substituição de equipamentos quando necessário.
6. Segurança e conformidade regulamentar: O departamento de produção deve cumprir as normas regulamentares e de segurança para proteger funcionários e clientes e evitar responsabilidades legais. Isso pode envolver o estabelecimento de procedimentos de segurança, treinamento de equipe em protocolos de segurança e garantia de que equipamentos e processos estejam em conformidade com os padrões regulamentares.
7. Melhoria contínua: O departamento de produção deve melhorar continuamente seus processos para aumentar a eficiência, reduzir o desperdício e melhorar a qualidade do produto ou serviço. Isso pode envolver a implementação de princípios de manufatura enxuta, adoção de novas tecnologias ou automação ou uso de análise de dados para identificar áreas de melhoria.

## Logística
[&#x1F53C;](#Sumário:) Clique para voltar para o sumário

#### <u>Indicadores:</u>
1. Entrega no prazo: mede a porcentagem de pedidos que são entregues antes ou na data de entrega prometida. Uma alta taxa de entrega no prazo pode melhorar a satisfação e a fidelidade do cliente.
    * Cálculo: (Número de pedidos entregues no prazo / Número total de pedidos) x 100
        * Defina o período de tempo que deseja avaliar a entrega no prazo. 
        * Identifique o número total de pedidos que foram entregues no prazo durante o período definido.
        * Divida o número de pedidos entregues no prazo pelo número total de pedidos e multiplique por 100 para obter a porcentagem de entregas no prazo.
    * Origem da informação: Registros de vendas para encontrar o número total de pedidos e o número de entregas no prazo
    * Apuração: Mensalmente ou trimestralmente
    * Polaridade: Positivo (Quanto maior o valor, melhor)
    * Visualização: Gráfico de barras ou gráfico de linha que mostra a taxa de entrega no prazo ao longo do tempo, incluindo uma meta e destacando as variações significativas em relação à meta.
    * Resultado: Identificar áreas que precisam de melhorias e a tomar medidas corretivas para aumentar a taxa de entrega no prazo.
    
<br>    
    
2. Prazo de entrega do pedido: mede o tempo que leva desde que um pedido é feito até que seja entregue. Um lead time mais curto pode melhorar a satisfação do cliente e reduzir os custos de estoque.
    * Cálculo: Média do tempo de entrega = (Soma do tempo de entrega de cada pedido avaliado / Número total de pedidos avaliados)
        * Defina o período de tempo que deseja avaliar o prazo de entrega do pedido. 
        * Identifique a data em que o pedido foi feito e a data em que o pedido foi entregue.
        * Calcule a diferença de tempo entre a data do pedido e a data de entrega para cada pedido avaliado.
        * Calcule a média do tempo de entrega de todos os pedidos avaliados.
    * Origem da informação: Registros de vendas para encontrar as datas de cada pedido e a data de entrega.
    * Apuração: Mensalmente ou trimestralmente.
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: um gráfico de barras ou um gráfico de linha que mostra a média do tempo de entrega ao longo do tempo, incluindo uma meta e destacando as variações significativas em relação à meta.
    * Resultado: Identificar áreas que precisam de melhorias e a tomar medidas corretivas para reduzir o prazo de entrega do pedido. Além disso, é importante acompanhar a distribuição do tempo de entrega dos pedidos, para entender se há um padrão específico de atrasos em determinados períodos ou em produtos específicos.
    
<br>    
    
3. Custo de transporte por unidade: mede o custo de transporte de cada unidade de produto. Um custo menor pode melhorar a lucratividade e a competitividade.
    * Cálculo: Custo de transporte por unidade = Custo total do transporte / Número total de unidades transportadas
        * Defina o período de tempo que deseja avaliar o custo de transporte por unidade. 
        * Identifique o custo total do transporte durante o período definido.
        * Identifique o número total de unidades transportadas durante o período definido.
        * Divida o custo total do transporte pelo número total de unidades transportadas para obter o custo de transporte por unidade.
    * Origem da informação: Registros de vendas para encontrar o custo total do transporte e o número total de unidades transportadas.
    * Apuração: Mensalmente ou trimestralmente.
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: um gráfico de barras ou um gráfico de linha que mostra o custo de transporte por unidade ao longo do tempo, incluindo uma meta e destacando as variações significativas em relação à meta.
    * Resultado: Identificar áreas que precisam de melhorias e a tomar medidas corretivas para reduzir o custo de transporte por unidade. Além disso, é importante acompanhar a distribuição do custo de transporte por unidade por produto, para entender se há um padrão específico de custos em determinados produtos ou rotas de transporte.
    
<br>    
    
4. Precisão do pedido: mede a porcentagem de pedidos enviados sem erros. Uma alta taxa de precisão de pedidos pode reduzir o retrabalho e as devoluções e melhorar a satisfação do cliente.
    * Cálculo: Precisão do pedido = (Número de pedidos enviados sem erros / Número total de pedidos enviados) x 100%
        * Defina o período de tempo que deseja avaliar a precisão do pedido. 
        * Identifique o número total de pedidos enviados durante o período definido.
        * Identifique o número de pedidos enviados sem erros durante o período definido.
        * Calcule a porcentagem de pedidos enviados sem erros em relação ao número total de pedidos enviados durante o período definido.
    * Origem da informação: Coletar informações do departamento de logística, de controle de qualidade e dos registros de vendas para identificar os pedidos enviados sem erros e o número total de pedidos enviados.
    * Apuração: Mensalmente ou trimestralmente
    * Polaridade: Positivo (Quanto maior o valor, melhor)
    * Visualização: um gráfico de barras ou um gráfico de pizza que mostra a porcentagem de pedidos enviados sem erros em relação ao número total de pedidos enviados durante o período definido, incluindo uma meta e destacando as variações significativas em relação à meta.
    * Resultado: Identificar áreas que precisam de melhorias e a tomar medidas corretivas para aumentar a precisão do pedido. Além disso, é importante acompanhar a distribuição dos erros por tipo de erro, para entender se há um padrão específico de erros em determinados produtos ou rotas de envio.
    
<br>    
    
5. Taxa de danos na remessa: mede a porcentagem de remessas danificadas durante o transporte. Uma baixa taxa de danos pode reduzir custos e melhorar a satisfação do cliente.
    * Cálculo: Taxa de danos na remessa = (Número de remessas danificadas / Número total de remessas enviadas) x 100%
        * Defina o período de tempo que deseja avaliar a taxa de danos na remessa. 
        * Identifique o número total de remessas enviadas durante o período definido.
        * Identifique o número de remessas danificadas durante o período definido.
        * Calcule a porcentagem de remessas danificadas em relação ao número total de remessas enviadas durante o período definido.
    * Origem da informação: Coletar informações do departamento de logística e de controle de qualidade para identificar as remessas danificadas e o número total de remessas enviadas durante o período definido.
    * Apuração: Mensalmente ou trimestralmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: um gráfico de barras ou um gráfico de pizza que mostra a porcentagem de remessas danificadas em relação ao número total de remessas enviadas durante o período definido, incluindo uma meta e destacando as variações significativas em relação à meta.
    * Resultado: Identificar áreas que precisam de melhorias e a tomar medidas corretivas para reduzir a taxa de danos na remessa. Além disso, é importante acompanhar a distribuição dos danos por tipo de dano e por rota de envio, para entender se há um padrão específico de danos em determinados produtos ou rotas de envio.
    
<br>    
    
6. Desempenho da transportadora: Mede o desempenho das transportadoras que uma empresa utiliza para transportar seus produtos. As métricas podem incluir desempenho pontual, tempo de trânsito e taxas de danos.
    * Cálculo:
        * Defina as métricas que deseja avaliar para medir o desempenho da transportadora. Por exemplo, você pode avaliar o desempenho pontual, o tempo de trânsito e as taxas de danos.
        * Identifique as transportadoras que a empresa utiliza para transportar seus produtos.
        * Para cada métrica escolhida, colete os dados relevantes para avaliar o desempenho de cada transportadora. Por exemplo, para o desempenho pontual, você pode coletar dados sobre a porcentagem de remessas entregues no prazo pela transportadora.
        * Calcule a média ponderada para cada métrica, levando em consideração o volume de remessas transportadas por cada transportadora.
        * Analise os resultados para identificar as transportadoras com melhor e pior desempenho e as áreas que precisam de melhoria.
    * Origem da informação: Coletar informações do departamento de logística e financeiro para identificar as transportadoras utilizadas e os custos associados a cada uma delas, além de dados de desempenho, como tempo de trânsito e porcentagem de remessas entregues no prazo.
    * Apuração: Trimestralmente
    * Polaridade: Positivo (Quanto maior o valor, melhor)
    * Visualização: um gráfico de barras ou um gráfico de radar que mostra o desempenho de cada transportadora em cada métrica avaliada. É importante destacar as transportadoras com melhor e pior desempenho em cada métrica e apresentar variações significativas em relação à média. 
    * Resultado: Identificar áreas que precisam de melhorias e a tomar medidas corretivas para melhorar o desempenho da transportadora e reduzir os custos de transporte. Além disso, é importante acompanhar a distribuição dos danos por tipo de dano e por rota de envio, para entender se há um padrão específico de danos em determinados produtos ou rotas de envio.
    
<br>    
    
7. Utilização do depósito: Mede a porcentagem do espaço do depósito que está sendo utilizado. Uma alta taxa de utilização pode indicar que o depósito está sendo usado de forma eficiente, enquanto uma baixa taxa pode indicar espaço desperdiçado e aumento de custos.
    * Cálculo: Utilização do depósito = (espaço de armazenamento utilizado / espaço total disponível) x 100
    * Origem da informação: Os dados podem ser obtidos por meio de um sistema de gerenciamento de armazém (WMS), que rastreia a movimentação de estoque e a utilização de espaço de armazenamento. 
    * Apuração: Mensalmente ou trimestralmente
    * Polaridade: Positivo (Quanto maior o valor, melhor)
    * Visualização: um gráfico de barras horizontais que mostre a porcentagem de utilização do depósito, com a linha de base marcando 100%. 

<br>

#### <u>Principais problema de negócio:</u>
1. Falta de planejamento: Uma falha no planejamento pode resultar em atrasos na entrega, excesso de estoque ou falta de estoque, o que pode levar a insatisfação do cliente e aumento de custos.
2. Ineficiência no gerenciamento de estoque: O estoque é um dos principais ativos de uma empresa e gerenciá-lo de forma eficiente é fundamental para manter a saúde financeira da empresa. Problemas como excesso de estoque, falta de estoque ou obsolescência de produtos podem levar a perdas financeiras significativas.
3. Problemas de comunicação: Uma comunicação ineficiente entre os diferentes departamentos ou com os clientes pode levar a atrasos na entrega, problemas de qualidade e insatisfação do cliente.
4. Falhas na gestão de transportes: A escolha inadequada de transportadores, rotas ineficientes, atrasos e perdas de carga podem levar a custos elevados e insatisfação do cliente.
5. Problemas de qualidade: Falhas na qualidade dos produtos ou na embalagem podem levar a danos durante o transporte, o que pode resultar em devoluções e insatisfação do cliente.
6. Falta de treinamento: A falta de treinamento adequado para os funcionários pode levar a erros na gestão do estoque, falhas na comunicação e falta de eficiência na gestão da logística.
7. Falta de tecnologia: A falta de investimento em tecnologia, como sistemas de gerenciamento de armazém ou de transporte, pode levar a uma gestão ineficiente da logística.

## Jurídico
[&#x1F53C;](#Sumário:) Clique para voltar para o sumário

#### <u>Indicadores:</u>
1. Taxa de resolução de casos: mede a porcentagem de casos que foram resolvidos em um período específico. Isso pode indicar a eficiência do departamento jurídico em lidar com casos e resolver problemas rapidamente.
    * Cálculo: Dividir o número de casos resolvidos pelo número total de casos no período e multiplicar o resultado por 100 para obter a porcentagem.
    * Origem da informação: Os dados para esse KPI podem ser obtidos por meio de sistemas de gestão de casos ou relatórios internos do departamento jurídico.
    * Apuração: Mensalmente ou trimestralmente
    * Polaridade: Positivo (Quanto maior o valor, melhor)
    * Visualização: um gráfico de linha que acompanha a evolução da taxa de resolução de casos ao longo do tempo, permitindo avaliar a eficiência do departamento em lidar com diferentes volumes de trabalho. 

<br>

2. Tempo médio de resolução de casos: mede o tempo médio necessário para resolver um caso. Isso pode indicar a eficiência do departamento jurídico em lidar com casos e resolver problemas rapidamente.
    * Cálculo: É necessário obter a diferença entre a data em que o caso foi aberto e a data em que foi resolvido. Em seguida, é possível calcular a média dos tempos de resolução de todos os casos no período desejado.
    * Origem da informação: No sistema de gestão de casos do departamento jurídico, que registra a data de abertura e a data de fechamento de cada caso.
    * Apuração: Mensalmente ou trimestralmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: um gráfico de barras ou de linhas, que apresente o tempo médio de resolução de casos ao longo do tempo. Além disso, pode ser interessante apresentar também o tempo médio de resolução por tipo de caso ou por advogado, para identificar possíveis gargalos no processo de resolução de casos.

<br>

3. Número de processos judiciais: mede o número de processos judiciais em que a empresa está envolvida. Isso pode indicar a eficácia das políticas e práticas da empresa em prevenir disputas legais.
    * Cálculo: Contar o número de processos judiciais em que a empresa está envolvida dentro do período de apuração.
    * Origem da informação: Esses dados podem ser obtidos por meio do departamento jurídico da empresa ou por meio de sistemas de gerenciamento de casos judiciais. 
    * Apuração: Anualmente ou trimestralmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: um gráfico de barras ou um gráfico de linhas que exibe o número de processos judiciais ao longo do tempo.
    * Resultado: é útil destacar as principais causas dos processos e quais são os departamentos mais afetados, a fim de ajudar a empresa a identificar áreas em que precisa melhorar para evitar futuras disputas legais.

<br>

4. Custo legal: mede o custo total de todas as atividades jurídicas realizadas pela empresa em um período específico. Isso pode incluir honorários advocatícios, custos de litígio, custos de mediação e arbitragem e outros custos relacionados.
    * Cálculo: somar todos os custos relacionados às atividades jurídicas da empresa durante um período específico e dividir pelo número de casos ou pelo número de horas de trabalho de advogados da empresa durante esse período.
    * Origem da informação: registros financeiros da empresa, na área de contabilidade e finanças, bem como nos relatórios e planilhas de acompanhamento de atividades jurídicas do departamento jurídico.
    * Apuração: mensal, trimestral ou anual, dependendo da necessidade da empresa.
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: um gráfico de linhas ou barras que apresente o valor total de custo legal em cada período de apuração. 
    * Resultado: a comparação dos valores com metas e orçamentos estabelecidos pela empresa pode ajudar a identificar possíveis desvios e oportunidades de melhoria.
    
<br>    

5. Risco legal: avalia o risco potencial de litígio ou outros problemas legais que possam afetar a empresa. Isso pode incluir riscos relacionados a contratos, propriedade intelectual, questões trabalhistas e outras áreas de preocupação.
    * Cálculo: Pode variar de acordo com a metodologia escolhida pela empresa, mas uma forma comum é atribuir uma pontuação para cada área de preocupação, com base na probabilidade de ocorrência e no impacto financeiro e de reputação que o problema pode causar. 
        * Por exemplo, a probabilidade de uma disputa de propriedade intelectual pode ser classificada em uma escala de 1 a 5, sendo 1 muito improvável e 5 muito provável, enquanto o impacto financeiro e de reputação pode ser classificado em uma escala de 1 a 10, sendo 1 baixo impacto e 10 alto impacto. A pontuação é então calculada multiplicando a probabilidade pelo impacto, e a pontuação total é somada para fornecer uma avaliação geral do risco legal da empresa.
    * Origem da informação: Podem ser obtidos por meio de uma revisão interna das áreas de risco legal, por meio de relatórios de auditoria ou por meio de consultas com especialistas externos. 
    * Apuração: trimestral ou anual, dependendo da complexidade dos negócios da empresa.
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: um gráfico de barras ou um painel de controle, que mostra a pontuação de risco legal para cada área de preocupação e a pontuação total. O painel de controle pode ser configurado para fornecer alertas quando a pontuação total atingir um nível crítico, permitindo que a empresa tome medidas proativas para gerenciar o risco legal.

<br>

6. Satisfação do cliente interno: mede a satisfação dos clientes internos (ou seja, outros departamentos e funcionários da empresa) com os serviços jurídicos prestados. Isso pode ajudar a identificar áreas de melhoria e garantir que as necessidades da empresa sejam atendidas de forma eficaz.
    * Cálculo: é necessário realizar pesquisas ou enquetes com os departamentos e funcionários da empresa que utilizam os serviços jurídicos. As perguntas devem estar relacionadas ao atendimento, qualidade dos serviços prestados, prazos, efetividade e outras áreas de interesse. Os resultados das pesquisas podem ser expressos em uma escala numérica ou porcentagem, onde valores mais altos indicam maior satisfação.
    * Origem da informação: Os dados podem ser buscados por meio de ferramentas de pesquisa ou de feedback dos usuários dos serviços jurídicos. 
    * Apuração: anualmente ou a cada seis meses para acompanhar as mudanças ao longo do tempo.
    * Polaridade: Positivo (Quanto maior o valor, melhor)
    * Visualização: um gráfico de barras ou um gráfico de linha mostrando a evolução da satisfação do cliente interno ao longo do tempo.

<br>

#### <u>Principais problema de negócio:</u>
1. Excesso de demanda: o departamento jurídico muitas vezes é responsável por lidar com uma grande quantidade de demandas e processos judiciais, o que pode levar a uma sobrecarga de trabalho e atrasos na resposta a solicitações e prazos.
2. Prazos apertados: muitas vezes, os prazos para apresentação de defesas, recursos e outros documentos legais são curtos, o que pode gerar pressão e estresse para os profissionais do departamento jurídico.
3. Falta de recursos: alguns departamentos jurídicos podem ter orçamentos limitados, o que pode impedir a contratação de profissionais qualificados e o investimento em tecnologia e ferramentas necessárias para o desempenho das atividades.
4. Conformidade regulatória: as empresas precisam se adequar a uma série de leis e regulamentações, o que pode ser um desafio para o departamento jurídico, especialmente em setores altamente regulamentados.
5. Conflitos de interesse: os advogados do departamento jurídico precisam garantir que suas atividades estejam em conformidade com o código de ética da profissão, o que pode incluir evitar conflitos de interesse e garantir a confidencialidade das informações dos clientes da empresa.
6. Comunicação ineficiente: uma comunicação ineficiente entre os membros do departamento jurídico e outras áreas da empresa pode levar a problemas na gestão de riscos e no cumprimento de normas e regulamentações.

## Estoque / Armazém
[&#x1F53C;](#Sumário:) Clique para voltar para o sumário

#### <u>Indicadores:</u>
1. Taxa de rotatividade de estoque: mede quantas vezes o estoque de uma empresa é vendido e substituído em um período específico. Um alto índice de rotatividade de estoque indica uma gestão de estoque eficiente.
    * Cálculo: (Custo dos produtos vendidos / Valor médio do estoque) x 12 (para representar o número de vezes que o estoque é vendido em um ano)
    * Origem da informação: obtidos do sistema de gestão de estoque ou armazém da empresa, onde são registrados os custos dos produtos vendidos e o valor médio do estoque.
    * Apuração: mensalmente ou trimestralmente
    * Polaridade: Positivo (Quanto maior o valor, melhor) = Quanto maior a taxa de rotatividade de estoque, mais rápido os produtos estão sendo vendidos e maior é o lucro da empresa.
    * Visualização: um gráfico de linhas, onde é possível acompanhar a evolução da taxa de rotatividade de estoque ao longo do tempo.
    
<br>    
    
2. Custo de manutenção do estoque / Custo por unidade armazenada: é o custo associado à manutenção do estoque, incluindo custos de armazenamento, seguro e obsolescência. O tempo de permanência em armazém, a dimensão das mercadorias, o seu peso e outras características que obrigam a condições especiais de armazenagem e movimentação são geralmente os fatores que contribuem para a determinação do custo em armazém.
    * Cálculo: Gastos com o armazém / de produtos em armazém (Real)
    * Origem da informação: Armazém
    * Apuração: Mensalmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: Gráfico de Colunas com linha de média (Real Custos por unidade armazenada x Tempo em meses)
    * Resultado: Este indicador pode ainda ser determinado por tipo de mercadoria, por armazém e por cliente. A sua análise deve permitir identificar eventuais custos excessivos e proceder à racionalização do armazenamento de acordo com as necessidades permanentes e evitando acumulação de excesso de stock.
    
<br>    
    
3. Margem bruta de retorno sobre o investimento (GMROI): mede a lucratividade do estoque comparando o lucro bruto obtido com o investimento no estoque. Um GMROI mais alto indica que a empresa está lucrando mais com seu estoque.
    * Cálculo: dividindo-se a margem bruta pelo custo médio dos estoques. A margem bruta é a diferença entre o preço de venda e o custo dos produtos vendidos. O custo médio dos estoques é o valor médio dos produtos mantidos em estoque durante um determinado período de tempo.
    * Origem da informação: dados financeiros, como vendas, custo dos produtos vendidos e valor médio dos estoques. Esses dados podem ser encontrados no sistema de gestão do armazém, no sistema de contabilidade ou no sistema de gestão financeira da empresa.
    * Apuração: mensalmente ou trimestralmente
    * Polaridade: Positivo (Quanto maior o valor, melhor) = indica a capacidade do departamento de estoque/armazém de gerar lucros em relação ao investimento em estoque.
    * Visualização: um gráfico de linhas ou um gráfico de barras.
    
<br>    
    
4. Taxa de falta de estoque: mede a porcentagem de tempo em que o estoque de uma empresa fica sem estoque. Uma alta taxa de falta de estoque indica uma má gestão de estoque, o que pode levar à perda de vendas e à insatisfação do cliente.
    * Cálculo: número de vezes que ocorreu uma falta de estoque em relação ao número total de vezes que um item deveria ter sido disponibilizado no estoque durante um período de tempo específico.
    * Origem da informação:  podem ser obtidos a partir dos registros de estoque e das informações de compras e vendas.
    * Apuração: mensal ou trimestral.
    * Polaridade: Negativa (Quanto menor o valor, melhor) = já que a falta de estoque pode levar a atrasos nas entregas e perda de vendas, afetando negativamente a receita da empresa.
    * Visualização: um gráfico de barras, que permite comparar o número de vezes que ocorreu falta de estoque em relação ao número de vezes que um item deveria ter sido disponibilizado no estoque durante um período de tempo específico.
    
<br>    
    
5. Lead time: é o tempo que uma empresa leva para receber o estoque após fazer um pedido. Prazos de entrega mais curtos podem ajudar uma empresa a responder mais rapidamente às mudanças na demanda e evitar a falta de estoque.
    * Cálculo: mede o tempo necessário para que um pedido seja processado desde o momento em que é recebido até o momento em que é entregue ao cliente.
    * Origem da informação: podem ser coletados a partir de sistemas de gerenciamento de estoque e vendas
    * Apuração: mensal ou trimestral
    * Polaridade: Negativa (Quanto menor o valor, melhor) = quanto maior for o tempo de lead time, pior será a eficiência da empresa em processar pedidos e entregar produtos aos clientes
    * Visualização: um gráfico de linha ou de barras, que permite uma fácil comparação entre os tempos de lead time em diferentes períodos.
    
<br>    
    
6. Tempo de ciclo do pedido: é o tempo necessário para processar e atender um pedido do cliente. Tempos de ciclo de pedidos mais curtos podem ajudar uma empresa a melhorar a satisfação do cliente e aumentar as vendas. 
    * Cálculo: Σ (Data do pedido de encomenda – Data da expedição da encomenda) / de expedições (Dias ou horas)
    * Origem da informação: Armazém
    * Apuração: Mensalmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: Gráfico de Colunas com linhas da média e média do setor (Tempo média de execução x Tempo em meses)
    * Resultado: Quanto menor for o tempo médio de expedição, maior será a eficiência interna e melhor será vista a empresa pelos seus clientes. Em alguns setores de atividade, pode ser considerado um fator crítico de competitividade, uma vez que existem clientes que consideram fundamental poderem ter acesso rapidamente aos produtos que adquirem.
    
<br>    
    
7. Dias de estoque disponível: mede quantos dias de vendas uma empresa tem em estoque. Um número mais alto indica que uma empresa tem excesso de estoque e pode correr o risco de incorrer em custos de manutenção ou obsolescência.
    * Cálculo: dividindo-se o valor total do inventário pela média diária de vendas, ou seja, a quantidade total de estoque disponível dividido pela quantidade média de vendas diárias.
    * Origem da informação: obter o valor total do inventário e a quantidade média de vendas diárias. Essas informações podem ser obtidas diretamente do sistema de gerenciamento de estoque da empresa.
    * Apuração: mensalmente
    * Polaridade: Positivo (Quanto maior o valor, melhor) = Quanto maior o número de dias de estoque disponível, melhor será a capacidade da empresa em atender a demanda dos clientes e reduzir o risco de falta de estoque.
    * Visualização: um gráfico de barras ou um gráfico de linha que mostra a evolução do número de dias de estoque disponível ao longo do tempo.

<br>

8. Porcentagem da capacidade de armazém utilizada: Conhecer o grau de rentabilização do espaço do armazém é fundamental para perceber se existe eficiência na sua utilização. Conhecendo o valor, a empresa está em condições de decidir sobre se o espaço que utiliza é excessivo ou insuficiente para a atividade que desenvolve. Quando não existe um layout funcional na disponibilização das existências em armazém, originam-se sempre situações de desperdício de espaço. Para além de conhecer a capacidade de armazém utilizada atualmente, é também importante saber qual foi a sua média desde o início do ano.
    * Cálculo: Área utilizada (m3) / Área total do armazém (m3) (%)
    * Origem da informação: Armazém
    * Apuração: Mensalmente
    * Polaridade: Positivo (Quanto maior o valor, melhor)
    * Visualização: Gráfico de coluna empilhada (Área utilizada - m³ x Área disponível – m³ x Tempo em meses)
    
<br>

9. Porcentagem nas vendas dos gastos com armazém: Este indicador pretende identificar o peso dos gastos com o armazenamento das mercadorias ou matérias-primas nas vendas da empresa. É um indicador que incentiva a racionalização dos gastos. Importa calcular este indicador em termos acumulados, de forma a dispersar uma eventual sazonalidade das vendas. Existindo uma componente de gastos fixos, o valor deste indicador irá aumentar nos momentos em que as vendas sejam reduzidas.
    * Cálculo: Gastos com armazém / Volume de vendas (%)
    * Origem da informação: Armazém
    * Apuração: Trimestralmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: Gráfico de Linha (% Peso nas vendas dos gastos com armazém x Tempo em trimestre)

<br>

10. Tempo médio de permanência das existências em armazém: Este indicador mede o tempo médio que as existências ficam no armazém. É de extrema utilidade para analisar quais as existências que ficam mais tempo no armazém e as que ficam menos. Pode ainda ser calculado para produtos acabados e mercadorias. É um indicador importante para efeitos de análise do ciclo de exploração.
    * Cálculo: (Existências Médias de Matérias-primas/Custo das Matérias Consumidas) × 365 (Dias)
    * Origem da informação: Armazém
    * Apuração: Mensalmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: Gráfico de Colunas com linha média (Tempo médio de permanência em dias x Tempo em meses)

<br>

11. Número de artigos obsoletos: são aqueles que já não cumprem as condições essenciais para continuar a ser comercializados pela empresa. Por exemplo: um artigo que caiu em desuso. Poderão ainda ter algum valor, mas não serão vendidos nas mesmas condições do inventário normal.
    * Cálculo: Σ do de artigos obsoletos (N°)
    * Origem da informação: Armazém
    * Apuração: Mensalmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: Gráfico de coluna (N° de artigos obsoletos x Tempo em meses)
    * Resultado: O aumento do de artigos obsoletos é um dos sintomas de má gestão do inventário. Para diminuir o prejuízo dos artigos obsoletos, devem ser efetuadas promoções para escoar esses inventários. No setor das tecnologias, o risco de os produtos ficarem obsoletos é muito elevado, já que o período de tempo de uso é geralmente curto (exemplo: computador).
    
<br>

12. Porcentagem de devoluções de mercadorias não aceites no destino pelos clientes: Indica a taxa de mercadorias devolvidas pelo cliente. Permite assim avaliar o nível de insatisfação do cliente e a incapacidade de detectar antecipadamente inconformidades nas mercadorias a expedir (% de inconformidades detectadas na expedição das mercadorias/produtos). É um indicador de qualidade de serviço.  
    * Cálculo: N° de encomendas devolvidas/ de encomendas expedidas (%)
    * Origem da informação: Armazém
    * Apuração: Mensalmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: Gráfico de Colunas com eixo secundário (% de devoluções x N° de encomendas devolvidas x N° de encomendas expedidas x Tempo em meses)
    * Resultado: Em períodos de tempo em que existem muitas encomendas, existirá uma probabilidade maior das encomendas serem devolvidas pelos clientes devido a inconformidades nos produtos ou da encomenda. Caberá à empresa reforçar a verificação das encomendas e dos produtos. 

<br>

13. Porcentagem de inconformidades detectadas na expedição das mercadorias/produtos: Identifica o peso das inconformidades detectadas na fase de distribuição. Permite avaliar a qualidade do processo de distribuição e a ocorrência de inconformidades, de forma a alertar para possíveis melhorias. É um indicador que mede a qualidade do serviço de expedição de mercadorias. As inconformidades no envio podem ser de vários tipos. Por exemplo, mercadorias não solicitadas, quantidades de mercadoria diferentes das encomendadas, mercadorias com defeitos, etc.
    * Cálculo: Número de encomendas com inconformidades / Número de encomendas expedidas (%)
    * Origem da informação: Armazém
    * Apuração: Mensalmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: Gráfico de Colunas com eixo secundário (% de inconformidades x N° de encomendas com inconformidades x N° de encomendas expedidas x Tempo em meses)

<br>

14. Porcentagem de inconformidades detectadas na aceitação das mercadorias/produtos: Identifica o peso das mercadorias aceites pelo armazém que não estavam de acordo com o pedido de encomenda. No ato de recepção de mercadorias, verifica-se a conformidade do documento de guia de remessa ou de transporte que acompanha a mercadoria/produtos, de forma a detectar as inconformidades. Este indicador avalia a qualidade da atividade na recepção de mercadorias. As inconformidades na recepção podem ser de vários tipos. Por exemplo, mercadorias não solicitadas, quantidades de mercadoria diferentes das encomendadas, mercadorias com defeitos, etc.
    * Cálculo: Número de encomendas com inconformidades / Número de encomendas recepcionadas (%)
    * Origem da informação: Armazém
    * Apuração: Mensalmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: Gráfico de Colunas com eixo secundário (% de inconformidade x N° de encomendas com inconformidades x N° de encomendas recepcionadas x Tempo em meses)

<br>

15. Porcentagem de encomendas entregues em tempo: Nos dias de hoje, os clientes são cada vez mais exigentes com o cumprimento dos prazos de entrega estabelecidos. A obtenção de valores elevados neste indicador demonstra que os processos internos estão a funcionar de forma eficiente e é claramente um sinal da qualidade de serviço prestada pela empresa.
    * Cálculo: N° de encomendas entregues em tempo / total de encomendas (%)
    * Origem da informação: Armazém
    * Apuração: Mensalmente
    * Polaridade: Positivo (Quanto maior o valor, melhor)
    * Visualização: Gráfico de colunas empilhadas com eixo secundário (% Encomendas entregues x N° entregues em tempo x N° Total de encomendas x Tempo em meses)
    * Resultado: Em períodos de tempo em que existem muitas encomendas, existirá uma probabilidade maior de as encomendas não serem entregues dentro dos prazos estabelecidos. Caberá à empresa gerir as expectativas dos seus clientes, de modo a não prejudicar a sua imagem e eventualmente afetar futuras encomendas.
    
<br>

#### <u>Principais problema de negócio:</u>
1. Precisão do estoque: o departamento de estoque deve garantir que os níveis de estoque sejam precisos e atualizados. Isso pode envolver a implementação de sistemas de rastreamento de estoque, realização de contagens físicas regulares de estoque e reconciliação de discrepâncias de estoque.
2. Falta de estoque e excesso de estoque: o departamento de estoque deve equilibrar os níveis de estoque para evitar falta de estoque, que pode levar à perda de vendas e insatisfação do cliente, e excesso de estoque, que pode comprometer o capital e aumentar os custos de armazenamento.
3. Previsão de demanda: o departamento de estoque deve prever a demanda de produtos acabados, matérias-primas e suprimentos para garantir que os níveis de estoque apropriados sejam mantidos. Isso pode envolver o uso de dados históricos de vendas, análise de tendências de mercado e colaboração com outros departamentos, como vendas e marketing.
4. Interrupções na cadeia de suprimentos: o departamento de estoque pode contar com uma rede complexa de fornecedores e vendedores para entregar matérias-primas e suprimentos. Interrupções na cadeia de suprimentos, como atrasos ou problemas de qualidade, podem afetar os níveis de estoque e a satisfação do cliente.
5. Estoque obsoleto: o departamento de estoque deve gerenciar o estoque obsoleto, como produtos descontinuados ou matérias-primas vencidas, para evitar a imobilização de capital e espaço de armazenamento.
6. Gerenciamento de custos: o departamento de estoque deve gerenciar os custos de estoque, como armazenamento, transporte e custos de manuseio, para garantir que o estoque seja gerenciado de maneira eficiente e econômica.
7. Relatórios e análises: o departamento de estoque deve fornecer relatórios precisos e oportunos sobre os níveis de estoque, custos e tendências. Isso pode envolver o uso de ferramentas de análise de dados para identificar oportunidades de melhorias de processo, economia de custos e otimização de estoque.
8. Restrições de espaço: O departamento de armazém deve garantir que tem espaço suficiente para armazenar mercadorias de forma eficiente e segura. As restrições de espaço podem limitar a quantidade de estoque que pode ser armazenado, levando a atrasos e ineficiências.
9. Rotação de estoque: O departamento de almoxarifado deve gerenciar a rotação de estoque para garantir que as mercadorias sejam armazenadas e distribuídas de maneira eficiente e segura. Isso pode envolver a implementação de um sistema de primeiro a entrar, primeiro a sair (FIFO) ou primeiro a expirar, primeiro a sair (FEFO) para garantir que o estoque mais antigo seja usado ou vendido antes do estoque mais novo.
10. Gerenciamento de mão de obra: O departamento de depósito deve gerenciar sua força de trabalho com eficiência para garantir que as mercadorias sejam recebidas, armazenadas e distribuídas de maneira oportuna e econômica. Isso pode envolver escalar a equipe adequadamente, treinar a equipe em procedimentos de segurança e manuseio e fornecer incentivos para incentivar a produtividade.
11. Manutenção de equipamentos: O departamento de almoxarifado deve manter seus equipamentos, como empilhadeiras e esteiras transportadoras, para garantir que operem com eficiência e segurança. Isso pode envolver o agendamento de manutenção de rotina, reparo de equipamentos conforme necessário e substituição de equipamentos quando necessário.
12. Segurança e conformidade regulamentar: O departamento de armazém deve cumprir as normas regulamentares e de segurança para proteger funcionários e clientes e evitar responsabilidades legais. Isso pode envolver o estabelecimento de procedimentos de segurança, treinamento de equipe em protocolos de segurança e garantia de que equipamentos e processos estejam em conformidade com os padrões regulamentares.
13. Melhoria contínua: O departamento de almoxarifado deve melhorar continuamente seus processos para aumentar a eficiência, reduzir o desperdício e melhorar a segurança. Isso pode envolver a implementação de princípios enxutos, a adoção de novas tecnologias ou automação ou o uso de análise de dados para identificar áreas de melhoria.

## Ambiente

#### <u>Indicadores:</u>
1. Consumo de energia: Este KPI mede a quantidade de energia utilizada por uma empresa e pode ajudar a identificar oportunidades de melhorias de eficiência energética. O somatório de todas as fontes de energia utilizadas na produção é convertido em toneladas equivalente de petróleo (tep).
    * Cálculo: Consumo de energia (tep) / Volume de produção (tonelada) (tep)
    * Origem da informação: Direção de produção 
    * Apuração: Mensalmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: Gráfico de coluna (Consumo especifico de energia (tep) x Tempo em meses)
    * Resultado: Para a energia elétrica, considerou-se que 1kWh = 215 × 10-6 tep. Gás natural = 1,077; Gás petróleo liquefeito = 1.130; Fuel óleo = 0,965; Gasóleo = 1,034 e Gasolina = 1,075 (tep / tonelada).
    
<br>    
    
2. Pegada de carbono: Este KPI mede a quantidade total de emissões de gases de efeito estufa (geralmente medido em equivalente de CO2) gerado pelas atividades de uma empresa. As empresas podem usar esse KPI para acompanhar o progresso em direção às metas de redução de emissões e identificar áreas de melhoria.
    * Cálculo: Para calcular a pegada de carbono, a empresa deve identificar as principais fontes de emissão de gases de efeito estufa, que podem incluir transporte, uso de energia, produção de resíduos, entre outros. A partir daí, é possível estimar a quantidade de emissões associadas a cada atividade por meio de fatores de emissão (por exemplo, a quantidade de dióxido de carbono emitido por litro de combustível). 
    * Origem da informação: Os dados podem ser coletados por meio de medições diretas, registros de consumo de energia e combustível, ou outras fontes relevantes.
    * Apuração: anual ou trimestral, dependendo da frequência de coleta de dados e dos objetivos da empresa.
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: um gráfico de linha ou de barras, mostrando a evolução das emissões ao longo do tempo e as metas de redução estabelecidas pela empresa. Também pode ser útil apresentar os resultados em termos absolutos e relativos (por exemplo, a pegada de carbono por unidade produzida), para facilitar a comparação entre diferentes períodos e setores da empresa.
    
<br>    
    
3. Consumo específico de água: Este KPI mede a quantidade de água usada por uma empresa e pode ajudar a identificar oportunidades de conservação de água e melhorias de eficiência. Este indicador apresenta o consumo de água por tonelada produzida de produto. Também pode ser apresentado de forma a indicar os m3/litros de água necessários para a produção de uma peça. É acima de tudo um indicador para medir a eficiência da produção.
    * Cálculo: Consumo de água (m³) / Volume de produção (tonelada) (m³)
    * Origem da informação: Direção de produção 
    * Apuração: Mensalmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: Gráfico de coluna (Consumo especifico de água - m³ x Tempo em meses)
    
<br>    
    
4. Geração de resíduos: Este KPI mede a quantidade de resíduos gerados por uma empresa e pode ajudar a identificar oportunidades de redução, reutilização e reciclagem de resíduos. Este indicador apresenta o total em kg de resíduos sólidos gerados no processo produtivo da empresa. Resíduos são substâncias, produtos ou objetos que ficaram incapazes de utilização para os fins para que foram produzidos, ou são restos de um processo de produção, transformação ou utilização.
    * Cálculo: Quantidade de resíduos sólidos (kg) / Volume de produção (tonelada) (kg)
    * Origem da informação: Direção de produção 
    * Apuração: Mensalmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: Gráfico de colunas (Kg Quantidade especifica de resíduos sólidos x Tempo em meses)
    
<br>    
    
5. Incidentes ambientais: Este KPI mede o número de incidentes ambientais (como derramamentos ou vazamentos) que ocorrem nas operações de uma empresa. O rastreamento dessa métrica pode ajudar as empresas a identificar áreas onde podem existir riscos ambientais e implementar medidas para evitar futuros incidentes.
    * Cálculo: Para calcular o KPI de Incidentes Ambientais, é necessário primeiro definir o que é considerado um incidente ambiental e como os dados serão coletados e categorizados. Em geral, um incidente ambiental pode ser definido como um evento não planejado que cause danos ou riscos ao meio ambiente, como derramamentos de produtos químicos, vazamentos de óleo, contaminação da água, etc.
    * Origem da informação: Para coletar os dados, as empresas devem estabelecer procedimentos claros para relatar e registrar incidentes ambientais, geralmente por meio de um sistema de gerenciamento ambiental. Os dados podem ser coletados por local, por tipo de incidente, por frequência ou outras categorias relevantes.
    * Apuração: Mensalmente ou trimestralmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: um gráfico de barras ou um gráfico de linhas que mostra o número de incidentes ambientais ao longo do tempo, categorizados por tipo de incidente ou local.
    
<br>    
    
6. Conformidade ambiental: Este KPI mede a conformidade de uma empresa com as regulamentações e licenças ambientais. Garantir a conformidade pode ajudar as empresas a evitar penalidades e danos à reputação.
    * Cálculo: Para calcular a empresa precisa monitorar o cumprimento das regulamentações e licenças ambientais que se aplicam a suas atividades. Isso pode incluir leis e regulamentos locais, estaduais, nacionais e internacionais. 
    * Origem da informação: Os dados podem ser obtidos através de auditorias internas e externas, inspeções regulatórias, relatórios de conformidade e outras fontes.
    * Apuração: pode variar de acordo com a frequência das auditorias e inspeções regulatórias, mas é comum que seja medido trimestral ou anualmente.
    * Polaridade: Vai depender do cumprimento das regulamentações e licenças ambientais 
    * Visualização: um gráfico de barras ou pizza, apresentando a porcentagem de conformidade alcançada em relação às regulamentações e licenças aplicáveis. Também pode ser útil mostrar tendências ao longo do tempo, para identificar áreas que precisam de melhoria contínua.
    
<br>    
    
7. Sustentabilidade da cadeia de suprimentos: Este KPI mede o impacto ambiental da cadeia de suprimentos de uma empresa, incluindo seus fornecedores e parceiros logísticos. As empresas podem usar essa métrica para identificar oportunidades de reduzir o impacto ambiental de suas operações e cadeia de suprimentos.
    * Cálculo: a empresa precisa rastrear as emissões de gases de efeito estufa (geralmente medido em equivalente de CO2) e outros impactos ambientais, como o uso de água e energia, em toda a cadeia de suprimentos. 
    * Origem da informação: por meio de questionários enviados aos fornecedores ou por meio de sistemas de gerenciamento da cadeia de suprimentos.
    * Apuração: pode variar dependendo da frequência de coleta de dados. No entanto, é recomendável que a empresa colete dados regularmente para identificar tendências e oportunidades de melhoria ao longo do tempo.
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: um gráfico de barras ou um gráfico de linhas que mostre as emissões ou impactos ambientais totais ao longo do tempo, segmentados por fornecedor ou categoria de produto. 
    * Resultado: Permitirá que a empresa identifique os fornecedores ou áreas da cadeia de suprimentos com maior impacto ambiental e implemente medidas para reduzir esses impactos.
    
<br>

8. Carga específica de efluente gasoso: Este indicador apresenta a quantidade de poluente × (por exemplo: CO; NOx; CO2; Partículas; COV; etc.) emitida para a atmosfera por tonelada de produção. As emissões na atmosfera têm uma especial importância devido a diversos impactos ambientais e na saúde humana.
    * Cálculo: Quantidade de poluente × (Kg)/Volume de produção (kg)
    * Origem da informação: Direção de produção 
    * Apuração: Mensalmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: Gráfico de coluna (Kg carga especifica de efluentes gasosos x Tempo em meses)

<br>

9. % de reciclagem de água: Este indicador apresenta a relação entre a água recuperada e a água consumida no processo produtivo. Em princípio, quanto maior a taxa de reciclagem da água maior é a eficiência produtiva da empresa e menor será o impacto em termos ambientais.
    * Cálculo: Água recuperada (m3) / Consumo de água (m3) (%)
    * Origem da informação: Direção de produção 
    * Apuração: Mensalmente
    * Polaridade: Positivo (Quanto maior o valor, melhor)
    * Visualização: Gráfico de Coluna empilhado com eixo secundário (% Taxa de reciclagem da água x Água recuperada em m³ x Consumo de água em m³ x Tempo em meses)
    
<br>

10. Taxa de papel reciclado: Este indicador expressa-se em kg ou tonelada e apresenta a relação entre o papel enviado para reciclagem e o papel adquirido pela empresa. Também se costuma utilizar este indicador com uma variante, de modo a perceber o peso do papel reciclado adquirido pela empresa.
    * Cálculo: Papel enviado para reciclagem (Kg)/Papel comprado (Kg) (%)
    * Origem da informação: Direção de produção 
    * Apuração: Mensalmente
    * Polaridade: Positivo (Quanto maior o valor, melhor)
    * Visualização: 2 Gráficos de coluna empilhada (Kg de papel reciclado x Kg papel adquirido x Tempo em meses) e Gráfico de linha (% de papel reciclado x Tempo em meses)
    * Resultado: Pode acontecer que este indicador apresente um valor superior a 100%. Nestes casos, a quantidade de papel reciclado foi superior ao adquirido, uma vez que a reciclagem incorporou papel não adquirido pela empresa, mas que teve origem em outras fontes da empresa, por exemplo, papel com origem nos documentos vindos do exterior. Importa acompanhar os indicadores em termos acumulados.

<br>

11. Quantidade de resíduos valorizados: Este indicador apresenta a quantidade de resíduos sólidos cujo destino é a valorização através da reutilização, reciclagem ou incineração em instalações de incineração de resíduos com valorização energética.
    * Cálculo: Σ da quantidade de resíduos valorizados (Tonelada)
    * Origem da informação: Direção de produção 
    * Apuração: Mensalmente
    * Polaridade: Positivo (Quanto maior o valor, melhor)
    * Visualização: 2 Gráficos de coluna (Kg resíduos valorizados x Tempo em meses) e Gráfico de linha (% Taxa de valorização de resíduos x Tempo em meses)
    * Resultado: Por oposição a este indicador, existe a quantidade de resíduos eliminados cujo destino é a deposição em aterro ou tratamento para posterior eliminação. Para se perceber a dimensão da valorização, é comum utilizar-se o indicador “taxa de valorização” que permite perceber o peso dos resíduos sólidos valorizados relativamente ao total de resíduos produzidos.

<br>

12. Taxa  de resíduos perigosos: Este indicador apresenta o peso dos resíduos perigosos relativamente ao total de resíduos gerados na empresa. Os resíduos perigosos podem ser classificados em função do seu carácter tóxico, corrosivo, explosivo, radioativo, etc.
    * Cálculo: Quantidade de resíduos perigosos (tonelada) / Quantidade de resíduos sólidos (%)
    * Origem da informação: Direção de produção 
    * Apuração: Mensalmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: 2 Gráficos de coluna (Ton de resíduos perigosos x Tempo em meses) e Gráfico de linha (% Taxa de resíduos perigosos x Tempo em meses)
    * Resultado: Uma das formas de minimizar o valor deste indicador passa pela conhecida sigla dos 3R’s: Reduzir, reutilizar e reciclar
    
<br>

13. % de fornecedores avaliados ambientalmente: Este indicador apresenta o peso dos fornecedores que foram avaliados ambientalmente mediante avaliação do seu sistema de gestão da qualidade e/ou ambiental. Desta forma, a empresa impõe padrões de referência a todos os fornecedores que pretendam fornecer a empresa.
    * Cálculo: Número de fornecedores avaliados ambientalmente / total de fornecedores (%)
    * Origem da informação: Direção de produção 
    * Apuração: Mensalmente
    * Polaridade: Positivo (Quanto maior o valor, melhor)
    * Visualização: 2 Gráficos de coluna empilhada (N° de fornecedores avaliados x N° Total de fornecedores x Tempo em meses) e Gráfico de linha (% de fornecedores avaliados x Tempo em meses)
    * Resultado: A empresa incentiva assim os seus fornecedores a desenvolver o seu sistema de gestão da qualidade e ambiental. Deve-se proceder a uma reavaliação anual, de modo a determinar se os fornecedores mantêm os requisitos.

<br>

14. Custos ambientais: Apresenta o total de custos ambientais no que diz respeito à gestão da energia, água, resíduos e emissões gasosas. A identificação dos custos ambientais permite que as empresas possam identificar os processos mais sustentáveis para a sua atividade, mantendo a sua competitividade.
    * Cálculo: Σ dos custos ambientais com tratamento, transporte, taxas, deposição, etc. (energia, água, resíduos e emissões gasosas) (Real)
    * Origem da informação: Direção de produção e Direção Financeira
    * Apuração: Mensalmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: Gráfico de coluna (Real Custos ambientais x Tempo em meses) com média em linha
    
<br>

#### <u>Principais problema de negócio:</u>
1. Conformidade com os regulamentos: O departamento ambiental deve garantir que a empresa cumpra todos os regulamentos ambientais relevantes, como padrões de emissões, regulamentos de descarte de resíduos e limites de uso de água. O descumprimento desses regulamentos pode resultar em multas, ações judiciais e danos à reputação da empresa.
2. Avaliação de impacto ambiental: A área de meio ambiente deve avaliar o impacto ambiental das atividades e produtos da empresa. Isso pode envolver a realização de avaliações de impacto ambiental, análise de dados sobre emissões e resíduos e identificação de oportunidades para reduzir a pegada ambiental da empresa.
3. Práticas sustentáveis: A área de meio ambiente deve promover práticas sustentáveis dentro da empresa, como redução do uso de energia, minimização de desperdícios e promoção do uso de recursos renováveis. Isso pode envolver a implementação de iniciativas de sustentabilidade, como programas de energia verde, programas de redução de resíduos e programas de conservação de água.
4. Relatórios ambientais: O departamento ambiental deve relatar o desempenho ambiental da empresa às partes interessadas, como reguladores, investidores e clientes. Isso pode envolver a preparação de relatórios ambientais, a divulgação de dados ambientais e a resposta a consultas das partes interessadas.
5. Emergências ambientais: O departamento de meio ambiente deve responder às emergências ambientais, como derramamentos ou vazamentos, para mitigar seu impacto no meio ambiente e evitar danos às pessoas e à vida selvagem. Isso pode envolver o desenvolvimento de planos de resposta a emergências, treinamento de pessoal em procedimentos de emergência e coordenação com agências reguladoras.
6. Treinamento de conformidade: O departamento ambiental deve treinar os funcionários sobre regulamentos, políticas e procedimentos ambientais para garantir que eles estejam cientes de suas responsabilidades e possam ajudar a empresa a cumprir os padrões ambientais.
7. Melhoria contínua: O departamento ambiental deve melhorar continuamente o desempenho ambiental da empresa, identificando áreas de melhoria, implementando as melhores práticas e monitorando o progresso ao longo do tempo. Isso pode envolver a definição de metas ambientais, o acompanhamento de indicadores-chave de desempenho e o envolvimento com as partes interessadas para solicitar feedback e ideias para melhorias.

## Pesquisa e Desenvolvimento

#### <u>Indicadores:</u>
1. Número de novas patentes registradas: a quantidade de patentes registradas pelo departamento de P&D pode ser um indicador do progresso da empresa em inovação e desenvolvimento de novos produtos.
    * Cálculo: calculado contabilizando o número de patentes que foram registradas durante um período de tempo específico.
    * Origem da informação: podem ser buscados nos registros de patentes do departamento, em bancos de dados de patentes e em relatórios de pesquisa e desenvolvimento.
    * Apuração: mensal, trimestral, semestral ou anual.
    * Polaridade: Positivo (Quanto maior o valor, melhor) = uma vez que um aumento no número de patentes registradas é um indicativo de que a empresa está investindo em pesquisa e desenvolvimento, o que pode resultar em novos produtos e/ou processos que agreguem valor ao negócio.
    * Visualização: um gráfico de barras ou linha, que permite comparar a quantidade de patentes registradas em diferentes períodos de tempo.

<br>

2. Taxa de sucesso do lançamento de novos produtos: a taxa de sucesso do lançamento de novos produtos pode ser medida pelo número de produtos lançados que alcançaram ou superaram as expectativas de vendas.
    * Cálculo: quantidade de lançamentos de novos produtos bem-sucedidos dividido pela quantidade total de lançamentos de novos produtos.
    * Origem da informação: buscados em registros de vendas, feedback de clientes, relatórios de produção, entre outros.
    * Apuração: mensal, trimestral, semestral ou anual.
    * Polaridade: Positivo (Quanto maior o valor, melhor) = quanto maior a taxa de sucesso de lançamento de novos produtos, melhor é o desempenho da empresa.
    * Visualização: um gráfico de barras ou um gráfico de linha, mostrando a evolução da taxa de sucesso ao longo do tempo.

<br>

3. Orçamento de P&D: o orçamento dedicado ao departamento de P&D pode ser um indicador de quanto a empresa está investindo em inovação.
    * Cálculo: quantidade de recursos financeiros investidos em atividades de pesquisa e desenvolvimento da empresa.
    * Origem da informação: obtidos a partir de registros financeiros e contábeis da empresa, como balanços, demonstrativos de resultados e relatórios de investimentos em P&D.
    * Apuração: mensalmente, trimestralmente ou anualmente
    * Polaridade: Positivo (Quanto maior o valor, melhor) = quanto maior for o orçamento destinado à P&D, maiores são as chances de a empresa desenvolver novos produtos, melhorar processos e se manter competitiva no mercado.
    * Visualização: gráficos de barras, linhas ou pizza são boas opções para representar visualmente o orçamento de P&D de forma clara e objetiva.

<br>

4. Tempo de lançamento de novos produtos: o tempo que leva para lançar um novo produto no mercado pode ser um indicador da eficiência do departamento de P&D e da capacidade da empresa de se adaptar rapidamente às mudanças no mercado.
    * Cálculo: diferença entre a data de início do desenvolvimento e a data de lançamento oficial do produto.
    * Origem da informação: através de um sistema de gestão de projetos, que registra as datas de início e fim do desenvolvimento do produto e a data de lançamento oficial. Também podem ser utilizadas informações de relatórios de equipe e de processos do departamento de Pesquisa e Desenvolvimento.
    * Apuração: é recomendado que seja mensal, para permitir uma análise mais detalhada do desempenho do departamento.
    * Polaridade: Negativa (Quanto menor o valor, melhor) = quanto menor o tempo de lançamento de novos produtos, melhor o desempenho do departamento de Pesquisa e Desenvolvimento.
    * Visualização: um gráfico de linhas que mostre o tempo decorrido entre a data de início do desenvolvimento e a data de lançamento oficial do produto, permitindo comparar o tempo de lançamento entre diferentes produtos e identificar eventuais atrasos.

<br>

5. Taxa de sucesso de testes de protótipo: a taxa de sucesso dos testes de protótipo pode indicar a eficácia do departamento de P&D em desenvolver soluções que atendam às necessidades dos clientes e sejam viáveis comercialmente.
    * Cálculo: quantidade de testes de protótipo bem-sucedidos dividido pelo número total de testes de protótipo realizados em um determinado período de tempo. O resultado é geralmente apresentado em porcentagem.
    * Origem da informação: obtidos a partir dos registros e relatórios de testes de protótipos realizados pelo departamento de Pesquisa e Desenvolvimento.
    * Apuração: mensal, trimestral, semestral ou anual
    * Polaridade: Positivo (Quanto maior o valor, melhor) = quanto maior for a taxa de sucesso de testes de protótipo, melhor será o desempenho do departamento
    * Visualização: gráfico de linha pode ser uma boa opção para mostrar a evolução da taxa de sucesso de testes de protótipo ao longo do tempo. Além disso, um gráfico de barras pode ser utilizado para comparar a taxa de sucesso de diferentes tipos de protótipos.
    
<br>

6. Percentual de despesas de P&D em relação às receitas: o percentual de despesas de P&D em relação às receitas pode ser um indicador de quanto a empresa está investindo em inovação em relação à sua receita total.
    * Cálculo: dividindo-se os custos de P&D pela receita total e multiplicando por 100. A fórmula é: (Despesas de P&D / Receita Total) * 100.
    * Origem da informação: obtidos no departamento financeiro da empresa, que geralmente é responsável por acompanhar e registrar as receitas e despesas da organização.
    * Apuração: mensal, trimestral, semestral ou anual
    * Polaridade: Positivo (Quanto maior o valor, melhor) = uma vez que a empresa está investindo em pesquisa e desenvolvimento como uma estratégia para aumentar sua competitividade e inovação
    * Visualização: um gráfico de linhas ou um gráfico de barras, que pode mostrar a evolução do percentual de despesas de P&D em relação às receitas ao longo do tempo.

<br>

7. Número de ideias de produtos geradas: o número de ideias de produtos geradas pelo departamento de P&D pode indicar a capacidade da equipe de P&D de encontrar novas oportunidades de mercado e desenvolver soluções inovadoras.
    * Cálculo: contabilizando o número de ideias ou conceitos de produtos que são criados ou propostos pela equipe de P&D em um determinado período de tempo. Essas ideias podem ser originadas internamente ou através de pesquisas de mercado, análise de tendências e outras fontes de inspiração.
    * Origem da informação: coletados através de reuniões com a equipe de P&D, registros de ideias em plataformas de gerenciamento de projetos ou outros sistemas de gestão utilizados pela empresa.
    * Apuração: mensal, trimestral, semestral ou anual
    * Polaridade: Positivo (Quanto maior o valor, melhor) = quanto maior o número de ideias geradas, maior é a chance de encontrar oportunidades de negócio e desenvolver novos produtos bem-sucedidos.
    * Visualização: um gráfico de barras ou de linhas, que mostre a evolução do número de ideias geradas ao longo do tempo. Também pode ser interessante apresentar uma lista com as ideias propostas e o status de cada uma delas, como em análises de pipeline de projetos.

<br>

8. Tempo médio de ciclo de desenvolvimento de produtos: o tempo médio de ciclo de desenvolvimento de produtos pode ser um indicador da eficiência da equipe de P&D em desenvolver soluções de forma rápida e eficaz.
    * Cálculo: dividir o tempo total de desenvolvimento pela quantidade de produtos desenvolvidos no período.
    * Origem da informação: obtidos por meio do acompanhamento do processo de desenvolvimento de cada produto, registrando as datas de início e fim de cada etapa.
    * Apuração: mensal ou trimestral
    * Polaridade: Negativa (Quanto menor o valor, melhor) = pois quanto maior o tempo de ciclo de desenvolvimento de produtos, maior é o impacto no desempenho e competitividade da empresa.
    * Visualização: uma linha do tempo que mostra as datas de início e fim de cada produto desenvolvido no período, com uma média do tempo de desenvolvimento para cada produto.
    
<br>

#### <u>Principais problema de negócio:</u>
1. Falta de financiamento: a falta de financiamento pode limitar o departamento de P&D, impedindo que ele desenvolva projetos ambiciosos e inovadores.
2. Dificuldades em encontrar e reter talentos: encontrar e reter talentos altamente qualificados pode ser um desafio, especialmente em um mercado competitivo, o que pode afetar a qualidade e a eficiência do trabalho desenvolvido.
3. Falta de alinhamento estratégico: o departamento de P&D pode não estar alinhado com a estratégia geral da empresa, o que pode levar a projetos que não atendem às necessidades da empresa ou dos clientes.
4. Problemas de gerenciamento de projetos: projetos de P&D podem ser complexos e exigir gerenciamento cuidadoso para garantir que sejam entregues dentro do prazo e do orçamento.
5. Falta de clareza nas especificações do produto: a falta de clareza nas especificações do produto pode levar a erros no desenvolvimento, resultando em custos extras, tempo de desenvolvimento prolongado e produtos que não atendem às expectativas do cliente.
6. Problemas de comunicação: a comunicação inadequada entre membros da equipe de P&D e outras partes interessadas, como equipes de vendas e marketing, pode levar a problemas de expectativas e falta de coordenação.
7. Falha em priorizar projetos: é importante que a equipe de P&D priorize seus projetos para que possam ser concluídos dentro do prazo e do orçamento, mas a falha em priorizar projetos pode levar a desperdício de tempo e recursos.
8. Dificuldades em adaptar-se a mudanças: o departamento de P&D pode ter dificuldades em adaptar-se a mudanças repentinas nas condições de mercado ou nos requisitos do cliente, o que pode afetar a eficácia de seus projetos.
9. Falta de colaboração e compartilhamento de conhecimento: a falta de colaboração e compartilhamento de conhecimento entre a equipe de P&D pode limitar a inovação e impedir que ideias sejam geradas e desenvolvidas de forma eficaz.
10. Problemas de padronização e controle de qualidade: problemas com padronização e controle de qualidade podem levar a inconsistências nos resultados, falhas em testes e projetos que não atendem às especificações do cliente.

## Atendimento ao Cliente
[&#x1F53C;](#Sumário:) Clique para voltar para o sumário

#### <u>Indicadores:</u>
1. Tempo médio de espera (TME): o tempo médio que um cliente leva para ser atendido por um agente ou receber uma resposta a uma pergunta. 
    * Cálculo: Dividindo o tempo total que os clientes esperaram pelo número total de atendimentos no período
    * Origem da informação: a partir de um software de gestão de atendimento ao cliente ou de um sistema de monitoramento de chamados, que registram o tempo que cada cliente aguardou para ser atendido.
    * Apuração: Pode ser diário, semanal, mensal ou até mesmo anual, dependendo da frequência de atendimento ao cliente e da necessidade de monitorar o desempenho do atendimento
    * Polaridade: Negativa (Quanto menor o valor, melhor) = reduzir o tempo médio de espera dos clientes
    * Visualização: um gráfico de linhas ou de barras que apresente o tempo médio de espera ao longo do período apurado. Também é interessante incluir a meta estabelecida para o TME e indicar se o valor está acima ou abaixo dessa meta.
    * Resultado: O resultado é o tempo médio que cada cliente passou aguardando atendimento.

<br>

2. Tempo médio de atendimento (TMA): o tempo médio que um agente leva para resolver o problema ou atender às necessidades do cliente.
    * Cálculo: Dividindo o tempo total de atendimento de todos os clientes pelo número total de atendimentos no período.
    * Origem da informação: a partir de um software de gestão de atendimento ao cliente ou de um sistema de monitoramento de chamados, que registram o tempo que cada atendimento levou para ser concluído.
    * Apuração: Pode ser diário, semanal, mensal ou até mesmo anual, dependendo da frequência de atendimento ao cliente e da necessidade de monitorar o desempenho do atendimento.
    * Polaridade: Pode ser tanto positiva quanto negativa (um TMA baixo pode indicar que os atendimentos estão sendo realizados de forma eficiente e satisfatória para os clientes e um TMA alto pode indicar que os atendimentos estão sendo feitos de forma muito rápida, sem a devida atenção e cuidado com as demandas dos clientes, o que pode gerar insatisfação e prejudicar a qualidade do atendimento.)
    * Visualização: É um gráfico de linhas ou de barras que apresente o tempo médio de atendimento ao longo do período apurado. Também é interessante incluir a meta estabelecida para o TMA e indicar se o valor está acima ou abaixo dessa meta.
    * Resultado: O resultado é o tempo médio que cada atendimento levou para ser concluído.

<br>

3. Taxa de abandono: a porcentagem de clientes que desistem de uma chamada, chat ou e-mail antes de serem atendidos.
    * Cálculo: Dividindo o número de chamadas abandonadas pelo número total de chamadas recebidas, e multiplicando o resultado por 100
    * Origem da informação: a partir de um software de gestão de atendimento ao cliente ou de um sistema de monitoramento de chamados, que registram o número de chamadas abandonadas e o número total de chamadas recebidas. 
    * Apuração: Pode ser diário, semanal, mensal ou até mesmo anual, dependendo da frequência de atendimento ao cliente e da necessidade de monitorar o desempenho do atendimento.
    * Polaridade: Negativa (Quanto menor o valor, melhor) = indica a insatisfação dos clientes com o atendimento.
    * Visualização: um gráfico de barras que apresente a porcentagem de chamadas abandonadas ao longo do período apurado. Também é interessante incluir a meta estabelecida para a taxa de abandono e indicar se o valor está acima ou abaixo dessa meta.
    * Resultado: O resultado é a porcentagem de chamadas que foram abandonadas pelos clientes antes de serem atendidas.

<br>

4. Nível de serviço (NS): a porcentagem de chamadas, chats ou e-mails que são atendidos dentro de um determinado período de tempo. 
    * Cálculo: dividindo o número de chamadas atendidas dentro do tempo de espera estabelecido pela empresa pelo número total de chamadas recebidas, e multiplicando o resultado por 100
    * Origem da informação: a partir de um software de gestão de atendimento ao cliente ou de um sistema de monitoramento de chamados, que registram o número de chamadas atendidas dentro do tempo de espera e o número total de chamadas recebidas.
    * Apuração: Pode ser diário, semanal, mensal ou até mesmo anual, dependendo da frequência de atendimento ao cliente e da necessidade de monitorar o desempenho do atendimento. 
    * Polaridade: Positivo (Quanto maior o valor, melhor) = Quanto maior a porcentagem de chamadas atendidas dentro do tempo de espera, melhor é a experiência do cliente com o atendimento e maior é a probabilidade de que ele volte a entrar em contato com a empresa.
    * Visualização: é um gráfico de linha que apresente a evolução da porcentagem de chamadas atendidas dentro do tempo de espera ao longo do período apurado. Também é interessante incluir a meta estabelecida para o NS e indicar se o valor está acima ou abaixo dessa meta.
    * Resultado: O resultado é a porcentagem de chamadas atendidas dentro do tempo de espera.

<br>

5. Satisfação do cliente: a avaliação do cliente sobre a qualidade do serviço que ele recebeu. 
    * Cálculo: Mede o grau de satisfação do cliente em relação ao serviço prestado pelo departamento, calculado através de pesquisas de satisfação, avaliações ou feedbacks dos clientes.
    * Origem da informação: através de questionários de pesquisa de satisfação, avaliações em tempo real após o atendimento ao cliente, feedbacks diretos dos clientes, entre outras formas.
    * Apuração: cada trimestre ou semestre
    * Polaridade: Positivo (Quanto maior o valor, melhor)
    * Visualização: um gráfico de barras ou de linhas, que mostre a evolução do índice de satisfação ao longo do tempo e permita comparar os resultados entre períodos.


<br>

6. Primeira resolução de chamadas: a porcentagem de chamadas que são resolvidas na primeira tentativa.
    * Cálculo: divisão do número de chamadas resolvidas na primeira interação pelo número total de chamadas recebidas.
    * Origem da informação: podem ser obtidos através dos registros do sistema de atendimento ao cliente, que devem registrar cada interação com o cliente, incluindo o número de chamadas recebidas e resolvidas na primeira interação.
    * Apuração: período mensal, para que haja tempo suficiente para que uma quantidade significativa de chamadas seja registrada.
    * Polaridade: Positivo (Quanto maior o valor, melhor) = é um indicador de eficiência e qualidade no atendimento ao cliente.
    * Visualização: um gráfico de barras ou um gráfico de pizza, que podem mostrar a proporção de chamadas resolvidas na primeira interação de forma clara e simples.


<br>

7. Taxa de repetição de chamadas: a porcentagem de clientes que ligam novamente para tratar do mesmo problema.
    * Cálculo: dividindo o número total de chamadas que precisaram ser repetidas pelo número total de chamadas recebidas, multiplicando o resultado por 100 para obter a porcentagem.
    * Origem da informação: podem ser encontrados no sistema de registro de chamadas ou em ferramentas de monitoramento de chamadas, onde é possível acompanhar o histórico de cada atendimento e identificar quais precisaram ser repetidos.
    * Apuração: mensal ou semestral. 
    * Polaridade: Negativa (Quanto menor o valor, melhor) = indica que o atendimento não foi eficiente e o problema não foi solucionado na primeira tentativa
    * Visualização: gráficos de barras ou linhas, que permitem comparar a taxa de repetição de chamadas ao longo do tempo e identificar tendências.


<br>

8. Taxa de conversão: a porcentagem de clientes que foram atendidos e converteram-se em vendas.
    * Cálculo: dividindo o número de chamadas ou contatos que resultaram em uma venda ou solução pelo número total de chamadas ou contatos recebidos no mesmo período de tempo e multiplicando por 100 para obter uma porcentagem
    * Origem da informação: a partir de sistemas de gerenciamento de relacionamento com o cliente (CRM) ou de sistemas de gerenciamento de chamadas (CMS), que registram as informações de cada chamada ou contato.
    * Apuração: mensal, trimestral ou anual.
    * Polaridade: Positivo (Quanto maior o valor, melhor) = uma taxa mais alta indica que a equipe está sendo eficaz em converter chamadas ou contatos em vendas ou soluções para o cliente.
    * Visualização: um gráfico de barras ou um gráfico de linha, que mostram a evolução do KPI ao longo do tempo.

<br>

9. Custo por interação: o custo médio de cada interação com o cliente (por exemplo, telefone, chat ou e-mail).
    * Cálculo: É calculado dividindo o custo total do canal de atendimento pelo número de interações realizadas nesse canal.
    * Origem da informação: podem ser obtidos do departamento financeiro da empresa, que deve fornecer informações detalhadas sobre os custos associados a cada canal de atendimento.
    * Apuração: Mensalmente ou trimestralmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: Um gráfico de barras ou um gráfico de linha, que pode mostrar a evolução do custo por interação ao longo do tempo e ajudar a identificar tendências.

<br>

10. Tempo de espera na fila: o tempo que um cliente espera antes de ser atendido. 
    * Cálculo: calculado pela soma do tempo que cada cliente passa na fila de espera até ser atendido, dividido pelo número total de clientes atendidos em um determinado período de tempo.
    * Origem da informação: encontrados no sistema de atendimento ao cliente, que geralmente registra o tempo de espera de cada cliente na fila de espera antes de ser atendido.
    * Apuração: Mensalmente ou semanalmente
    * Polaridade: Negativa (Quanto menor o valor, melhor)
    * Visualização: Um gráfico de linha, que pode mostrar a evolução do tempo médio de espera na fila ao longo do tempo, permitindo que a equipe de atendimento identifique tendências e faça ajustes para melhorar o tempo de espera dos clientes.
    
<br>

#### <u>Principais problema de negócio:</u>
1. Longos tempos de espera: clientes podem ficar frustrados quando precisam esperar muito tempo para serem atendidos, o que pode levar a uma experiência negativa.
2. Falta de comunicação eficaz: se a comunicação entre o cliente e o agente de atendimento não for clara e eficaz, o cliente pode ficar ainda mais confuso e frustrado.
3. Falta de treinamento adequado: se os agentes de atendimento ao cliente não receberem treinamento adequado, eles podem não ter as habilidades necessárias para lidar com problemas e clientes difíceis.
4. Falta de empoderamento dos agentes: se os agentes não tiverem autonomia para tomar decisões e resolver problemas, isso pode levar a uma experiência insatisfatória para o cliente.
5. Falta de acesso a informações do cliente: se o agente não tiver acesso a informações relevantes do cliente, como histórico de compras ou problemas anteriores, pode ser difícil para eles resolverem problemas de forma eficaz.
6. Falta de personalização: os clientes esperam que suas interações com as empresas sejam personalizadas, e a falta de personalização pode levar a uma experiência negativa.
7. Falta de acompanhamento: se o agente não fizer um acompanhamento após a resolução do problema, o cliente pode sentir que sua preocupação não foi levada a sério.
8. Problemas técnicos: se houver problemas técnicos com o sistema de atendimento ao cliente, como falhas no software ou na conexão com a internet, isso pode afetar negativamente a experiência do cliente.
9. Falta de empatia: se os agentes de atendimento ao cliente não demonstrarem empatia e compreensão em relação aos problemas do cliente, isso pode levar a uma experiência insatisfatória.

## Compliance
[&#x1F53C;](#Sumário:) Clique para voltar para o sumário

#### <u>Indicadores:</u>
1. Taxa de conformidade: mede a porcentagem de operações da empresa que estão em conformidade com as regulamentações aplicáveis.
    * Cálculo: pela proporção entre o número de ações em conformidade com as políticas e regulamentações internas e externas do negócio e o número total de ações auditadas.
    * Origem da informação: podem ser coletados por meio de auditorias internas e externas que avaliam a conformidade das atividades realizadas pelo departamento
    * Apuração: O período de apuração pode variar de acordo com a frequência das auditorias e o ciclo de avaliação da conformidade das atividades.
    * Polaridade: Positivo (Quanto maior o valor, melhor) = quanto maior a taxa de conformidade, melhor é o desempenho do departamento Compliance.
    * Visualização: um gráfico de barras que demonstra a evolução da taxa de conformidade ao longo do tempo, permitindo a comparação entre os resultados de diferentes períodos.

<br>

2. Taxa de detecção de violações: mede a porcentagem de violações identificadas pelo departamento de Compliance em relação ao total de operações analisadas.
    * Cálculo: o número de violações detectadas em um determinado período de tempo dividido pelo número total de atividades monitoradas ou auditadas no mesmo período de tempo
    * Origem da informação: obtidos por meio de sistemas de monitoramento e auditoria de atividades de negócios, relatórios de incidentes de conformidade e dados de reclamações de clientes.
    * Apuração: O período de apuração pode variar, dependendo da periodicidade do monitoramento e da auditoria das atividades de negócios e das reclamações de clientes.
    * Polaridade: Negativa (Quanto menor o valor, melhor) = uma vez que um aumento na taxa de detecção de violações pode indicar um aumento nos riscos de conformidade e, portanto, um maior potencial de perda financeira e de reputação para a empresa.
    * Visualização: um gráfico de linhas ou barras, que mostrasse a evolução da taxa de detecção de violações ao longo do tempo, permitindo a identificação de tendências e padrões.

<br>

3. Tempo médio de resposta: mede o tempo médio que o departamento de Compliance leva para responder a uma violação ou relatório de não conformidade.
    * Cálculo: coletar dados sobre o tempo que a equipe de Compliance leva para responder às solicitações e consultas recebidas
    * Origem da informação: obtidos por meio de registros internos de atendimento, como sistemas de tickets ou ferramentas de gestão de chamados.
    * Apuração: mensal ou trimestral.
    * Polaridade: Negativa (Quanto menor o valor, melhor) = um tempo médio de resposta alto indica que a equipe de Compliance está levando mais tempo do que o esperado para responder às solicitações e consultas relacionadas às políticas de conformidade da empresa, o que pode impactar negativamente a eficácia do departamento.
    * Visualização: um gráfico de linha ou de barras que apresente o tempo médio de resposta ao longo do tempo, com destaque para as variações significativas. Também é possível apresentar este KPI em um painel de indicadores, juntamente com outros indicadores de desempenho do departamento de Compliance.

<br>

4. Número de violações resolvidas: mede o número de violações que foram resolvidas com sucesso pelo departamento de Compliance.
    * Cálculo: contagem do número de violações identificadas pelo departamento de Compliance que foram efetivamente resolvidas dentro de um período específico.
    * Origem da informação: podem ser obtidos a partir dos registros de violações identificadas pelo departamento de Compliance, bem como das informações sobre a resolução dessas violações.
    * Apuração: mensal, trimestral, semestral ou anual
    * Polaridade: Positivo (Quanto maior o valor, melhor) = quanto maior o número de violações resolvidas, melhor é o desempenho do departamento de Compliance em garantir a conformidade da organização com as regulamentações e políticas internas.
    * Visualização: um gráfico de barras ou um gráfico de linhas que exibisse a evolução do número de violações identificadas e resolvidas ao longo do tempo.

<br>

5. Número de relatórios de não conformidade recebidos: mede o número de relatórios de não conformidade recebidos pelo departamento de Compliance em um determinado período de tempo.
    * Cálculo: a quantidade de relatórios de não conformidade recebidos por esse departamento dentro de um período específico.
    * Origem da informação: buscados nos registros do departamento Compliance, que devem registrar todos os relatórios recebidos.
    * Apuração: mensal ou trimestral
    * Polaridade: Negativa (Quanto menor o valor, melhor) = já que um aumento no número de relatórios de não conformidade recebidos indica que há um maior número de problemas a serem resolvidos pelo departamento Compliance.
    * Visualização: um gráfico de barras ou um gráfico de linhas mostrando a evolução do número de relatórios recebidos ao longo do tempo

<br>

6. Taxa de participação em treinamentos: mede a porcentagem de funcionários que participaram dos treinamentos de Compliance oferecidos pelo departamento.
    * Cálculo: Número de funcionários que participaram dos treinamentos / Número total de funcionários do departamento de Compliance x 100%
    * Origem da informação: podem ser obtidos por meio dos registros de participação nos treinamentos realizados pelo departamento de Compliance, como por exemplo, os registros de presença em aulas ou de conclusão de cursos e treinamentos.
    * Apuração: mensal, trimestral ou anual
    * Polaridade: Positivo (Quanto maior o valor, melhor) = quanto maior a taxa de participação em treinamentos, melhor será o desempenho do departamento de Compliance na conscientização e capacitação dos seus funcionários em relação às políticas e procedimentos de conformidade.
    * Visualização: um gráfico de barras ou um gráfico de linha, que apresente a taxa de participação em treinamentos ao longo do tempo. Também pode ser útil apresentar a taxa de participação de cada funcionário individualmente em um gráfico de pizza ou de rosca.

<br>

7. Número de políticas e procedimentos atualizados: mede o número de políticas e procedimentos atualizados pelo departamento de Compliance em relação ao total de políticas e procedimentos da empresa.
    * Cálculo: contabilizando o número de políticas e procedimentos atualizados em um determinado período
    * Origem da informação: nos registros de atualizações e revisões das políticas e procedimentos do departamento Compliance.
    * Apuração: mensal ou trimestral
    * Polaridade: Positivo (Quanto maior o valor, melhor) = uma vez que o aumento do número de políticas e procedimentos atualizados indica que o departamento está em conformidade com as normas e regulamentos em constante mudança e, portanto, mais preparado para lidar com possíveis riscos e violações.
    * Visualização: um gráfico de linha que demonstra a evolução do número de políticas e procedimentos atualizados ao longo do tempo

<br>

8. Número de auditorias realizadas: mede o número de auditorias realizadas pelo departamento de Compliance em um determinado período de tempo.
    * Cálculo: total de auditorias realizadas pelo departamento Compliance durante um determinado período.
    * Origem da informação: obtidos por meio de relatórios de auditoria, planilhas de acompanhamento ou sistemas de gerenciamento de auditorias.
    * Apuração: mensal ou trimestral
    * Polaridade: Positivo (Quanto maior o valor, melhor) = quanto maior o número de auditorias realizadas pelo departamento Compliance, maior é a efetividade do programa de conformidade e menor é o risco de não conformidades e sanções.
    * Visualização: um gráfico de barras ou um gráfico de linha, comparando o número de auditorias realizadas em diferentes períodos.

<br>

9. Taxa de resolução de não conformidades: mede a porcentagem de não conformidades resolvidas pelo departamento de Compliance em relação ao total de não conformidades identificadas.
    * Cálculo: dividindo o número de não conformidades resolvidas pelo número total de não conformidades identificadas no período e multiplicando o resultado por 100 para obter a porcentagem.
    * Origem da informação: obtidos a partir de relatórios de auditoria, relatórios de inspeção, relatórios de incidentes e reclamações de clientes
    * Apuração: mensal ou trimestral
    * Polaridade: Positivo (Quanto maior o valor, melhor) = quanto maior a taxa de resolução de não conformidades, melhor é o desempenho do departamento Compliance.
    * Visualização: um gráfico de barras ou um gráfico de linha, mostrando a tendência ao longo do tempo.

<br>

10. Nível de satisfação do cliente interno: mede o nível de satisfação dos funcionários da empresa com os serviços prestados pelo departamento de Compliance.
    * Cálculo: calculado através de uma pesquisa de satisfação que busca avaliar o grau de satisfação dos colaboradores da empresa com relação aos serviços e suporte prestados pelo departamento de Compliance.
