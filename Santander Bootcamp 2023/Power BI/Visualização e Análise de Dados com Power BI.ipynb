### Businesse Intelligence:
Consiste em uma área composta por estratégias e tecnologias utilizadas pelas empresa a fim de realizar análise de dados e gerenciamento das informações.
* Objetivo: Transformar dados em conhecimento
* Entendimento: Fatores conhecidos e não conhecidos
* Foco: Operações e melhorias
* Evolução do conceito de BI: Relatorios > Relatórios mais sofisticados (Exemplos: Comportamentos de clientes, monitoramento de ativos e previbilidade de vendas)
* Exemplos de métricas: Receitas recorrente, Produtividade, Margem bruta de lucro, Retorno sobre o investimento e número total de clientes
* Futuro: É entender os diferentes escopos, variedade dos dados, especificidade e granularidade de dados

<br>

### Tipos de análises:
* Descritiva (Caracterizar e sumarizar para entender o comportamento dos dados) = O que aconteceu?
* Diagnósticas (Processo probabilístico e encontrar relações de causa e efeito no cenarios descrito pelos dados) = Por que aconteceu?
* Preditiva (Tentativa de prever um cenário utilizando probabilidade e análise de tendências) = O que vai acontecer?
* Prescritivo (Resolução de um problema, subsídio dos impactos e prescrever uma solução com base no diagnóstico) = O que fazer?

<br>

## Fundamentos Teóricos sobre ETL e ELT (Processamento de Dados) <br>
* O que é ETL (Extract, Transform and Load): 
    * Técnicas de extração de inúmeras fontes de dados (CRM, Cloud, Filas, SGBD, Legados e Files) > Coleta de dados brutos (Não estruturados, estruturados e semi) > Limpeza
    * Técnicas de transformação (Conversão de um formato para outro "colocando uma estrutura", dependendo onde será armazenado)
    * Técnicas de carregamento (Repositório central ou distribuidos, Data Warehouse/Banco de dados relacional ou não relacional, Banco de dados físico ou na Nuvem) > Análise

<br>

* O que é ELT (Extract, Load and Transform):
    * Técnicas de extração de inúmeras fontes de dados (CRM, Cloud, Filas, SGBD, Legados e Files) > Coleta de dados brutos (Não estruturados, estruturados e semi) > Limpeza
    * Técnicas de carregamento (Repositório central ou distribuidos, Data Warehouse/Banco de dados relacional ou não relacional, Banco de dados físico ou na Nuvem)
    * Técnicas de transformação (Conversão de um formato para outro "colocando uma estrutura", dependendo onde será armazenado) > Análise 
    
<br>

#### Vantagens da ELT: 
* Mudança de pespectiva de dados faz com que o processo seja mais ágil (Otimização de tempo), quando carregamos antes de transforma-los, principalmente para os dados não estruturados.  
* Os analistas que realizam a transformação dos dados (no ELT) em vez do Engenheiro de dados (no ETL), ou seja, menor dependência da TI e o analista tem um papel principal.
* Tempo de carregamento (Sistemas integrado para realizar o carregamento diretamente no repositório > Delay menor) 
* Tempo de transformação (Consome muito tempo no ETL)
* Tempo de manutenção (Custo com atualização)
* Complexidade de implementação (Custo com atualização e maior capacidade de armazenamento)
* Limitação dos dados (Quais dados estão disponíveis)
* Suporte a Data Warehouses
* Usabilidade (Vai depender da abordagem do processo)

<br>

#### Qual escolher:
* O ETL foi criado quando armazenar era muito caro, porém o processamento dos dados é mais lento
* O ELT o armazenamento é feito diretamente no repositório central e o processo de dados é mais rápido.

### Fundamentos de BI: KPIs e Métricas <br>

* KPIS (Key Performance Indicator): Medir desempenho e o progresso da empresa 
    * Meça o progresso em direção às principais metas de negócios
    * Eles têm uma perspectiva de alto nível
    * Eles são relevantes em diferentes departamentos
    * Usado para tomada de decisões estratégicas
* Visão dos Stakeholders sobre a empresa (Ter uma visão da empresa)
* Varia de acordo com contexto de cada empresa
* Quais são os benefícios?
    * Contabilização da performance dos colabores (Realizar um Track/Rastreamento ou progresso da performance)
    * Permite ajuste dos gerentes (Adequação das estratégias para alavancar a performance novamente)
    * Todos na mesma página (Ter a mesma perspectiva, ter um alvo comum)
    * Avalia a saúde da empresa como um todo ou por setores (Visão objetiva e definida de como está performance)
* Categorias 
    * Operacional (Como está o dia a dia, processo mais curto)
    * Estratégica (Objetivo mais amplo, longo prazo) 
    * Leading ou Lagging (Adiantado ou Atrasado)
    * Qualitativa ou Quantitativa
* Níveis
    * Nível de Companhia (Organizacional - Geral)
    * Nível de Departamento (Setor - Específico)
    * Nível do Projeto
* Entender qual é o objetivo do KPI
    * Por que esse resultado é importante?
    * Quem é responsavel pelo resultado?
    * Como o resultado pode ser influenciado?
    * Como você vai medir o progresso?
    * Com que frequência você medirá o progesso?
    * Como você saberá se atingiu seus objetivos?
* Criação: S(Específico), M (Mensuravel), A (Alcançavel), R (Realista/Relevante) e T (Temporal)
* Observações: Os KPIS precisam ser simples e claros, ter uma comunicação para divulga-los e precisa de um refinamento ao longo do tempo.

<br>

* Métricas: 
    * Meça o desempenho de atividades ou processos de negócios específicos.
    * Eles têm uma perspectiva de nível inferior
    * Eles são relevantes para departamentos ou áreas de negócios específicos
    * Tendem a ser operacionais ou táticos

### Criando Dashboard Interativos <br>

O que é? Os dashboards permitem que os consumidores de relatório criem um artefato de dados direcionados que é personalizado apenas para eles. Enquanto um relatório de Power BI usa dados de um conjunto de dados, um dashboard do Power BI pode conter visuais de diferentes conjuntos de dados.

<br>

##### Power BI Serviço > Dashboard: <br>
* Vários conjuntos de dados ou reports (Os dashboards permitem que um usuário fixe visuais de diferentes relatórios e conjuntos de dados em uma tela, tornando mais simples agrupar o que é importante para o usuário. Os relatórios, por outro lado, estão mais concentrados em conseguir visualizar e aplicar transformações a um conjunto de dados.)
* Não possuem nos painéis os Filtro, Visualização e Campos (São visuais fixados e visuais de relatórios distintos)
* Podem ter apenas uma página
* Dataset analisado > Report
* Ambos podem ser atualizados
* Manipulação dos blocos (Os blocos podem ser originados de uma infinidade de lugares, incluindo de relatórios, conjuntos de dados, outros dashboards, Microsoft Excel, SQL Server Reporting Services e muito mais. Ao fixar um elemento de relatório em um dashboard, você cria uma conexão direta entre o dashboard e o relatório do qual o instantâneo veio. Um dos maiores benefícios de um dashboard é a possibilidade de fixar um visual que é originado de um conjunto de dados diferente.)
> https://learn.microsoft.com/pt-br/power-bi/create-reports/service-dashboard-tiles
* Alertas para notificar usuários quando alterar medidores, cartões e KPIS (Esse recurso está disponível para quem tiver acesso ao dashboard, não apenas ao proprietário do dashboard.)
> https://learn.microsoft.com/pt-br/power-bi/create-reports/service-set-data-alerts
* Coerência na utilização de temas (São Templates JSON e Possibilidade de aplicação geral)
> https://community.fabric.microsoft.com/t5/Themes-Gallery/bd-p/ThemesGallery <br>
> https://learn.microsoft.com/pt-br/power-bi/create-reports/service-dashboard-themes
* Os dashboards devem ser uma coleção de várias fontes, não apenas uma "plataforma de lançamento" para relatórios.
> https://learn.microsoft.com/pt-br/power-bi/create-reports/service-create-dashboard-mobile-phone-view
* Caixa de Pergunta P e R para fazer sobre os dados
