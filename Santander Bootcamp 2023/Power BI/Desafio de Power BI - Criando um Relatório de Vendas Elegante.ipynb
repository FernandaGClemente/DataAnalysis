# Criando um Relatrio de Vendas Elegante com Power BI

#### Contexto: Voc  um cientista de dados no Santander e recebeu a tarefa de criar um relatrio no Power BI Desktop com base na Sample Financials disponibilizada pela prpria Microsoft. <br>
#### Passo a passo realizado no projeto: <br>
1. Importao do conjunto de dados para o Power BI (Arquivo Financial Sample.xlxs);
2. Transformao dos dados no Editor Power Query;
    * Alterao do nome da tabela para "Amostra Financeira";
    * Incluiso do parmetro do caminho do arquivo Financial Sample.xlxs;
    * Alterao dos nomes de todas as colunas para portugus;
    * Alterao do tipo de dados de todas as colunas;
    * Criao da tabela Calendrio dos anos 2013 e 2014, incluindo as colunas Ano, Ms, Dia, Incio do ms e Nome do ms;
3. Na exibio de modelo: Criao da relao da tabela Calendrio com a Tabela Amostra Financeira;
4. No modo de exibio de tabela: Alterao das formataes das colunas, caso seja necessrio;
5. Criao das 3 telas de fundos de todas as pginas no Power Point; 
6. Na exibio de relatrio: Incluso das telas de fundos criadas nas respectivas pginas;
7. Criao dos Botes de navegaes das pginas;
    * Na pgina Menu foi criados 2 botes: Um para voltar para a pgina de Vendas (Desenho de Seta) e outro para voltar para a pgina de Resultado (Desenho de Seta)
    * Na pgina Venda foram criado 3 botes para:
        * Para voltar para a pgina Menu (Desenho de Casa)
        * Para voltar para a pgina de Resultado (Desenho de Seta)
        * Para limpar todos os filtros (Borracha)
        * Para selecionar o grfico de Barra das vendas por segmento
        * Para selecionar o grfico de Matriz das vendas por segmento
        * Para selecionar o grfico de Barra das vendas por produto
        * Para selecionar o grfico de Matriz das vendas por produto
        * Para selecionar o grfico de Barra das vendas por pas
        * Para selecionar o grfico de Mapa das vendas por pas
    * Na pgina Resultado foram criados 3 botes: Um para voltar para a pgina Menu (Desenho de Casa), outro para voltar para a pgina de Vendas (Desenho de Seta) e o terceiro para Limpar todos os filtros (Borracha)
8. Criao dos grficos e filtros da pgina Vendas;
    * Criao de 5 cartes com os dados da Venda, Unidade vendidas, Descontos, Custo
    * Criao do Grfico de Linha com os dados da Venda por ms;
    * Criao do Grfico de Barra Clusterizado com os dados da Venda por segmento;
    * Criao do Grfico de Barra Clusterizado com os dados da Venda por produto;
    * Criao do Grfico de Mapa com os dados da Venda por pas;
    * Criao do filtro de data;
9. Criao dos grficos e filtros da pgina Resultado;
    * Criao do Grfico de Radar com os dados do Lucro por produto;
    * Criao do Grfico de rvore Hierarquica com os dados do Preo de venda total, Ano e Pas;
    * Criao do Grfico Treemap com os dados do Lucro por segmento;
    * Criao do Grfico de cascata com os dados do Lucro por trimestre;
    * Criao do filtro de ano;
10. Para finalizar, foi feito a publicao dos relatorios no Meu Workspace
