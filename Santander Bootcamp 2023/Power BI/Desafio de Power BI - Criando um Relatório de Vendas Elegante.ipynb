# Criando um Relatório de Vendas Elegante com Power BI

#### Contexto: Você é um cientista de dados no Santander e recebeu a tarefa de criar um relatório no Power BI Desktop com base na Sample Financials disponibilizada pela própria Microsoft. <br>
#### Passo a passo realizado no projeto: <br>
1. Importação do conjunto de dados para o Power BI (Arquivo Financial Sample.xlxs);
2. Transformação dos dados no Editor Power Query;
    * Alteração do nome da tabela para "Amostra Financeira";
    * Incluisão do parâmetro do caminho do arquivo Financial Sample.xlxs;
    * Alteração dos nomes de todas as colunas para português;
    * Alteração do tipo de dados de todas as colunas;
    * Criação da tabela Calendário dos anos 2013 e 2014, incluindo as colunas Ano, Mês, Dia, Início do mês e Nome do mês;
3. Na exibição de modelo: Criação da relação da tabela Calendário com a Tabela Amostra Financeira;
4. No modo de exibição de tabela: Alteração das formatações das colunas, caso seja necessário;
5. Criação das 3 telas de fundos de todas as páginas no Power Point; 
6. Na exibição de relatório: Inclusão das telas de fundos criadas nas respectivas páginas;
7. Criação dos Botões de navegações das páginas;
    * Na página Menu foi criados 2 botões: Um para voltar para a página de Vendas (Desenho de Seta) e outro para voltar para a página de Resultado (Desenho de Seta)
    * Na página Venda foram criado 3 botões para:
        * Para voltar para a página Menu (Desenho de Casa)
        * Para voltar para a página de Resultado (Desenho de Seta)
        * Para limpar todos os filtros (Borracha)
        * Para selecionar o gráfico de Barra das vendas por segmento
        * Para selecionar o gráfico de Matriz das vendas por segmento
        * Para selecionar o gráfico de Barra das vendas por produto
        * Para selecionar o gráfico de Matriz das vendas por produto
        * Para selecionar o gráfico de Barra das vendas por país
        * Para selecionar o gráfico de Mapa das vendas por país
    * Na página Resultado foram criados 3 botões: Um para voltar para a página Menu (Desenho de Casa), outro para voltar para a página de Vendas (Desenho de Seta) e o terceiro para Limpar todos os filtros (Borracha)
8. Criação dos gráficos e filtros da página Vendas;
    * Criação de 5 cartões com os dados da Venda, Unidade vendidas, Descontos, Custo
    * Criação do Gráfico de Linha com os dados da Venda por mês;
    * Criação do Gráfico de Barra Clusterizado com os dados da Venda por segmento;
    * Criação do Gráfico de Barra Clusterizado com os dados da Venda por produto;
    * Criação do Gráfico de Mapa com os dados da Venda por país;
    * Criação do filtro de data;
9. Criação dos gráficos e filtros da página Resultado;
    * Criação do Gráfico de Radar com os dados do Lucro por produto;
    * Criação do Gráfico de Árvore Hierarquica com os dados do Preço de venda total, Ano e País;
    * Criação do Gráfico Treemap com os dados do Lucro por segmento;
    * Criação do Gráfico de cascata com os dados do Lucro por trimestre;
    * Criação do filtro de ano;
10. Para finalizar, foi feito a publicação dos relatorios no Meu Workspace
