### O debate R versus Python: <br>
##### R
* Características comuns: 
    - Código aberto 
    - Dados armazenados em data frames 
    - Fórmulas e funções prontamente disponíveis 
    - Comunidade para desenvolvimento e suporte de código
* Vantagens únicas: 
    - Manipulação de dados, visualização de dados e pacotes de estatísticas 
    - Abordagem de "bisturi" para dados: encontre pacotes para fazer o que quiser com os dados

### Planilhas, SQL e R: uma comparação: <br>
##### Cada ferramenta tem seus próprios pontos fortes e fracos, mas todas tornam o processo de análise de dados mais tranquilo e eficiente. Existem duas coisas principais que todos os três têm em comum: filtros e funções
* O que é?
    * Planilhas: Um programa que usa linhas e colunas para organizar dados e permite análise e manipulação por meio de fórmulas, funções e recursos integrados
    * SQL: Uma linguagem de programação de banco de dados usada para se comunicar com bancos de dados para realizar uma análise de dados
    * R: Uma linguagem de programação de uso geral usada para análise estatística, visualização e outras análises de dados
* Qual é a principal vantagem?
    * Planilhas: Inclui uma variedade de ferramentas e recursos de visualização
    * SQL: Permite que os usuários manipulem e reorganizem os dados conforme necessário para auxiliar na análise
    * R: Fornece uma linguagem acessível para organizar, modificar e limpar quadros de dados e criar visualizações de dados criteriosas
* Com quais conjuntos de dados funciona melhor?
    * Planilhas: Conjuntos de dados menores
    * SQL: Conjuntos de dados maiores
    * R: Conjuntos de dados maiores
* Qual é a fonte dos dados?
    * Planilhas: Inserido manualmente ou importado de uma fonte externa
    * SQL: Acessado de um banco de dados externo
    * R: Carregado com R quando instalado, importado do seu computador ou carregado de fontes externas
* Onde normalmente são armazenados os dados da minha análise?
    * Planilhas: Em um arquivo de planilha no seu computador
    * SQL: Dentro de tabelas no banco de dados acessado
    * R: Em um arquivo R no seu computador
* Eu uso fórmulas e funções?
    * Planilhas: Sim
    * SQL: Sim
    * R: Sim
* Posso criar visualizações?
    * Planilhas: Sim
    * SQL: Sim, usando uma ferramenta adicional como um sistema de gerenciamento de banco de dados (SGBD) ou uma ferramenta de business intelligence (BI)
    * R: Sim

### Fundamentos de programação em R <br>
##### Conceitos básicos de R:  
* Funções: são um conjunto decódigo reutilizável usado para executar tarefas específicas em R. As funções começam com nomes de funções como imprimir ou colar, egeralmente são seguidos por um ou mais argumentos entre parênteses.Um argumento é a informação que uma função em R precisa para ser executada. Ex: Saber mais sobre a função → ?função() + Enter
* Comentários: Os comentários são úteis quando você deseja descrever ou explique o que está acontecendo no seu código. Comentários devem ser usados para tornar um script R mais legível.Um comentário não deve ser tratado como código, por isso colocaremos um # na frente dele.
* Variáveis: Uma variável é uma representação de um valor em R que pode ser armazenado porusar mais tarde durante a programação.Variáveis também podem ser chamadas de objetos. O nome de uma variável deve começar com uma letra etambém pode conter números e sublinhados.
    * Operador de atribuição: <-
    * Sintaxe: NomeDaVariavel <- Valor
* Tipos de dados 
* Vetores: um vetor é um grupo de elementos de dados do mesmo tipo armazenado em uma sequência em R. Você pode criar um vetor usando a função combinada. Em R esta função é apenas a letra c seguida pelos valores que você deseja em seu vetor entre parênteses.
    * Operador de atribuição: <-
    * Função c() (chamada de função “combinar”) combina vários valores em um vetor. 
    * Sintaxe: NomeDovetor <- c(Valor)
    * Para criar um vetor de inteiros usando a função c(), você deve colocar a letra “L” diretamente após cada número.
    * Determinar o tipo de vetor: typeof() → Exemplo: <br>
        typeof(c(1L , 3L, 5L))
    * Determinar o comprimento de vetor: length() → Exemplo: <br>
        x <- c(1L , 3L, 5L) <br>
        length(x)
    * Verificar se um vetor é de um tipo específico usando uma função is: is.logical(), is.double(), is.integer(), is.character() → Exemplo: <br>
        x <- c(1L , 3L, 5L) <br>
        is.integer(x)
    * Todos os tipos de vetores podem ser nomeados. Os nomes são úteis para escrever código legível e descrever objetos em R. Você pode nomear os elementos de um vetor com a função names() <br>
        x <- c(1L , 3L, 5L) <br>
        names(x) <- c("a", "b", "c") <br>

##### Na programação, uma estrutura de dados é um formato para organizar e armazenar dados. É importante entender as estruturas de dados porque você trabalhará com elas frequentemente ao usar R para análise de dados. As estruturas de dados mais comuns na linguagem de programação R incluem: 
* Vetores: 
    * Vetores atômicos = Existem seis tipos principais de vetores atômicos: lógico, inteiro, duplo, caractere (que contém strings), complexo e bruto 
    * Listas = As listas são diferentes dos vetores atômicos porque seus elementos podem ser de qualquer tipo. As listas podem até conter outras listas.
        * Sintaxe: list(Valor)
        * Se quiser descobrir quais tipos de elementos uma lista contém, você pode usar a função str(). Para isso, coloque o código da lista entre parênteses da função. 
        * As listas podem ser nomeadas quando são criadas: list('Nome1' = Valor1, 'Nome2' = Valor2, 'Nome3' = Valor3)
* DataFrames:
    * Se precisar criar manualmente um quadro de dados em R, você pode usar a função data.frame() . A função data.frame() recebe vetores como entrada. 
    * Sintaxe: data.frame(NomeVariavel = c(ValorObs), NomeVariavel = c(ValorObs))
    * Exemplo: data.frame(x = c(1, 2, 3) , y = c(1.5, 5.5, 7.5))

    * Uma matriz é uma coleção bidimensional de elementos de dados. Isso significa que possui linhas e colunas. Por outro lado, um vetor é uma sequência unidimensional de elementos de dados. Mas, assim como os vetores, as matrizes só podem conter um único tipo de dados. 
    * Para criar uma matriz em R, você pode usar a função array() . A função array() possui dois argumentos principais que você insere entre parênteses. Primeiro, adicione um vetor. O vetor contém os valores que você deseja colocar na matriz. A seguir, adicione pelo menos uma dimensão de matriz. Você pode optar por especificar o número de linhas ou o número de colunas usando o código nrow = ou ncol = . 
    * Sintaxe: matrix(Vetor), nrow ou ncol = Nr)
    * Exemplo: matrix(c(3:8), nrow = 2) ou matrix(c(3:8), ncol = 2)

#### <u>Depuração</u> <br>

#### A depuração bem-sucedida de qualquer código R começa com o diagnóstico correto do problema. A primeira etapa para diagnosticar o problema em seu código é entender o que você esperava que ocorresse. Então, você pode identificar o que realmente ocorreu e como isso se diferenciou de suas expectativas. <br>

#### Por exemplo, imagine que você deseja executar a função glimpse() para obter uma exibição resumida do conjunto de dados dos pinguins . Você escreve o seguinte código:

    Glimpse(penguins)

#### Quando você executa a função, você obtém o seguinte resultado:

    Error in Glimpse(penguins) : could not find function "Glimpse"

#### Você esperava uma exibição do conjunto de dados. Em vez disso, você recebeu uma mensagem de erro. O que deu errado? Nesse caso, o problema pode ser diagnosticado como um erro estilístico: você escreveu Glimpse com um "G" maiúsculo, mas o código diferencia maiúsculas de minúsculas e requer um "g" minúsculo. Se você executar o código glimpse(penguins) você obterá o resultado esperado. <br>

#### Ao diagnosticar o problema, é mais provável que você – e qualquer outra pessoa que possa ajudar depurar seu código – entenderá o problema se você fizer as seguintes perguntas:
* Qual foi a sua contribuição?
* O que você estava esperando?
* O que você ganhou?
* Como o resultado se diferencia de suas expectativas originais?
* Suas expectativas estavam corretas em primeiro lugar?
 
#### Alguns bugs são difíceis de descobrir, e descobrira causa do problema pode ser um desafio. Se você se deparar com mensagens de erro ou precisar de ajuda com um bug, start pesquisando on-line para obter informações sobre ele. #### a) Sensibilidade a maiúsculas e minúsculas:

##### Em R, você precisa ter certeza de que cada parêntese de abertura em sua função tenha um parêntese de fechamento e cada aspa de abertura tenha uma aspa de fechamento. 
    O RStudio informa que você tem um colchete de abertura incomparável. 
#### c) Usando o sinal de mais para adicionar camadas 
##### No ggplot2, você precisa adicionar um sinal de mais (“+”) ao seu código ao adicionar uma nova camada ao seu gráfico. Colocar o sinal de mais no lugar errado é um erro comum. O sinal de mais deve sempre ser colocado no final de uma linha de código e não no início de uma linha.
    Erro 1: 
    ggplot(dados = pinguins)
    + geom_point(mapping = aes(x = flipper_length_mm, y = body_mass_g))
    Mensagem de erro: Não é possível usar `+.gg()` com um único argumento. Você acidentalmente colocou + em uma nova linha? 
    
    Erro 2: 
    ggplot(dados = pinguins)%>%   
    geom_point(mapping = aes(x = flipper_length_mm, y = body_mass_g))
    Mensagem de erro: `data` deve ser um quadro de dados ou outro objeto coercível por `fortify()`, não um objeto S3 com classe gg/ggplot
    
    Código correto:
    ggplot(dados = pinguins) + 
    geom_point(mapping = aes(x = flipper_length_mm, y = body_mass_g))

#### <u>Suavização</u>

##### A suavização permite a detecção de uma tendência de dados mesmo quando você não consegue perceber facilmente uma tendência nos pontos de dados plotados. A funcionalidade de suavização do Ggplot2 é útil porque adiciona uma linha de suavização como outra camada a um gráfico; a linha de suavização ajuda os dados a fazerem sentido para um observador casual.
##### Tipo de suavização:
* Suavização de loess: O processo de suavização loess é melhor para suavizar gráficos com menos de 1.000 pontos.
        
        ggplot(dados, aes(x=, y=)) + geom_point() + geom_smooth(método="loess")
        
* Suavização de jogo: A suavização Gam, ou suavização generalizada de modelo aditivo, é útil para suavizar gráficos com um grande número de pontos. 
        
        ggplot(dados, aes(x=, y=)) + geom_point() + geom_smooth(método="gam", fórmula = y ~s(x))

#### Qual a diferença entre a visualização de dados em um arquivo R Markdown e a visualização de dados em um programa como o Tableau? 
#### O R Markdown é uma linguagem de marcação que mescla texto e código R, permitindo a criação de documentos dinâmicos que integram análises de dados, visualizações e explicação textual. As visualizações de dados são geradas por meio de scripts de programação, oferecendo um alto nível de personalização e controle sobre os elementos gráficos.
#### Por outro lado, o Tableau é uma ferramenta de business intelligence e visualização de dados que oferece uma interface intuitiva baseada em arrastar e soltar. Os usuários podem criar visualizações interativas sem a necessidade de escrever código, facilitando a exploração e a apresentação de dados de maneira mais acessível para usuários não técnicos.
#### Em resumo, a diferença está na abordagem técnica: o R Markdown é mais orientado para quem trabalha com análise de dados e programação, enquanto o Tableau é mais voltado para profissionais que buscam criar visualizações de dados sem conhecimentos avançados de programação. <br>
