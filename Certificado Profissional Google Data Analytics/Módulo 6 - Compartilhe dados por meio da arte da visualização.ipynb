## Visualizações de dados eficazes <br>
#### Uma boa maneira de pensar na visualização de dados é que ela pode ser a diferença entre a confusão total e a compreensão real de um problema. Criar visualizações de dados eficazes é uma tarefa complexa.<br>
#### Primeiro, você revisará duas estruturas que são úteis para pensar sobre como organizar as informações em sua visualização. As estruturas podem ajudá-lo a organizar seus pensamentos sobre a visualização de dados e fornecer uma lista de verificação útil para referência.
* Método McCandless: lista quatro elementos de uma boa visualização de dados: 
    1. Informações (dados): Os dados com os quais você está trabalhando. Sem informações ou dados, você não poderá comunicar suas descobertas com sucesso.
    2. História (conceito): Uma narrativa ou conceito claro e convincente. A história permite que você compartilhe seus dados de maneiras significativas e interessantes. Sem uma história, sua visualização é informativa, mas não muito inspiradora. 
    3. Meta/Objetivo (função) : Um objetivo ou função específica para o visual. O objetivo da visualização de dados torna os dados úteis e utilizáveis. Sem uma meta, sua visualização ainda pode ser informativa, mas não pode gerar insights acionáveis.
    4. Forma visual (metáfora): Um uso eficaz de metáfora ou expressão visual.é o que dá estrutura à visualização de dados e a torna bonita. Sem forma visual, seus dados ainda não são visualizados.

* Verificação Trifecta dos gráficos de lixo eletrônico de Kaiser Fung: Essa abordagem é um conjunto útil de perguntas que pode ajudar os consumidores de visualização de dados a criticar o que estão consumindo e determinar quão eficaz é. O Checkup tem três perguntas:
    1. Qual é a questão prática? 
    2. O que dizem os dados?
    3. O que o visual diz? 
#### Segundo, você explorará atributos pré-atentivos e como eles podem ser usados para afetar a forma como as pessoas pensam sobre suas visualizações.
#### Atributos pré-atentivos são os elementos de uma visualização de dados que as pessoas reconhecem automaticamente, sem esforço consciente. Os blocos de construção básicos e essenciais que tornam os recursos visuais imediatamente compreensíveis são chamados de marcas e canais. 
* Marcas são objetos visuais básicos como pontos, linhas e formas. Cada marca pode ser dividida em quatro qualidades:
    1. Posição - Onde uma marca específica está no espaço em relação a uma escala ou a outras marcas
    2. Tamanho - quão grande, pequena, longa ou alta é uma marca
    3. Forma - Se um objeto específico recebe uma forma que comunica algo sobre ele
    4. Cor – Qual é a cor da marca
* Canais são aspectos visuais ou variáveis que representam características dos dados. Canais são basicamente marcas usadas para visualizar dados. Os canais variam em termos de eficácia na comunicação de dados com base em três elementos: 
    1. Precisão – Os canais são úteis para estimar com precisão os valores representados? (Por exemplo, a cor é muito precisa ao comunicar diferenças categóricas, como maçãs e laranjas. Mas é muito menos eficaz ao distinguir dados quantitativos como 5 de 5,5.)
    2. Popout – Quão fácil é distinguir certos valores de outros? (Há muitas maneiras de chamar a atenção para partes específicas de um visual, e muitas delas aproveitam atributos pré-atentivos, como comprimento, tamanho, largura da linha, forma, delimitação, matiz e intensidade da linha.)
    3. Agrupamento – Quão bom é um canal para comunicar grupos existentes nos dados? (Considere a proximidade, semelhança, fechamento, conectividade e continuidade do canal.)

#### Princípios de design: <br>

#### Existem nove princípios básicos de design que os analistas de dados devem considerar ao construir suas visualizações.   
1. Equilíbrio : O design de uma visualização de dados é equilibrado quando os principais elementos visuais, como cor e forma, são distribuídos uniformemente. Isso não significa que você precisa de simetria completa, mas sua visualização não deve ter um lado distraindo o outro. Se a sua visualização de dados for equilibrada, isso pode significar que as linhas usadas para criar os gráficos têm comprimento semelhante em ambos os lados ou que o espaço entre os objetos é igual. 
2. Ênfase: Sua visualização de dados deve ter um ponto focal, para que seu público saiba onde se concentrar. Em outras palavras, suas visualizações devem enfatizar os dados mais importantes para que os usuários os reconheçam primeiro. Usar cor e valor é uma maneira eficaz de fazer isso acontecer. Ao usar cores contrastantes, você pode garantir que os elementos gráficos — e os dados mostrados nesses elementos — se destaquem. 
3. Movimento: Movimento pode se referir ao caminho que o olho do espectador percorre enquanto observa uma visualização de dados ou movimento literal criado por animações. O movimento na visualização de dados deve imitar a maneira como as pessoas costumam ler. Você pode usar linhas e cores para chamar a atenção do visualizador pela página. 
4. Padrão: você pode usar formas e cores semelhantes para criar padrões em sua visualização de dados. Isso pode ser útil de várias maneiras diferentes. Por exemplo, você pode usar padrões para destacar semelhanças entre diferentes conjuntos de dados ou quebrar um padrão com uma forma, cor ou linha exclusiva para dar mais ênfase.
5. Repetição: A repetição de tipos, formas ou cores de gráficos aumenta a eficácia da sua visualização.
6. Proporção: Proporção é outra forma de demonstrar a importância de determinados dados. Usar várias cores e tamanhos ajuda a demonstrar que você está chamando a atenção para um visual específico em detrimento de outros. Se você criar um gráfico em um painel maior que os outros, estará chamando a atenção para ele. É importante certificar-se de que cada gráfico reflita e visualize com precisão a relação entre os valores nele contidos.
7. Ritmo: Refere-se à criação de uma sensação de movimento ou fluxo em sua visualização. O ritmo está intimamente ligado ao princípio do movimento. Se o seu design finalizado não criar um fluxo com sucesso, você pode querer reorganizar alguns dos elementos para melhorar o ritmo.
8. Variedade: suas visualizações devem ter alguma variedade nos tipos de gráficos, linhas, formas, cores e valores que você usa. A variedade mantém o público envolvido. Mas é bom encontrar o equilíbrio, pois muita variedade pode confundir as pessoas.
9. Unidade: O último princípio é a unidade. Isso significa que a visualização final dos dados deve ser coesa. Se o visual estiver desarticulado ou mal organizado, será confuso e opressor.

#### Qual é a melhor maneira de contar a história com meus dados?
* Gráfico de linha: É usado para rastrear alterações em períodos curtos e longos de tempo. Quando existem alterações menores, é melhor usar gráficos de linhas do que gráficos de barras. Os gráficos de linhas também podem ser usados para comparar alterações no mesmo período de tempo para mais de um grupo. 
* Gráfico de colunas: Usam o tamanho para contrastar e comparar dois ou mais valores, usando altura ou comprimento para representar os valores específicos.  
* Mapa de calor: Semelhante aos gráficos de barras, os mapas de calor também usam cores para comparar categorias em um conjunto de dados. Eles são usados principalmente para mostrar relações entre duas variáveis e usam um sistema de codificação por cores para representar valores diferentes.
* Gráfico de pizza: É um gráfico circular dividido em segmentos que representam proporções correspondentes à quantidade que representa, principalmente quando se trata de partes de um todo.
* Gráfico de dispersão: Mostram relações entre diferentes variáveis. Os gráficos de dispersão são normalmente usados para duas variáveis de um conjunto de dados, embora variáveis adicionais possam ser exibidas.
* Gráfico de distribuição: exibe a distribuição de vários resultados em um conjunto de dados.

<br>

#### Revendo cada um desses exemplos visuais, onde você percebe que eles se enquadram em relação ao seu tipo de dados? Padrões significativos podem assumir várias formas, como:
* Mudança: Esta é uma tendência ou exemplo de observações que se tornam diferentes ao longo do tempo. Uma ótima maneira de medir alterações nos dados é por meio de um gráfico de linhas ou colunas.
* Clustering: Uma coleção de pontos de dados com valores semelhantes ou diferentes. Isso é melhor representado por meio de um gráfico de distribuição.
* Relatividade: São observações consideradas em relação ou em proporção a outra coisa. Você provavelmente já viu exemplos de dados da relatividade em um gráfico de pizza.
* Classificação: Esta é uma posição em uma escala de realização ou status. Os dados que requerem classificação são melhor representados por um gráfico de colunas.
* Correlação: Isso mostra um relacionamento mútuo ou conexão entre duas ou mais coisas. Um gráfico de dispersão é uma excelente forma de representar esse tipo de padrão de dados.

## Correlação e causalidade <br>
#### A correlação nas estatísticas é a medida do grau em que duas variáveis se movem em relação uma à outra. Um exemplo de correlação é a ideia de que “À medida que a temperatura sobe, as vendas de sorvetes também sobem”. É importante lembrar que correlação não significa que um evento cause outro. Mas indica que eles têm um padrão ou um relacionamento entre si. Se uma variável sobe e a outra variável também sobe, é uma correlação positiva. Se uma variável sobe e a outra desce, é uma correlação negativa ou inversa. Se uma variável subir e a outra permanecer praticamente a mesma, não há correlação. <br>
#### Causalidade refere-se à ideia de que um evento leva a um resultado específico. Por exemplo, quando um raio cai, ouvimos o trovão (onda sonora) causado pelo aquecimento e resfriamento do ar causado pelo raio. O relâmpago causa trovão. <br>
#### Por que é importante diferenciar entre correlação e causalidade? 
#### Ao tirar conclusões da análise de dados, você precisa ter certeza de não assumir uma relação causal entre elementos de seus dados quando há apenas uma correlação. Em sua análise de dados, lembre-se de: 
* Analise criticamente quaisquer correlações que você encontrar 
* Examine o contexto dos dados para determinar se uma causalidade faz sentido (e pode ser apoiada por todos os dados)
* Compreenda as limitações das ferramentas que você usa para análise

## Os dados crescem em árvores de decisão <br>
#### Uma árvore de decisão é uma ferramenta de tomada de decisão que permite que você, analista de dados, tome decisões com base em perguntas-chave que você pode fazer a si mesmo. Cada pergunta na árvore de decisão de visualização ajudará você a tomar uma decisão sobre recursos críticos para sua visualização. Abaixo está um exemplo de árvore de decisão básica para guiá-lo na tomada de uma decisão baseada em dados sobre qual visualização é a melhor maneira de contar sua história. Observe que existem muitos tipos diferentes de árvores de decisão que variam em complexidade e podem fornecer decisões mais aprofundadas. <br>

#### Comece com sua história: <br>
#### Comece avaliando o tipo de dados que você possui e faça uma série de perguntas para determinar a melhor fonte visual:
* Seus dados possuem apenas uma variável numérica? Se você tiver dados que possuem uma variável numérica contínua, um histograma ou gráfico de densidade são os melhores métodos para traçar seus dados categóricos. Dependendo do seu tipo de dados, um gráfico de barras pode até ser apropriado neste caso. 
* Existem vários conjuntos de dados? Para casos que lidam com mais de um conjunto de dados, considere um gráfico de linha ou de pizza para uma representação precisa dos seus dados. Um gráfico de linhas conectará vários conjuntos de dados em uma única linha contínua, mostrando como os números mudaram ao longo do tempo. Um gráfico de pizza é bom para dividir um todo em várias categorias ou partes. 
* Você está medindo as mudanças ao longo do tempo? Um gráfico de linhas geralmente é adequado para traçar tendências ao longo do tempo. Porém, quando as alterações são maiores, um gráfico de barras é a melhor opção.
* As relações entre os dados precisam ser mostradas? Quando você tem duas variáveis para um conjunto de dados, é importante apontar como uma afeta a outra. Variáveis que combinam bem são melhor plotadas em um gráfico de dispersão. No entanto, se houver muitos pontos de dados, a relação entre as variáveis pode ser obscurecida, de modo que um mapa de calor pode ser uma representação melhor nesse caso.

### Design thinking para melhoria de visualização: <br>
#### Quando você traz o design thinking para o seu trabalho, você está tentando identificar estratégias alternativas para suas visualizações que podem não ficar claras imediatamente.Você tem que desafiar seu próprio pensamento e explorar diferentes maneiras de abordar os problemas e encontrar soluções. Design thinking é um processo usado para resolver problemas complexos de forma centrada no usuário. A visualização de dados é o estágio perfeito do seuanálise para aplicar uma mentalidade baseada no usuário. <br>

#### Se você usar o design thinking ao planejar e criar sua visualização de dados, você estará tomar decisões com base nas necessidades das pessoas que as visualizarão. Dessa forma, seu público ficará envolvido e iluminado pela forma como você visualiza suas descobertas. Embora o processo de design thinking venha em muitas formas diferentes, todos eles têm estágios ou fases. No espírito do design thinking,essas fases não precisam seguir uma ordem definida. <br>

#### 5 fases do design thinking: 
* Na fase de empatia: Pensar nas emoções e necessidades do público-alvo para a visualização dos dados, sejam partes interessadas, membros da equipe ou o público em geral.
* Na fase de definição: Descobrir exatamente o que seu público precisa dos dados, ajuda a encontrar as necessidades, seus problemas e seus insights. Você poderia usar esta fase para pensar sobre quais dados mostrar em sua visualização.
* Na fase de idealização: Gerando ideias para visualização de dados. Isso pode envolver a criação de rascunhos de sua visualização com cores diferentes combinações ou talvez experimentando diferentes formas. Criar tantos exemplos quanto possível ajudará você a refinar suas ideias. A chave aqui é sempre lembrar do seu público ao ter ideias eestratégias. Você quer pensar em como pode posicionar suas visualizações para atenderas necessidades e expectativas do seu público.
* Na fase da protótipação: Reunindo visualizações para testes e feedback. Podem criar protótipos da próxima versão do painel com visualizações de dados novas e aprimoradas.
* Na fase de teste: Mostrar visualizações de protótipos às pessoas antes que as partes interessadas as vejam para revisão e aprovação. Faça melhorias incrementais nas visualizações de dados ao longo do tempo.

## Título, legendas e rótulos <br>
#### Títulos, legendas, rótulos e anotações ajudam você a transformar suas visualizações de dados em exibições mais significativas. Afinal, você deseja convidar o seu público para a sua apresentação e mantê-lo envolvido. Quando você apresenta uma visualização, eles devem ser capazes de processar e compreender as informações que você está tentando compartilhar nos primeiros cinco segundos. <br>
* Um título é uma linha de palavras impressas em letras grandes emparte superior da visualização para comunicar quais dados estão sendo apresentados.É o que chama a atenção que faz seu público deseja ler mais. É melhor manter as coisas simples. Deixe-o em negrito ou alguns tamanhos maiores queo resto do texto ecoloque-o diretamente acima do gráfico,alinhado à esquerda.
    * - Conteúdo : descreva resumidamente os dados - Comprimento : geralmente a largura do quadro de dados - Posição : acima dos dados
    * - Use linguagem breve - Não use letras maiúsculas - Não use itálico - Não use siglas - Não use abreviações - Não use humor ou sarcasmo
* Uma legenda apoia o título adicionando mais contexto e descrição.  Adicionar uma legenda ajudará o público a entender melhor os detalhes associados ao seu gráfico. Use um estilo de fonte que corresponda ao resto os elementos do gráfico e local o subtítulo diretamente abaixo do título.
    * - Conteúdo : Esclareça o contexto dos dados - Comprimento : igual ou menor que o título - Posição : Diretamente abaixo do título
    * - Use um tamanho de fonte menor que o do título - Não use palavras indefinidas - Não use letras maiúsculas, negrito ou itálico - Não use siglas - Não use abreviações
* Um rótulo em uma visualização identifica os dados em relação a outros dados. Os dados também podem ser rotulados diretamente em um gráfico, em vez de por meio de uma legenda do gráfico. Isso ocorre porque muitos gráfico suse propriedades visuais diferentes, como cores ou formas para representar diferentes valores de dados. Uma lenda ou chaveidentifica o significado de vários elementos emuma visualização de dados e pode ser usado comouma alternativa à rotulagem direta dos dados.
    * - Conteúdo : Substitui a necessidade de legendas - Comprimento : Geralmente menos de 30 caracteres - Posição : Próximo aos dados ou abaixo ou ao lado dos eixos
    * - Use apenas algumas palavras - Use códigos de cores bem pensados - Use textos explicativos para apontar para os dados - Não use letras maiúsculas, negrito ou itálico
* Uma anotação explica brevemente os dados ou ajuda a focar o público em um aspecto específico dos dados em uma visualização. 
    * - Conteúdo : Chama a atenção para determinados dados - Comprimento : Varia, limitado pelo espaço aberto - Posição : Imediatamente próximo aos dados anotados
    * - Não use letras maiúsculas, negrito ou itálico - Não use texto girado - Não distraia os espectadores dos dados

## Projetando um gráfico em 60 minutos <br>
#### Preparação (5 min): Crie o espaço mental e físico necessário para um ambiente de pensamento abrangente. Isso significa ter espaço para pensar em como deseja que seus dados apareçam, considerando a quantidade e o tipo de dados que você possui. <br>
#### Fale e ouça (15 min): Identifique o objeto do seu trabalho chegando ao “pergunta atrás da pergunta” e estabelecendo expectativas. Faça perguntas e concentre-se realmente no feedback das partes interessadas em relação aos seus projetos para ajudá-lo a aprimorar a forma de organizar seus dados. <br>
#### Esboço e design (20 min): Esboce sua abordagem para o problema. Defina o tempo e o resultado do seu trabalho para ter uma ideia clara e concisa do que você está elaborando. <br>

## Insights ao vivo e estáticos <br>
#### Identificar se os dados são ativos ou estáticos depende de alguns fatores:
* Qual é a idade dos dados?
* Quanto tempo até que os insights fiquem obsoletos ou não sejam mais válidos para a tomada de decisões?
* Esses dados ou análises precisam ser atualizados regularmente para permanecerem valiosos?

<br>

#### Os dados estáticos envolvem o fornecimento de capturas de tela ou instantâneos em apresentações ou a construção de painéis usando instantâneos de dados. <br>
#### PRÓS:
* Pode controlar rigorosamente uma narrativa pontual dos dados e insights
* Permite que análises complexas sejam explicadas em profundidade para um público maior
#### CONTRAS:
* O Insight imediatamente começa a perder valor e continua perdendo valor quanto mais tempo os dados permanecem em estado estático
* Os instantâneos não conseguem acompanhar o ritmo da mudança de dados

<br>

#### Dados dinâmicos significam que você pode criar painéis, relatórios e visualizações conectados a dados atualizados automaticamente. <br>
#### PRÓS:
* Os painéis podem ser criados para serem mais dinâmicos e escaláveis
* Fornece os dados mais atualizados às pessoas que precisam deles, no momento em que precisam.
* Permite visualizações atualizadas e selecionadas dos dados com a capacidade de criar uma “única fonte de verdade” escalável para vários casos de uso
* Permite que ações imediatas sejam tomadas em dados que mudam com frequência
* Alivia o tempo/recursos gastos em processos para cada análise
#### CONTRAS:
* Pode utilizar recursos de engenharia para manter os pipelines ativos e escaláveis, o que pode estar fora do escopo da alocação de recursos de dados de algumas empresas
* Sem a capacidade de interpretar os dados, você pode perder o controle da narrativa, o que pode causar caos nos dados (ou seja, equipes chegando a conclusões conflitantes com base nos mesmos dados).
* Pode causar falta de confiança se os dados não forem tratados adequadamente

<br>

#### Os analistas precisam se familiarizar com o negócio e os dados para que possam recomendar quando uma análise estática atualizada é necessária ou deve ser atualizada. Além disso, esse insight de dados o ajudará a definir quais tipos de análises, visualizações e dados adicionais são recomendados para os tipos de decisões que a empresa precisa tomar.

## Dicas de apresentação <br>
#### O método McCandless é um conjunto de diretrizes para apresentações. Sugere que você comece com ideias amplas e gerais e depois trabalhe nos detalhes. As etapas do método McCandless incluem:

* Dica 1: Conheça o seu fluxo
#### Assim como em qualquer boa história, uma história de dados deve ter um bom enredo (tema e fluxo), um bom diálogo (pontos de discussão) e um ótimo final ou grande revelação (resultados e conclusões). #### Para desenvolver o fluxo certo para sua apresentação, tenha em mente o seu público. Faça a si mesmo estas duas perguntas para ajudá-lo a definir o fluxo geral e construir sua apresentação. <br>
#### a) Quem é meu público?
#### b) Qual é o objetivo da minha apresentação?
* Se o objetivo da sua apresentação é solicitar ou recomendar algo no final, como um discurso de vendas, você pode fazer com que cada slide atinja as recomendações no final. 
* Se o objetivo da sua apresentação é focar nos resultados da sua análise, cada slide pode ajudar a marcar o caminho para os resultados. Certifique-se de incluir bastante localização atual (visualizações das etapas de análise de dados) para demonstrar o caminho que você seguiu com os dados.
* Se o objetivo da sua apresentação é fornecer um relatório sobre a análise de dados, os seus slides devem resumir claramente os seus dados e as principais conclusões. Nesse caso, não há problema em deixar os dados serem a estrela ou falarem por si.

* Dica 2: prepare pontos de discussão e limite o texto nos slides
#### Ao criar cada slide da sua apresentação, prepare pontos de discussão (também chamados de notas do orador ) sobre o que você dirá. Torne mais fácil para o seu público ler os slides enquanto ainda presta atenção ao que você está dizendo. Em geral, siga a regra dos cinco segundos. Seu público não deve gastar mais de cinco segundos lendo qualquer bloco de texto em um slide. <br>
#### Saber exatamente o que você dirá ao explicar cada slide da apresentação também cria um fluxo natural para sua história. Os pontos de discussão ajudam a evitar pausas estranhas entre os tópicos. Os slides que resumem os dados também podem ser repetitivos (e enfadonhos). Se você preparar uma variedade de pontos de discussão interessantes sobre os dados, poderá manter seu público alerta e prestando atenção aos dados e à sua análise.

#### b) Apresentando seus dados de forma eficiente. Certifique-se de que cada slide conte uma parte única e importante da sua história de dados. Se um slide não for tão único, você pode pensar em combinar as informações desse slide com outro slide. <br>

## Antecipe a pergunta <br>
#### Pense nessas coisas com antecedência e frequentemente ao planejar uma sessão de perguntas e respostas. Quando você se sentir confiante de que entende completamente as expectativas das partes interessadas e os objetivos do projeto, você pode começar a identificar possíveis perguntas. Uma ótima maneira de identificar dúvidas do público é fazer um teste de sua apresentação. Gosto de chamar isso de “teste do colega”. Mostre sua apresentação ouseus dados visualizados para um colega que tenha nenhum conhecimento prévio de seu trabalho e veja quais perguntas eles fazem. <br>
#### Finalmente, esteja preparado para considerar e descreva para suas partes interessadas quaisquer limitações em seus dados. Você pode fazer isso analisando criticamente os padrões que você descobriu em seus dados para integridade. <br>
#### Antes da apresentação
* Monte e prepare suas perguntas.
* Discuta sua apresentação com seu gerente, outros analistas ou outros contatos amigáveis em sua organização.
* Pergunte a um gerente ou a outros analistas que tipo de perguntas normalmente eram feitas por seu público específico no passado.
* Busque comentários, feedback e perguntas na apresentação ou no documento de sua análise.
* Pelo menos 24 horas antes da apresentação, tente debater questões complicadas ou partes pouco claras que você possa encontrar - isso ajuda a evitar surpresas.
* Nunca é demais praticar o que você vai apresentar, para dar conta de alguma informação faltante ou simplesmente para acalmar os nervos.

## Manipulação de objeções <br>
#### Ao fazer uma apresentação para um público, seu objetivo principal é responder a uma tarefa de negócios. Uma tarefa de negócios é uma questão ou problema para o qual você usa dados para resolver – e uma apresentação demonstra como resolvê-lo. As tarefas de negócios podem ter uma variedade de contextos e escopos, portanto os detalhes da sua apresentação dependerão de vários fatores. <br>

#### Às vezes, você poderá receber perguntas ou objeções sobre sua apresentação. Isso é normal, pois o seu público deseja entender a sua apresentação da forma mais completa possível. Responder a essas perguntas e objeções de maneira clara, concisa e educada é crucial para fazer uma apresentação eficaz.<br>
#### As partes interessadas podem levantar objeções durante ou após a sua apresentação.Geralmente, essas objeções são sobre os dados, sua análise ou suas descobertas.<br>

#### a) Objeções sobre os dados podem significar algumas coisas diferentes.Às vezes, as partes interessadas podem perguntar onde você obteve os dados ede quais sistemas vieram, ou eles podem querer saber quais transformaçõesaconteceu antes de você trabalhar com ele ou quão atualizados e precisos são seus dados.Você pode adicionar um detalhamento mais detalhado em seuapêndice caso haja mais perguntas.<br>

#### b) Eles podem querer saber se sua análise é reproduzível, entãoajuda manter um registro de alterações documentando as etapas executadas. Dessa forma, outra pessoapoderia acompanhar e reproduzir seu processo.Você pode até criar um slide na seção de apêndices da sua apresentaçãoexplicando essas etapas, se achar necessário.<br>

#### c) Muitas vezes serão perguntas como: "Essas descobertas existemem períodos anteriores ou você controlou as diferenças em seus dados?"Seu público quer ter certeza de que seus resultados finais foram consideradosquaisquer possíveis inconsistências e que sejam precisas e úteis.<br>

#### Primeiro,pode ser útil comunicar quaisquer suposições sobre os dados,sua análise ou suas descobertas que podem ajudar a responder às suas perguntas. Segundo, explique por que sua análise pode ser diferente do esperado.Oriente seu público pelas variáveis que mudam os resultados para ajudá-losentenda como você chegou lá.E terceiro, algumas objeções têm mérito,especialmente se eles mencionarem algo em que você não tinha pensado antes.Se isso for verdade, você pode reconhecer que essas objeções são válidas etomar medidas para investigar mais detalhadamente.
