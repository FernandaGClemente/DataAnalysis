## Integridade dos dados: <br> 
é a precisão, integridade,consistência e confiabilidade dedados ao longo de seu ciclo de vida. Quando a integridade dos dados é baixa,pode causar qualquer coisa, desde a perda de um único pixel emuma imagem a uma decisão médica incorreta.Em alguns casos,uma peça que falta pode tornar todos os seus dados inúteis.A integridade dos dados pode sercomprometida de muitas maneiras diferentes.Há uma chance de que os dados possam sercomprometida toda vez que é replicada,transferidos ou manipulados de qualquer forma.

<br>

A replicação de dados é o processo dearmazenamento de dados em vários locais.Se você estiver replicando dados emtempos diferentes em lugares diferentes,há uma chance de seus dados estarem fora de sincronia.Esses dados carecem de integridadeporque pessoas diferentes podem não serusando os mesmos dados parasuas descobertas, o que pode causar inconsistências.

<br>

Há também a questão da transferência de dados,qual é o processo de copiardados de um dispositivo de armazenamentopara a memória ou de um computador para outro.Se a transferência de dados for interrompida,você pode acabar com um conjunto de dados incompleto,que pode não ser útil para suas necessidades.

<br>

O processo de manipulação de dados envolve a mudançaos dados para torná-los mais organizados e fáceis de ler.A manipulação de dados destina-se a tornaro processo de análise de dados mais eficiente,mas um erro durante o processopode comprometer a eficiência.

<br>

Finalmente, os dados também podem ser comprometidospor erro humano, vírus,malware, hacking e falhas do sistema.

<br>
Uma boa análise depende da integridade dos dados, e a integridade dos dados geralmente depende do uso de um formato comum. Exemplo: 
Restrição de dados > Definição

* Tipo de dados > Os valores devem ser de um determinado tipo: data, número, porcentagem, booleano, etc.
* Intervalo de dados > Os valores devem estar entre os valores máximo e mínimo predefinidos
* Obrigatório > Os valores não podem ser deixados em branco ou vazios
* Exclusivo > Os valores não podem ter uma duplicata
* Padrões de expressão regular (regex) > Os valores devem corresponder a um padrão prescrito
* Validação entre campos > Certas condições para vários campos devem ser satisfeitas
* Chave primária > (Somente bancos de dados) o valor deve ser exclusivo por coluna
* Conjunto de membros > (Somente bancos de dados) os valores de uma coluna devem vir de um conjunto de valores discretos
* Chave estrangeira > (Somente bancos de dados) os valores de uma coluna devem ser valores exclusivos provenientes de uma coluna em outra tabela
* Precisão > O grau em que os dados estão em conformidade com a entidade real que está sendo medida ou descrita
* Integridade > O grau em que os dados contêm todos os componentes ou medidas desejados
* Consistência > O grau em que os dados são repetíveis a partir de diferentes pontos de entrada ou coleta

## Objetivos e dados bem alinhados: <br>
Você pode obter insights poderosos e tirar conclusões precisas quando os dados estiverem bem alinhados aos objetivos de negócios. Como analista de dados, o alinhamento é algo que você precisará julgar. Um bom alinhamento significa que os dados são relevantes e podem ajudá-lo a resolver um problema de negócios ou determinar um curso de ação para atingir um determinado objetivo de negócios.

<br>

Dados limpos + alinhamento ao objetivo de negócios + limpeza de dados adicionais + variáveis recém-descobertas + restrições = conclusões precisas 
* Quando há dados claros e um bom alinhamento, você pode obter insights precisos e tirar conclusões que os dados suportam.
* Se houver um bom alinhamento, mas os dados precisarem ser limpos, limpe os dados antes de executar sua análise. 
* Se os dados estiverem apenas parcialmente alinhados com um objetivo, pense em como você pode modificar o objetivo ou usar restrições de dados para garantir que o subconjunto de dados se alinhe melhor com o objetivo de negócios.

## O que fazer quando você encontra um problema com seus dados: <br>
#### Problema de dados 1: sem dados 

a) Soluções possíveis: Reúna os dados em pequena escala para realizar uma análise preliminar e, em seguida, solicite mais tempo para concluir a análise depois de coletar mais dados.
Exemplos de soluções na vida real: Se você estiver pesquisando funcionários sobre o que eles pensam sobre um novo plano de desempenho e bônus, use uma amostra para uma análise preliminar. Em seguida, peça mais 3 semanas para coletar os dados de todos os funcionários.

b) Soluções possíveis: Se não houver tempo para coletar dados, faça a análise usando dados proxy de outros conjuntos de dados. Esta é a solução alternativa mais comum.
Exemplos de soluções na vida real: Se você estiver analisando os horários de pico para passageiros, mas não tiver os dados de uma cidade específica, use os dados de outra cidade com tamanho e demografia semelhantes.

<br>

#### Problema de dados 2: poucos dados
a) Soluções possíveis: Faça a análise usando dados de proxy juntamente com dados reais.
Exemplos de soluções na vida real: Se você estiver analisando tendências para proprietários de golden retrievers, aumente seu conjunto de dados incluindo os dados de proprietários de labradores.  

b) Soluções possíveis: Ajuste sua análise para alinhar com os dados que você já possui.
Exemplos de soluções na vida real: Se faltarem dados para jovens de 18 a 24 anos, faça a análise, mas observe a seguinte limitação em seu relatório: esta conclusão aplica-se apenas a adultos de 25 anos ou mais . 

<br>

##### Problema de dados 3: dados errados, incluindo dados com erros
a) Soluções possíveis: Se você tiver os dados errados porque os requisitos foram mal compreendidos, comunique os requisitos novamente.
Exemplos de soluções na vida real: Se você precisa dos dados para eleitores do sexo feminino e recebeu os dados dos eleitores do sexo masculino, reafirme suas necessidades.

b) Soluções possíveis: Identifique erros nos dados e, se possível, corrija-os na fonte procurando um padrão nos erros.
Exemplos de soluções na vida real: Se seus dados estiverem em uma planilha e houver uma instrução condicional ou booleana causando erros nos cálculos, altere a instrução condicional em vez de apenas corrigir os valores calculados.

c) Soluções possíveis: Se você não puder corrigir os erros de dados sozinho, poderá ignorar os dados errados e prosseguir com a análise se o tamanho da amostra ainda for grande o suficiente e ignorar os dados não causará viés sistemático.
Exemplos de soluções na vida real: Se o seu conjunto de dados foi traduzido de um idioma diferente e algumas das traduções não fazem sentido, ignore os dados com tradução ruim e prossiga com a análise dos outros dados.

<br>

## Calculando o tamanho da amostra: <br>

O objetivo é obter informações suficientes de um pequeno grupo dentro uma população para fazer previsões ou conclusões sobre toda a população. O tamanho da amostra ajuda garantir o grau em que você pode ser confiante de que suas conclusões representar com precisão a população.

<br>

Viés de amostragem é quando uma amostra não representa a população como um todo. Isso significa que alguns membros da população estão sendo super-representados ou sub-representados.

<br>
 
O uso de amostragem aleatória pode ajudar abordar alguns desses problemas com viés de amostragem. A amostragem aleatória é uma forma de selecionando uma amostra de uma população para que todo tipo possível da amostra tem uma chance igual de ser escolhida.

<br>

#### Terminologia: 
* População: Todo o grupo em que você está interessado para o seu estudo. Por exemplo, se você estiver pesquisando pessoas em sua empresa, a população seria todos os funcionários de sua empresa.
* Amostra: Um subconjunto de sua população. Assim como uma amostra de comida, ela é chamada de amostra porque é apenas um sabor. Portanto, se sua empresa for muito grande para pesquisar cada indivíduo, você pode pesquisar uma amostra representativa de sua população.
* Margem de erro: Como uma amostra é usada para representar uma população, espera-se que os resultados da amostra sejam diferentes do que seria se você tivesse pesquisado toda a população. Essa diferença é chamada de margem de erro. Quanto menor a margem de erro, mais próximos os resultados da amostra estarão do que seria se você tivesse pesquisado toda a população. 
* Nível de confiança: Quão confiante você está nos resultados da pesquisa. Por exemplo, um nível de confiança de 95% significa que, se você executar a mesma pesquisa 100 vezes, obterá resultados semelhantes 95 dessas 100 vezes. O nível de confiança é direcionado antes de você iniciar seu estudo porque afetará o tamanho de sua margem de erro no final de seu estudo. 
* Intervalo de confiança: A faixa de valores possíveis que o resultado da população estaria no nível de confiança do estudo. Este intervalo é o resultado da amostra +/- a margem de erro.
* Significado estatístico: A determinação de se o seu resultado pode ser devido ao acaso ou não. Quanto maior a significância, menos devido ao acaso.

<br>

#### Coisas para lembrar ao determinar o tamanho da sua amostra:

* Não use um tamanho de amostra menor que 30. Está estatisticamente comprovado que 30 é o menor tamanho de amostra onde um resultado médio de uma amostra começa a representar o resultado médio de uma população.
    * À medida que o tamanho da amostra aumenta, os resultados se assemelham mais à distribuição normal (em forma de sino) de um grande número de amostras. Uma amostra de 30 é o menor tamanho de amostra para o qual o CLT ainda é válido. 
    * https://www.investopedia.com/terms/c/central_limit_theorem.asp
* O nível de confiança mais comumente usado é 95%, mas 90% pode funcionar em alguns casos. 
* Para um nível de confiança mais alto , use um tamanho de amostra maior
* Para diminuir a margem de erro, use um tamanho de amostra maior
* Para maior significância estatística, use um tamanho de amostra maior

<br>

* Os tamanhos das amostras variam de acordo com o problema de negócios
* OS Tamanhos de amostra maiores têm um custo mais alto
* Saber o básico é útil > https://www.statisticssolutions.com/dissertation-resources/sample-size-calculation-and-sample-size-justification/sample-size-formula/

<br>

####  Uma calculadora de tamanho de amostra informa quantas pessoas você precisa entrevistar (ou coisas que precisa testar) para obter resultados que representem a população-alvo.  <br>
#### Existem calculadoras úteis online que podem ajudá-lo a encontrar o tamanho da amostra.Você precisa inserir o nível de confiança,tamanho da população e margem de erro. <br>
#### 1 - O nível de confiança é a probabilidade de que sua amostra reflete com precisão a população maior. Ter um nível de confiança de 99% é o ideal.Mas a maioria das indústrias espera pelo menos umNível de confiança de 90 ou 95 por cento. 
#### 2 - Margem de erro : A quantidade máxima que se espera que os resultados da amostra sejam diferentes daqueles da população real. Quanto mais próxima de zero a margem de erro, mais próximos serão seus resultados de suaamostra corresponderia aos resultados da população geral. Mais tecnicamente, a margem de erro define uma faixa de valores abaixo e acima do resultado médio da amostra. Espera-se que o resultado médio para toda a população esteja dentro desse intervalo. 
#### 3 - População : Este é o número total do qual você espera extrair sua amostra.
#### 4 - Amostra : Uma parte de uma população que é representativa da população.
#### 5 - Taxa de resposta estimada : se você estiver realizando uma pesquisa com indivíduos, essa é a porcentagem de pessoas que você espera que concluam sua pesquisa entre as que receberam a pesquisa.
* Depois de inserir suas informações em uma dessas calculadoras, ela fornecerá um tamanho de amostra recomendado. Lembre-se de que o tamanho da amostra calculado é o número mínimo para atingir o que você insere para nível de confiança e margem de erro. Se você estiver trabalhando com uma pesquisa, também precisará pensar na taxa de resposta estimada para descobrir quantas pesquisas precisará enviar. Por exemplo, se você precisar de uma amostra de 100 indivíduos e sua taxa de resposta estimada for de 10%, será necessário enviar sua pesquisa para 1.000 indivíduos para obter as 100 respostas necessárias para sua análise.
* Para outros estudos,as organizações podem apenas precisar saber que os resultados do teste ou pesquisa tê-los indo na direção certa.
* Fórmula do excel:

        =SEERRO(ARREDONDAR.PARA.CIMA((((PROCV(D6,'Pesquisa ZScore'!A51:C102,3,FALSO))^2*0.5*(1-0.5))/(F6/100)^2)/(1+((PROCV(D6,'Pesquisa ZScore'!A51:C102,3,FALSO))^2*0.5*(1-0.5))/((F6/100)^2*B6)), 0), "Erro (verificar entradas)")
        
* D6 = Nível de confiança (%)
* F6 = Margem de erro (%)
* B6 = Tamanho da população

## Pré-limpeza: <br>

* Ter dados suficientes: Dados insuficientes vem de apenas uma fonte, atualiza continuamente e estão incompletos, desatualizados e geograficamente limitados. Para lidar com dados insuficientes, pode identificar tendencias dentro dos dados disponiveis, aguardar mais dados se o tempo permitir, discutir com as partes interessadas e ajuste seus objetivos e pesquisar um novo conjunto de dados.

E saber quando parar de coletar dados permitirá que você conclua suas tarefas em tempo hábil sem sacrificar a integridade dos dados.

<br>

Se você não conectar os objetivos com os dados, sua análise pode não ser relevante para as partes interessadas. Finalmente, não entender quando parar de coletar dados pode levar a atrasos desnecessários na conclusão de tarefas. Ao concluir as atividades de pré-limpeza, você evita esses problemas.

## Por que a limpeza de dados é importante: <br>
* Dados sujos são dados incompletos, incorretos ou irrelevante para o problema que você está tentando resolver.
* Quando você trabalha com dados sujos, não pode ter certeza de que seus resultados estão corretos. Na verdade, você pode apostar que não serão.
* Ao trabalhar com dados limpos, você descobrirá que seus projetos funcionam com muito mais facilidade.
* Dados limpos são incrivelmenteimportante para uma análise eficaz.Se um dado for inserido emuma planilha ou banco de dados incorretamente,ou se for repetido,ou se um campo for deixado em branco,ou se os formatos de dados forem inconsistentes,o resultado são dados sujos. Pequenos erros podem levar agrandes consequências a longo prazo.
* Isso permite que você entenda e analiseinformações e identificar padrões importantes,conectar informações relacionadas e tirar conclusões úteis. 

<br>

Engenheiros de dados transformam dados emum formato útil paraanálise e dar-lhe uma infra-estrutura confiável.Isso significa que eles desenvolvem, mantêm,e bancos de dados de teste, processadores de dados e sistemas relacionados.

<br>

Os especialistas em armazenamento de dados desenvolvem processose procedimentos para armazenar e organizar dados de forma eficaz.Eles garantem que os dados estejam disponíveis,seguro e com backup para evitar perdas.

<br>

É importante lembrar:nenhum conjunto de dados é perfeito.É sempre uma boa ideia examinar elimpe os dados antes de iniciar a análise.

Um nulo é uma indicação de queum valor não existe em um conjunto de dados.Observe que não é o mesmo que um zero.No caso de uma pesquisa,um nulo significaria que os clientes ignoraram essa pergunta.Um zero significaria que eles forneceram zero como resposta.Para fazer sua análise,você primeiro precisaria limpar esses dados.O primeiro passo seria decidir o que fazer com esses valores nulos. Você pode filtrá-los ecomunicar que agora você tem um tamanho de amostra menor,ou você pode mantê-los e aprender com o fatoque os clientes não forneceram respostas.

#### Tipos de dados sujos: <br>
* 1 - Dados duplicados
    * Descrição: Qualquer registro de dados que apareça mais de uma vez
    * Causas Possíveis: Entrada manual de dados, importação de dados em lote ou migração de dados
    * Danos potenciais para as empresas: Métricas ou análises distorcidas, contagens ou previsões exageradas ou imprecisas ou confusão durante a recuperação de dados
* 2 - Dados desatualizados
    * Descrição: Quaisquer dados antigos que devem ser substituídos por informações mais recentes e precisas
    * Causas Possíveis: Pessoas mudando de função ou empresa, ou software e sistemas se tornando obsoletos
    * Danos potenciais para as empresas: Informações, tomadas de decisão e análises imprecisas
* 3 - Dados incompletos
    * Descrição: Quaisquer dados que estejam faltando campos importantes
    * Causas Possíveis: Coleta de dados imprópria ou entrada de dados incorreta
    * Danos potenciais para as empresas: Diminuição da produtividade, percepções imprecisas ou incapacidade de concluir serviços essenciais
* 4 - Dados incorretos/imprecisos
    * Descrição: Quaisquer dados completos, mas imprecisos
    * Causas Possíveis: Erro humano inserido durante a entrada de dados, informações falsas ou dados fictícios
    * Danos potenciais para as empresas: Insights imprecisos ou tomada de decisões com base em informações incorretas, resultando em perda de receita
* 5 - Dados inconsistentes
    * Descrição: Quaisquer dados que usam formatos diferentes para representar a mesma coisa
    * Causas Possíveis: Dados armazenados incorretamente ou erros inseridos durante a transferência de dados
    * Danos potenciais para as empresas: Pontos de dados contraditórios que levam à confusão ou incapacidade de classificar ou segmentar clientes

<br>

Isso inclui erros de ortografia e outros erros de texto,rótulos inconsistentes, formatos e faixa de campo,dados ausentes e duplicados.
Erros ortográficos, variações ortográficas, letras misturadas,pontuação inconsistente e erros de digitação em geral,acontecer quando alguém digita um dado incorretamente.

<br>

A validação de dados é uma ferramenta para verificara precisão e a qualidadedados antes de adicioná-los ou importá-los.A validação de dados é uma forma de limpeza de dados
* Validade (válidos): O conceito de usar princípios de integridade de dados para garantir que as medidas estejam em conformidade com as regras ou restrições de negócios definidas. Ex: Data collected five years ago used technology that is not approved or supported by the business
* Consistência (consistentes): O grau em que um conjunto de medidas é equivalente entre os sistemas. Ex: Date of store opening stored in both MM/DD/YYYY and MM/YY formats
* Precisão (Precisos): O grau de conformidade de uma medida com um padrão ou um valor verdadeiro. Ex: Addresses in the business database are identified as incorrect when compared to the public postal service database
* Completude (completos): O grau em que todas as medidas necessárias são conhecidas. Ex: NULL/missing value for the item “Number of employees per store” 

## Ferramentas e técnicas de limpeza de dados: <br>
* Remover dados indesejados, limpar o texto para remover espaços e espaços em branco extras, corrija erros de digitação e torne a formatação consistente.
* Antes de remover dados indesejados,é sempre uma boa prática fazer uma cópia do conjunto de dados.
* Normalmente, as duplicatas aparecem quando você combina conjuntos de dados de mais deuma fonte ou usando dados de vários departamentos dentro do mesmo negócio.
* A remoção de dados irrelevantes leva um pouco mais de tempo e esforço porque você tempara descobrir a diferença entre os dados que você precisa e os dados que você não precisa.Mas acredite em mim,tomar essas decisões economizará muito esforço no futuro.
* Espaços extras podem causar resultados inesperados ao classificar, filtrar ou pesquisar através de seus dados. Para remover esses espaços indesejados ou células em branco, você mesmo pode excluí-los. Ou, novamente, você pode confiar em suas planilhas,que oferecem muitas funções excelentes para remover espaços ou espaços em branco automaticamente.
* A próxima etapa de limpeza de dados envolve a correção de erros ortográficos,letras maiúsculas inconsistentes, pontuação incorreta e outros erros de digitação.
* O próximo passo é remover a formatação. Isso é particularmente importante quando você obtém dados de várias fontes diferentes.Cada banco de dados tem sua própria formatação,o que pode fazer com que os dados pareçam inconsistentes.

## Erros comuns a evitar: <br>
* Não verificar erros de ortografia : erros de ortografia podem ser tão simples quanto erros de digitação ou entrada. Na maioria das vezes, a ortografia incorreta ou erros gramaticais comuns podem ser detectados, mas fica mais difícil com coisas como nomes ou endereços. Por exemplo, se você estiver trabalhando com uma tabela de planilha de dados do cliente, poderá encontrar um cliente chamado “John” cujo nome foi inserido incorretamente como “Jon” em alguns lugares. A verificação ortográfica da planilha provavelmente não sinalizará isso, portanto, se você não verificar novamente os erros de ortografia e detectá-los, sua análise terá erros. 

* Esquecer-se de documentar erros : documentar seus erros pode economizar muito tempo, pois ajuda a evitar esses erros no futuro, mostrando como você os resolveu. Por exemplo, você pode encontrar um erro em uma fórmula em sua planilha. Você descobre que algumas das datas em uma de suas colunas não foram formatadas corretamente. Se você anotar essa correção, poderá consultá-la na próxima vez que sua fórmula for quebrada e obter uma vantagem na solução de problemas. Documentar seus erros também ajuda a acompanhar as mudanças em seu trabalho, para que você possa voltar atrás se uma correção não funcionou. 

* Não verificando valores incorretos : Um valor incorreto ocorre quando os valores são inseridos no campo errado. Esses valores ainda podem estar formatados corretamente, o que os torna mais difíceis de detectar se você não for cuidadoso. Por exemplo, você pode ter um conjunto de dados com colunas para cidades e países. Esses são os mesmos tipos de dados, portanto, são fáceis de confundir. Mas se você estivesse tentando localizar todas as instâncias da Espanha na coluna do país e a Espanha tivesse sido inserida por engano na coluna da cidade, você perderia os principais pontos de dados. Certificar-se de que seus dados foram inseridos corretamente é a chave para uma análise precisa e completa. 

* Negligenciar valores ausentes : valores ausentes em seu conjunto de dados podem criar erros e fornecer conclusões imprecisas. Por exemplo, se você estivesse tentando obter o número total de vendas dos últimos três meses, mas faltasse uma semana de transações, seus cálculos seriam imprecisos. Como prática recomendada, tente manter seus dados o mais limpos possível, mantendo a integridade e a consistência.

* Olhando apenas para um subconjunto dos dados : é importante pensar em todos os dados relevantes ao limpar. Isso ajuda a garantir que você entenda toda a história que os dados estão contando e que esteja prestando atenção a todos os possíveis erros. Por exemplo, se você estiver trabalhando com dados sobre padrões de migração de pássaros de diferentes fontes, mas limpar apenas uma fonte, talvez não perceba que alguns dos dados estão sendo repetidos. Isso causará problemas em sua análise mais tarde. Se você quiser evitar erros comuns como duplicatas, cada campo de seus dados requer atenção igual.

* Perder o controle dos objetivos de negócios : ao limpar dados, você pode fazer descobertas novas e interessantes sobre seu conjunto de dados, mas não deseja que essas descobertas o distraiam da tarefa em questão. Por exemplo, se você estiver trabalhando com dados meteorológicos para encontrar o número médio de dias chuvosos em sua cidade, também poderá observar alguns padrões interessantes sobre a queda de neve. Isso é realmente interessante, mas não está relacionado à pergunta que você está tentando responder agora. Ser curioso é ótimo! Mas tente não deixar que isso o distraia da tarefa em questão.  

* Não corrigir a origem do erro: corrigir o erro em si é importante. Mas se esse erro for realmente parte de um problema maior, você precisará encontrar a origem do problema. Caso contrário, você terá que continuar corrigindo o mesmo erro repetidamente. Por exemplo, imagine que você tem uma planilha de equipe que acompanha o progresso de todos. A tabela continua quebrando porque pessoas diferentes estão inserindo valores diferentes. Você pode continuar corrigindo todos esses problemas, um por um, ou pode configurar sua tabela para agilizar a entrada de dados para que todos estejam na mesma página. Abordar a origem dos erros em seus dados economizará muito tempo a longo prazo. 

* Não analisar o sistema antes da limpeza de dados: Se quisermos limpar nossos dados e evitar erros futuros, precisamos entender a causa raiz de seus dados sujos. Imagine que você é um mecânico de automóveis. Você descobriria a causa do problema antes de começar a consertar o carro, certo? O mesmo vale para os dados. Primeiro, você descobre de onde vêm os erros. Talvez seja por um erro de entrada de dados, por não configurar uma verificação ortográfica, falta de formatos ou duplicatas. Então, depois de entender de onde vêm os dados ruins, você pode controlá-los e manter seus dados limpos.

* Não fazer backup de seus dados antes da limpeza de dados : é sempre bom ser proativo e criar seu backup de dados antes de iniciar a limpeza de dados. Se o seu programa travar ou se suas alterações causarem um problema em seu conjunto de dados, você sempre poderá voltar para a versão salva e restaurá-la. O procedimento simples de fazer backup de seus dados pode economizar horas de trabalho e, o mais importante, uma dor de cabeça. 

* Não contabilizar a limpeza de dados em seus prazos/processos : todas as coisas boas levam tempo, e isso inclui a limpeza de dados. É importante ter isso em mente ao analisar seu processo e observar seus prazos. Quando você reserva um tempo para a limpeza de dados, isso ajuda a obter uma estimativa mais precisa de ETAs para as partes interessadas e pode ajudá-lo a saber quando solicitar um ETA ajustado.

## Limpeza de dados em planilhas: <br>
As planilhas são uma parte muito importante da análise de dados.Eles economizam tempo e esforço dos analistas de dadose nos ajude a eliminar erros todos os dias.
* Uma string de texto é um grupo de caracteres dentro de uma célula,na maioria das vezes composto de letras.
* Split é uma ferramenta que divide uma string de texto emo caractere especificado e colocacada fragmento em uma célula nova e separada. Dividir texto em colunas também é útil paracorrigir instâncias de números armazenados como texto.
* Uma função é um conjunto de instruções queexecuta um cálculo específicousando os dados em uma planilha.
* CONT.SE é uma função que retornao número de células que correspondem a um valor especificado.
* LEN é uma função que informa o comprimento dea cadeia de texto contandoo número de caracteres que contém.
* ESQUERDA é uma função que lhe dáum número definido de caracteres deo lado esquerdo de uma string de texto.
* DIREITA é uma função que lhe dáum número definido de caracteres deo lado direito de uma string de texto.
* MID é uma função que lhe dáum segmento do meio de uma string de texto.
* TRIM é uma função que remove entrelinhas,à direita e espaços repetidos nos dados.

##### Sistema de controle de versão
##### Veja como um sistema de controle de versão afeta uma alteração em uma consulta:
* Uma empresa possui versões oficiais de consultas importantes em seu sistema de controle de versão.
* Um analista garante que a versão mais atualizada da consulta seja a que eles alterarão. Isso se chama sincronização 
* O analista faz uma alteração na consulta.
* O analista pode pedir a alguém para revisar essa mudança. Isso é chamado de revisão de código e pode ser feito informal ou formalmente. Uma revisão informal pode ser tão simples quanto pedir a um analista sênior para dar uma olhada na mudança.
* Depois que um revisor aprova a alteração, o analista envia a versão atualizada da consulta para um repositório no sistema de controle de versão da empresa. Isso é chamado de confirmação de código. Uma prática recomendada é documentar exatamente qual foi a alteração e por que ela foi feita em uma área de comentários. Voltando ao nosso exemplo de consulta que extrai receita diária, um comentário pode ser: Receita atualizada para incluir a receita proveniente do novo produto, Calypso .
* Depois que a alteração for enviada , todos os outros na empresa poderão acessar e usar essa nova consulta ao sincronizar com as consultas mais atualizadas armazenadas no sistema de controle de versão.
* Se a consulta tiver um problema ou as necessidades de negócios forem alteradas, o analista poderá desfazer a alteração na consulta usando o sistema de controle de versão. O analista pode consultar uma lista cronológica de todas as alterações feitas na consulta e quem fez cada alteração. Então, depois de encontrar sua própria mudança, o analista pode reverter para a versão anterior.
* A consulta volta ao que era antes de o analista fazer a alteração. E todos na empresa também veem essa consulta original invertida.

## Glossário: <br>
* Abertura: O aspecto da ética de dados que promove o livre acesso, uso e compartilhamento de dados
* Agenda: Uma lista de compromissos agendados
* Alça de preenchimento: Uma caixa no canto inferior direito de uma célula selecionada da planilha que pode ser arrastada pelas células vizinhas para continuar uma instrução
* Algoritmo: Um processo ou conjunto de regras seguido para uma tarefa específica
* Amostra: Na análise de dados, um segmento de uma população que é representativo de toda a população
* Amostragem aleatória: Uma maneira de selecionar uma amostra de uma população de modo que todos os tipos possíveis de amostra tenham a mesma chance de serem escolhidos
* Amostragem imparcial: Quando a amostra da população que está sendo medida é representativa da população como um todo
* Análise de dados: A ciência dos dados
* Análise de dados: A coleta, transformação e organização de dados para tirar conclusões, fazer previsões e conduzir a tomada de decisões informadas
* Análise de lacunas: Um método para examinar e avaliar o estado atual de um processo, a fim de identificar oportunidades de melhoria no futuro
* Analista de informações :Alguém que coleta, transforma e organiza dados para tirar conclusões, fazer previsões e conduzir a tomada de decisões informadas
* Anonimização de dados: O processo de proteção dos dados privados ou confidenciais das pessoas, eliminando as informações de identificação
* Arquivo CSV (valores separados por vírgula):Um arquivo de texto delimitado que usa uma vírgula para separar valores
* Arquivo de áudio: Armazenamento de áudio digitalizado geralmente em MP3, AAC ou outro formato compactado
* Arquivo de vídeo: Uma coleção de imagens, arquivos de áudio e outros dados geralmente codificados em um formato compactado, como MP4, MV4, MOV, AVI ou FLV
