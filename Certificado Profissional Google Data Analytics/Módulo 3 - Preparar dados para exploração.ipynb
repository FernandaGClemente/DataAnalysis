Se você não coletar os dados usando seus próprios recursos, poderá obter dados de provedores de dados terceirizados ou terceirizados. ## Formatos de dados na prática <br>
### Dados primários x Dados secundários
* Coletado por um pesquisador de fontes em primeira mão
* Reunidos por outras pessoas ou de outras pesquisas
### Dados internos x Dados externos
* Dados que vivem dentro dos próprios sistemas de uma empresa
* Dados que vivem fora de uma empresa ou organização
### Dados contínuos x Dados discretos
* Dados que são medidos e podem ter quase qualquer valor numérico (Altura, Marcadores de tempo de execução em um vídeo e Temperatura)
* Dados que são contados e têm um número limitado de valores (Número de pessoas que visitam um hospital diariamente, Lotação máxima permitida da sala ou Bilhetes vendidos no mês corrente)
### Dados qualitativo x Dados quantitativo
* Medidas subjetivas e explicativas de qualidades e características
* Medidas específicas e objetivas de fatos numéricos
### Dados Nominal x Dados Ordinal
* Um tipo de dados qualitativos que não são categorizados com uma ordem definida (- Cliente pela primeira vez, cliente recorrente, cliente regular - Novo candidato a emprego, candidato existente, candidato interno - Nova listagem, listagem de preço reduzido, encerramento)
* Um tipo de dados qualitativos com uma ordem ou escala definida (- Classificações de filmes (número de estrelas: 1 estrela, 2 estrelas, 3 estrelas) - Seleções de votação de escolha classificada (1º, 2º, 3º) - Nível de renda (renda baixa, renda média, renda alta))
### Dados estruturados x Dados não estruturados
* Dados organizados em um determinado formato, como linhas e colunas
* Dados que não são organizados de maneira facilmente identificável

## A estrutura de dados <br>
### Duas categorias gerais de dados são: 
* Dados estruturados (Structured Data): organizados em um determinado formato, como linhas e colunas.
    * Tipos de dados definidos
    * Na maioria das vezes dados quantitativos
    * Fácil de organizar
    * Fácil de pesquisar
    * Fácil de analisar
    * Armazenados em bancos de dados relacionais e armazéns de dados
    * Contidos em linhas e colunas
    * Exemplos: Excel, Google Sheets, SQL, Dados do cliente, Registros telefônicos e Histórico de transações
* Dados não estruturados (Unstructured Data): não organizados de maneira fácil de identificar.
    * Tipos de dados variados
    * Na maioria das vezes dados qualitativos
    * Difícil de pesquisar
    * Oferece mais liberdade para análise
    * Armazenados em data lakes, data warehouses e bancos de dados NoSQL
    * Não pode ser colocado em linhas e colunas
    * Exemplos: mensagens de texto, comentários de mídia social, transcrições de chamadas telefônicas, vários arquivos de log, imagens, áudio e vídeo

## Níveis e técnicas de modelagem de dados <br>
### Os modelos de dados ajudam a manter os dados consistentes e permitem que as pessoas mapeiem como os dados são organizados. Um entendimento básico torna mais fácil para os analistas e outras partes interessadas entender seus dados e usá-los da maneira certa. <br>
### O que é modelagem de dados?
A modelagem de dados é o processo de criação de diagramas que representam visualmente como os dados são organizados e estruturados. Essas representações visuais são chamadas de modelos de dados . Você pode pensar na modelagem de dados como a planta de uma casa. A qualquer momento, pode haver eletricistas, carpinteiros e encanadores usando esse projeto. Cada um desses construtores tem uma relação diferente com a planta, mas todos precisam dela para entender a estrutura geral da casa. Os modelos de dados são semelhantes; usuários diferentes podem ter necessidades de dados diferentes, mas o modelo de dados fornece a eles uma compreensão da estrutura como um todo. 

<br>

### Níveis de modelagem de dados
Cada nível de modelagem de dados tem um nível diferente de detalhe.
* A modelagem de dados conceituais oferece uma visão de alto nível da estrutura de dados, por exemplo, como os dados interagem em uma organização. Por exemplo, um modelo de dados conceitual pode ser usado para definir os requisitos de negócios para um novo banco de dados. Um modelo de dados conceitual não contém detalhes técnicos.
* A modelagem lógica de dados concentra-se nos detalhes técnicos de um banco de dados, como relacionamentos, atributos e entidades. Por exemplo, um modelo de dados lógicos define como os registros individuais são identificados exclusivamente em um banco de dados. Mas não explica os nomes reais das tabelas do banco de dados. Esse é o trabalho de um modelo de dados físico.
* A modelagem de dados físicos descreve como um banco de dados opera. Um modelo de dados físico define todas as entidades e atributos usados; por exemplo, inclui nomes de tabelas, nomes de colunas e tipos de dados para o banco de dados.
> https://www.1keydata.com/datawarehousing/data-modeling-levels.html

<br>

### Técnicas de modelagem de dados
Existem muitas abordagens quando se trata de desenvolver modelos de dados, mas dois métodos comuns são o diagrama Entity Relationship Diagram (ERD) e o diagrama Unified Modeling Language (UML) . ERDs são uma maneira visual de entender o relacionamento entre entidades no modelo de dados. Os diagramas UML são diagramas muito detalhados que descrevem a estrutura de um sistema mostrando as entidades, atributos, operações e seus relacionamentos do sistema. Como analista de dados júnior, você precisará entender que existem diferentes técnicas de modelagem de dados, mas, na prática, provavelmente usará a técnica existente em sua organização.

## Compreendendo a lógica booleana <br>
As condições são criadas com operadores booleanos, incluindo AND, OR e NOT. Esses operadores são semelhantes aos operadores matemáticos e podem ser usados para criar instruções lógicas que filtram seus resultados. Os analistas de dados usam declarações booleanas para realizar uma ampla gama de tarefas de análise de dados, como criar consultas para pesquisas e verificar as condições ao escrever o código de programação. 

<br>

### Exemplo de lógica booleana
Imagine que você está comprando sapatos e está considerando certas preferências:
* Você comprará os sapatos apenas se forem rosa e cinza
* Você comprará os sapatos se forem totalmente rosa ou totalmente cinza, ou se forem rosa e cinza
* Você comprará os sapatos se forem cinza, mas não se forem rosa

<br>

### Operador AND
Sua condição é “Se a cor do sapato tiver qualquer combinação de cinza e rosa, você os comprará”. A instrução booleana quebraria a lógica dessa instrução para filtrar seus resultados por ambas as cores. Ele diria “SE (Color=”Grey”) E (Color=”Pink”) então compre-os.” O operador AND permite empilhar várias condições. 

<br>

### Operador OU
O operador OR permite que você avance se uma das duas condições for atendida. Sua condição é “Se os sapatos forem cinza ou rosa, você os comprará”. A declaração booleana seria “SE (Color=”Grey”) OU (Color=”Pink”) então compre-os.” 

<br>

### Operador NÃO
O operador NOT permite filtrar subtraindo condições específicas dos resultados. Sua condição é "Você comprará qualquer sapato cinza, exceto aqueles com qualquer vestígio de rosa". Sua declaração booleana seria “IF (Color="Grey") AND (Color=NOT "Pink") então compre-os."

## Transformando dados <br>
### O que é transformação de dados?
A transformação de dados é o processo de alteração do formato, estrutura ou valores dos dados. Como analista de dados, há uma boa chance de você precisar transformar os dados em algum momento para facilitar sua análise. 

<br>

### A transformação de dados geralmente envolve:
* Adicionar, copiar ou replicar dados 
* Excluindo campos (colunas) ou registros (linhas) 
* Padronizando os nomes das variáveis
* Renomear, mover ou combinar colunas em um banco de dados
* Juntando um conjunto de dados com outro
* Salvando um arquivo em um formato diferente. Por exemplo, salvar uma planilha como um arquivo de valores separados por vírgula (CSV).

<br>

### Por que transformar dados?
As metas para a transformação de dados podem ser: 
* Organização de dados : dados melhor organizados são mais fáceis de usar
* Compatibilidade de dados : diferentes aplicativos ou sistemas podem usar os mesmos dados
* Migração de dados : dados com formatos correspondentes podem ser movidos de um sistema para outro
* Mesclagem de dados : dados com a mesma organização podem ser mesclados
* Aprimoramento de dados : os dados podem ser exibidos com campos mais detalhados 
* Comparação de dados : comparações de maçãs com maçãs dos dados podem então ser feitas 

<br>

O viés de dados é um tipo de erro que sistematicamente distorce os resultados em uma determinada direção. Talvez as perguntas de uma pesquisa tivessem uma inclinação particular para influenciar as respostas ou talvez o grupo de amostra não fosse realmente representativo da população que está sendo estudada.

Até agora aprendemos que os preconceitos que temos como pessoas pode acabar criando dados tendenciosos, somos tendenciosos quando temos preferências baseadas em nossas próprias noções preconcebidas ou mesmo subconscientes. Quando os dados são tendenciosos, pode distorcer sistematicamente os resultados em uma certa direção, tornando-os não confiáveis. 

Viés de amostragem é quando uma amostra não é representativo da população como um todo. Você pode evitar isso certificando-se a amostra é escolhida aleatoriamente, para que todas as partes da população têm a mesma chance de serem incluídos. Se você não usar amostragem aleatória durante a coleta de dados, você acaba favorecendo um resultado.

A amostragem imparcial resulta em uma amostra que é representativo da população que está sendo medida.

Além do viés amostral, existe outros três tipos de viés de dados: viés do observador, viés de interpretação e viés de confirmação.

## Encontrar e identificar boas fontes de dados (ROCCC) <br>
* R (Reliable) = Confiável: Confiar que está obtendo precisão (não são imprecisos), informações completas (não são incompletos) e imparciais (não são tendenciosos) que foi examinado e comprovado adequado para uso.
* O (Original) = Original: Certifique-se de validá-lo com a fonte original e que não venha de uma segunda ou terceira fonte.
* C (Comprehensive) = Abrangente: As melhores fontes de dados contêm todas as informações críticas necessárias para responder à pergunta ou encontrar a solução.
* C (Current) = Corrente: As melhores fontes de dados são atuais e relevante para a tarefa em questão.
* C (Cited) = Citado: A citação ou examinadas torna a informação que está fornecendo mais credibilidade.

## Anonimização de dados <br>
### O que é anonimização de dados?
A anonimização de dados é o processo de proteção dos dados privados ou confidenciais das pessoas, eliminando esse tipo de informação. Normalmente, a anonimização de dados envolve apagamento, hash ou mascaramento de informações pessoais, geralmente usando códigos de comprimento fixo para representar colunas de dados ou ocultar dados com valores alterados.

<br>

### Que tipos de dados devem ser anonimizados?
Dados de saúde e financeiros são dois dos tipos de dados mais sensíveis. Essas indústrias dependem muito de técnicas de anonimização de dados. Esses dois setores geralmente passam por desidentificação , que é um processo usado para limpar os dados de todas as informações de identificação pessoal. A anonimização de dados é usada em praticamente todos os setores. Segue abaixo uma lista de dados que geralmente são anonimizados:
* Números de telefone
* Nomes
* Placas de licença e números de licença
* Números de segurança social
* Endereços IP
* Registros médicos
* Endereço de e-mail
* Fotografias
* Números de conta

<br>

### A anonimização dos dados é uma das maneiras de manter os dados privados e seguros!

## O debate sobre dados abertos <br>
### O que são dados abertos?
Na análise de dados, os dados abertos fazem parte da ética dos dados, que tem a ver com o uso ético dos dados. Abertura refere-se ao livre acesso, uso e compartilhamento de dados. Mas para que os dados sejam considerados abertos, eles devem:
* Estar disponível e acessível ao público como um conjunto de dados completo
* Ser fornecido em termos que permitam sua reutilização e redistribuição
* Permita a participação universal para que qualquer pessoa possa usar, reutilizar e redistribuir os dados
Os dados só podem ser considerados abertos quando atendem a todos esses três padrões. 

<br>

### Informações de identificação pessoal (PII) são dados com probabilidade razoável de identificar uma pessoa e divulgar informações sobre ela. É importante manter esses dados seguros . As PII podem incluir o endereço de uma pessoa, informações de cartão de crédito, número do seguro social, registros médicos e muito mais.<br>
### Todo mundo quer manter as informações pessoais sobre si mesmas privadas. Como os dados de terceiros estão prontamente disponíveis, é importante equilibrar a abertura dos dados com a privacidade dos indivíduos. <br>

## Bancos de dados em análise de dados: <br>

Os bancos de dados permitem que os analistas manipulem, armazenem e processem dados. Isso os ajuda a pesquisar os dados com muito mais eficiência para obter os melhores insights. 
Um banco de dados relacional é um banco de dados que contém uma série de tabelas que podem ser conectadas para mostrar relacionamentos. Basicamente, eles permitem que os analistas de dados organizem e vinculem dados com base no que os dados têm em comum.  

Em uma tabela não relacional, você encontrará todas as possíveis variáveis que você pode estar interessado em analisar todas agrupadas. Isso pode tornar muito difícil de classificar. Esta é uma das razões pelas quais os bancos de dados relacionais são tão comuns na análise de dados: eles simplificam muitos processos de análise e tornam os dados mais fáceis de encontrar e usar em um banco de dados inteiro. 

<br>

#### Normalização do banco de dados:
A normalização é um processo de organização de dados em um banco de dados relacional. Por exemplo, criar tabelas e estabelecer relacionamentos entre essas tabelas. É aplicado para eliminar a redundância de dados, aumentar a integridade dos dados e reduzir a complexidade em um banco de dados.

<br>

#### A chave para bancos de dados relacionais:
As tabelas em um banco de dados relacional são conectadas pelos campos que têm em comum. Você deve se lembrar de ter aprendido sobre chaves primárias e estrangeiras antes. Para relembrar rapidamente, uma chave primária é um identificador que faz referência a uma coluna na qual cada valor é exclusivo. Em outras palavras, é uma coluna de uma tabela usada para identificar exclusivamente cada registro dentro dessa tabela. O valor atribuído à chave primária em uma determinada linha deve ser exclusivo em toda a tabela. Por exemplo, se customer_id for a chave primária para a tabela customer, dois clientes nunca terão o mesmo customer_id. 

<br>

Por outro lado, uma chave estrangeira é um campo dentro de uma tabela que é uma chave primária em outra tabela. Uma tabela pode ter apenas uma chave primária, mas pode ter várias chaves estrangeiras. Essas chaves são o que cria os relacionamentos entre as tabelas em um banco de dados relacional, o que ajuda a organizar e conectar dados em várias tabelas no banco de dados.

<br>

Algumas tabelas não requerem uma chave primária. Por exemplo, uma tabela de receita pode ter várias chaves estrangeiras e não ter uma chave primária. Uma chave primária também pode ser construída usando várias colunas de uma tabela. Esse tipo de chave primária é chamado de chave composta . Por exemplo, se customer_id e location_id forem duas colunas de uma chave composta para uma tabela de clientes, os valores atribuídos a esses campos em qualquer linha devem ser exclusivos em toda a tabela.

## Inspecionando um conjunto de dados: <br>
Como analista de dados, você usará os dados para responder perguntas e resolver problemas. Ao analisar dados e tirar conclusões, você está gerando insights que podem influenciar decisões de negócios, gerar mudanças positivas e ajudar seus stakeholders a atingir seus objetivos. 

Antes de iniciar uma análise, é importante inspecionar seus dados para determinar se eles contêm as informações específicas necessárias para responder às perguntas dos interessados. Em qualquer conjunto de dados, pode ser o caso que:
* Os dados não estão lá
* Os dados são insuficientes
* Os dados estão incorretos

<br>

Inspecionar seu conjunto de dados ajudará você a identificar quais perguntas podem ser respondidas e quais dados ainda estão faltando. Você pode recuperar esses dados de uma fonte externa ou, pelo menos, recomendar às partes interessadas que outra fonte de dados seja usada. 

## Metadado: <br>
São informações usadas para descrever os dados que estão contidos em algo,como uma foto ou um e-mail. Lembre-se de que os metadados não são os dados em si. Em vez disso, são dados sobre os dados. Na análise de dados, os metadados ajudam os analistas de dados interpretar o conteúdo dos dados dentro de um banco de dados. É por isso que os metadados são tão importantes ao trabalhar com bancos de dados. Ele diz a um analista sobre o que são os dados. Isso torna possível colocar os dados para trabalhar resolver problemas e tomar decisões baseadas em dados. Como analista de dados, existem três tipos comuns de metadados que você encontrará: descritivo, estrutural e administrativo. 
* Metadados descritivos são metadados que descreve um pedaço de dados e podem ser usados para identificá-lo em um momento posterior.
* Metadados estruturais são metadados que indicam como um pedaço dos dados estão organizados e se fazem partede uma ou mais coletas de dados.
* Metadados administrativos são metadados queindica a fonte técnica de um ativo digital.Quando olhamos para os metadados dentro da foto,eram metadados administrativos.

<br>

Em essência, os metadados informam quem, o quê, quando, onde, qual, como e por quê dos dados.

<br>

Elementos de metadados:
* Título e descrição = Qual é o nome do arquivo ou site que você está examinando? Que tipo de conteúdo contém?
* Etiquetas e categorias = Qual é a visão geral dos dados que você tem? Os dados são indexados ou descritos de uma maneira específica? 
* Quem o criou e quando = De onde vieram os dados e quando foram criados? É recente ou existe há muito tempo?
* Quem o modificou pela última vez e quando = Houve alguma alteração nos dados? Se sim, as modificações foram recentes?
* Quem pode acessá-lo ou atualizá-lo = Este conjunto de dados é público? São necessárias permissões especiais para personalizar ou modificar o conjunto de dados?

<br>

Conhecer o conteúdo e o contexto de seus dados, bem como sua estrutura, é muito valioso em sua carreira como analista de dados. Ao analisar dados, é importante sempre entender o quadro completo. Não se trata apenas dos dados que você está visualizando, mas de como esses dados são reunidos. Os metadados garantem que você possa encontrar, usar, preservar e reutilizar dados no futuro. Lembre-se, será sua responsabilidade gerenciar e fazer uso dos dados em sua totalidade; metadados são tão importantes quanto os próprios dados.

<br>

## Repositório de metadados: <br>
É um banco de dados criado especificamente para armazenar metadados. Os repositórios de metadados podem ser armazenados em um local físico, ou podem ser virtuais, como os dados que existem na nuvem. Esses repositórios descrevem a origem dos metadados, mantê-lo em um formato acessível para quepode ser usado de forma rápida e fácil,e mantê-lo em uma estrutura comum para todos que possam precisar usá-lo. Os repositórios de metadados tornam isso mais fácil e rápido para reunir várias fontes para análise de dados. Eles fazem isso descrevendoo estado e a localização dos metadados, a estrutura das mesas no interior,e como os dados fluem pelo repositório. Eles até acompanham quem acessa os metadados e quando.

<br>

Metadados e repositorios de metadados são usados para criar uma única fonte de verdade, manter os dados consistentes e uniformes e garantir que os dados com os quais trabalhamos sejam precisos, precisa, relevante e oportuna. São armazenados em uma localização única e central e dá a informações padronizadas da empresa sobre todos os seus dados. Isto é feito de duas maneiras. Primeiro, os metadados incluem informações sobre onde cada sistema está localizado e onde os conjuntos de dados estão localizados dentro desses sistemas. Em segundo lugar, os metadados descrevem como todos os dados são conectados entre os vários sistemas.

<br>

É um processo para garantiro gerenciamento formal dos ativos de dados de uma empresa.Isso dá uma organizaçãomelhor controle de seus dados eajuda uma empresa a gerenciar problemasrelacionados à segurança e privacidade de dados,integridade, usabilidade,e fluxos de dados internos e externos. É importante observar que a governança de dados envolve maisdo que apenas padronizar a terminologia e os procedimentos.É sobre os papéis e responsabilidades deas pessoas que trabalham com os metadados todos os dias.Eles são especialistas em metadados eorganizar e manter os dados da empresa,garantindo que seja da mais alta qualidade possível.Essas pessoas criam identificação básica de metadadose informações de descoberta,descrever a maneira como diferentes conjuntos de dados trabalham juntos,e explicar os muitos tipos diferentes de recursos de dados.Especialistas em metadados também criam padrões muito importantesque todos seguem eos modelos usados para organizar os dados.

## Práticas recomendadas para convenções de nomenclatura de arquivos: <br>
Convenções de nomenclatura de arquivo ajudam a organizar, acessar, processar e analisar os dados:
* Trabalhe e concorde com as convenções de nomenclatura de arquivo no início de um projeto para evitar a renomeação de arquivos repetidas vezes.
* Alinhe a nomenclatura de seus arquivos com as convenções existentes de nomenclatura de arquivos de sua equipe ou empresa.
* Assegure-se de que seus nomes de arquivo sejam significativos; considere incluir informações como nome do projeto e qualquer outra coisa que o ajude a identificar rapidamente (e usar) o arquivo para a finalidade certa.
* Inclua a data e o número da versão nos nomes dos arquivos; formatos comuns são YYYYMMDD para datas e v## para versões (ou revisões).
* Crie um arquivo de texto como um arquivo de amostra com conteúdo que descreva (quebre) a convenção de nomenclatura de arquivo e um nome de arquivo que a aplique.
* Evite espaços e caracteres especiais em nomes de arquivos. Em vez disso, use travessões, sublinhados ou letras maiúsculas. Espaços e caracteres especiais podem causar erros em alguns aplicativos.
* Crie pastas e subpastas em uma hierarquia lógica para que os arquivos relacionados sejam armazenados juntos.
* Separe o trabalho em andamento do concluído para que os arquivos do projeto atual sejam mais fáceis de encontrar. Arquive os arquivos mais antigos em uma pasta separada ou em um local de armazenamento externo.
* Se o backup dos seus arquivos não for feito automaticamente, faça backup manualmente com frequência para evitar a perda de trabalhos importantes.

## Segurança de dados: <br>
Significa proteger dados de acesso não autorizado ou corrupção adotando medidas de segurança.
Os analistas de dados precisam encontrar uma maneira de equilibrar a segurança dos dados com suas necessidades reais de análise. Isso pode ser complicado - queremos manter nossos dados seguros e protegidos, mas também queremos usá-los o mais rápido possível para que possamos fazer observações significativas e oportunas. 
As empresas precisam encontrar maneiras de equilibrar suas medidas de segurança de dados com suas necessidades de acesso a dados.

* A criptografia usa um algoritmo exclusivo para alterar os dados e torná-los inutilizáveis por usuários e aplicativos que não conhecem o algoritmo. Este algoritmo é salvo como uma “chave” que pode ser usada para reverter a criptografia; portanto, se você tiver a chave, ainda poderá usar os dados em sua forma original.  
* A tokenização substitui os elementos de dados que você deseja proteger por dados gerados aleatoriamente, chamados de “token”. Os dados originais são armazenados em um local separado e mapeados para os tokens. Para acessar os dados originais completos, o usuário ou aplicativo precisa ter permissão para usar os dados tokenizados e o mapeamento de token. Isso significa que, mesmo que os dados tokenizados sejam hackeados, os dados originais ainda estarão seguros e protegidos em um local separado.
