## Seis fases da análise de dados: <br>
### Etapa 1: pergunte
É impossível resolver um problema se você não sabe o que é. Estas são algumas coisas a considerar:
* Defina o problema que você está tentando resolver
* Certifique-se de entender completamente as expectativas da parte interessada
* Concentre-se no problema real e evite distrações
* Colabore com as partes interessadas e mantenha uma linha de comunicação aberta
* Dê um passo para trás e veja toda a situação no contexto

Perguntas a serem feitas nesta etapa:
1. Quais são os meus stakeholders dizendo que são seus problemas?
2. Agora que identifiquei os problemas, como posso ajudar as partes interessadas a resolver suas dúvidas?

<br>

### Passo 2: Prepare-se 
Você decidirá quais dados precisa coletar para responder às suas perguntas e como organizá-los para que sejam úteis. Você pode usar sua tarefa de negócios para decidir: 
* Quais métricas medir
* Localize dados em seu banco de dados
* Crie medidas de segurança para proteger esses dados

Perguntas a serem feitas nesta etapa: 
1. O que eu preciso para descobrir como resolver esse problema?
2. Que pesquisa preciso fazer?

<br>

### Passo 3: Processo
Dados limpos são os melhores dados e você precisará limpar seus dados para se livrar de possíveis erros, imprecisões ou inconsistências. Isso pode significar:
* Usando funções de planilha para encontrar dados inseridos incorretamente
* Usando funções SQL para verificar se há espaços extras
* Removendo entradas repetidas
* Verificando o máximo possível se há viés nos dados

Perguntas a serem feitas nesta etapa: 
1. Quais erros ou imprecisões de dados podem atrapalhar a obtenção da melhor resposta possível para o problema que estou tentando resolver?
2. Como posso limpar meus dados para que as informações que tenho sejam mais consistentes?

<br>

### Passo 4: Analise 
Você vai querer pensar analiticamente sobre seus dados. Nesta fase, você pode classificar e formatar seus dados para facilitar: 
* Realizar cálculos
* Combine dados de várias fontes
* Crie tabelas com seus resultados

Perguntas a serem feitas nesta etapa:
1. Que história meus dados estão me contando?
2. Como meus dados vão me ajudar a resolver esse problema?
3. Quem precisa do produto ou serviço da minha empresa? Que tipo de pessoa tem mais probabilidade de usá-lo?

<br>

### Passo 5: Compartilhe
Todos compartilham seus resultados de maneira diferente, portanto, resuma seus resultados com visuais claros e atraentes de sua análise usando dados por meio de ferramentas como gráficos ou painéis. Esta é sua chance de mostrar aos stakeholders que você resolveu o problema deles e como chegou lá. Compartilhar certamente ajudará sua equipe:  
* Tome melhores decisões
* Tome decisões mais informadas
* Levar a resultados mais fortes
* Comunique com sucesso suas descobertas

Perguntas a serem feitas nesta etapa:
1. Como posso tornar o que apresento às partes interessadas envolvente e fácil de entender?
2. O que me ajudaria a entender isso se eu fosse o ouvinte?

<br>

### Etapa 6: Agir
Agora é hora de agir em seus dados. Você pegará tudo o que aprendeu com sua análise de dados e o colocará em prática. Isso pode significar fornecer recomendações às partes interessadas com base em suas descobertas, para que possam tomar decisões baseadas em dados.

Perguntas a serem feitas nesta etapa:
1. Como posso usar o feedback que recebi durante a fase de compartilhamento (etapa 5) para realmente atender às necessidades e expectativas da parte interessada?
2. Essas seis etapas podem ajudá-lo a dividir o processo de análise de dados em partes menores e gerenciáveis, o que é chamado de pensamento estruturado . Este processo envolve quatro atividades básicas:
    * Reconhecer o problema ou situação atual
    * Organizando as informações disponíveis 
    * Revelando lacunas e oportunidades
    * Identificando suas opções

Quando você está iniciando sua carreira como analista de dados, é normal se sentir puxado em algumas direções diferentes com sua função e expectativas. Seguir processos como os descritos aqui e usar habilidades de pensamento estruturado podem ajudá-lo a voltar aos trilhos, preencher quaisquer lacunas e informar exatamente o que você precisa.

O pensamento estruturado é o processo de reconhecer o problema atual ousituação, organizar as informações disponíveis,revelando lacunas e oportunidades e identificando as opções.Nesse processo, você aborda um problema vago e complexo, dividindo-o em partes menores.etapas e, em seguida, essas etapas levam você a uma solução lógica

Por definição, a maioria dos novos problemas que os analistas de dados enfrentam começam em território desconhecido. Cabe ao analista de dados e suas habilidades de resolução de problemas pensar estrategicamente, fazer boas perguntas e usar os dados para encontrar soluções para esses problemas. 

## Seis tipos de problemas <br>
A análise de dados é muito mais do que apenas inserir informações em uma plataforma para encontrar insights. Trata-se de resolver problemas. Para chegar à raiz desses problemas e encontrar soluções práticas, existem muitas oportunidades para o pensamento criativo. Não importa o problema, o primeiro e mais importante passo é entendê-lo. 

Os analistas de dados trabalham com uma variedade de problemas.Existe seis tipos comuns: 
1. Fazer previsões: Este tipo de problema envolve o uso de dados para fazer uma decisão informada sobre como as coisas podem ser no futuro.
2. Categorizar coisas: Isso significa atribuir informações a diferentes grupos ou clusters com base em características comuns.
3. Detectar algo incomum: Identificam dados, que é diferente da norma. Uma instância de detectar algo incomum no mundo real
4. Identificar temas: A identificação de temas leva a categorização como uma etapaainda mais, agrupando informações em conceitos mais amplos.
5. Descobrindo conexões: encontrar desafios semelhantes enfrentados por diferentes entidades,e, em seguida, combine dados e insights para resolvê-los
6. Encontrando padrões: usam dados para encontrarpadrões usando dados históricos paraentender o que aconteceu empassado e, portanto, é provável que aconteça novamente.

### Atualmente, empresas em muitos setores estão lidando com mudanças rápidas e incertezas crescentes. Mesmo as empresas bem estabelecidas estão sob pressão para acompanhar o que há de novo e descobrir o que vem a seguir. Para fazer isso, eles precisam fazer perguntas. Fazer as perguntas certas pode ajudar a desencadear as ideias inovadoras pelas quais tantas empresas anseiam hoje em dia. <br>

### As perguntas eficazes seguem a metodologia SMART. Isso significa que eles são específicos, mensuráveis, orientados para a ação, relevantes e com prazo definido.
* S = As perguntas específicas são simples, significativas efocado em um único tópico ou em algumas ideias intimamente relacionadas.Isso nos ajuda a coletar informações relevantes para o que estamos investigando.Se uma pergunta for muito geral,tente reduzi-lo concentrando-se em apenas um elemento.
* M = Questões mensuráveis podem ser quantificadas e avaliadas.
* A = Perguntas orientadas para a ação incentivam a mudança.Você deve se lembrar que a solução de problemas é ver o estado atual edescobrir como transformá-lo no estado futuro ideal.Bem, perguntas orientadas para a ação ajudam você a chegar lá.
* R = Questões relevantes importam,são importantes e têm significado para o problema que você está tentando resolver.
* T = As perguntas com limite de tempo especificam o tempo a ser estudado. Limita o leque de possibilidades e permite que o analista de dados se concentre em dados relevantes.

<br>

### Exemplo:
Quais recursos as pessoas procuram ao comprar um carro novo?
* Específico : a pergunta se concentra em um recurso específico do carro?
* Mensurável : a pergunta inclui um sistema de classificação de recursos?
* Orientado para a ação : a pergunta influencia a criação de pacotes de recursos diferentes ou novos?
* Relevante : a pergunta identifica quais recursos determinam ou prejudicam uma possível compra de carro?
* Limite de tempo : a pergunta valida os dados sobre os recursos mais populares dos últimos três anos? 
### As perguntas devem ser abertas. Essa é a melhor maneira de obter respostas que o ajudarão a qualificar ou desqualificar com precisão possíveis soluções para seu problema específico. Portanto, com base no processo de pensamento, possíveis perguntas SMART podem ser:
* Em uma escala de 1 a 10 (sendo 10 o mais importante), quão importante é o seu carro ter tração nas quatro rodas?
* Quais são os cinco principais recursos que você gostaria de ver em um pacote de carro?
* Quais recursos, se incluídos na tração nas quatro rodas, o deixariam mais inclinado a comprar o carro?
* Quanto mais você pagaria por um carro com tração nas quatro rodas?
* A tração nas quatro rodas se tornou mais ou menos popular nos últimos três anos?

Essas perguntas podem ajudá-lo a se concentrar em técnicas e análises que produzem resultados de interesse para as partes interessadas. Eles também esclarecem a data de vencimento da entrega, o que é importante saber para que você possa gerenciar seu tempo com eficiência. Quando você começa a trabalhar em um projeto, precisa fazer perguntas que se alinhem com o plano e as metas e o ajudem a explorar os dados. 
* Quando o projeto é devido?
* Há algum desafio específico a ser lembrado? 
* Quem são as principais partes interessadas neste projeto e o que elas esperam que este projeto faça por elas?
* A quem estou apresentando os resultados?
* Objetivos: Quais são os objetivos do mergulho profundo? Quais perguntas, se houver, devem ser respondidas por este mergulho profundo?
* Público: Quem são as partes interessadas? Quem está interessado ou preocupado com os resultados deste mergulho profundo? Quem é o público da apresentação?
* Tempo: Qual é o prazo para a conclusão? Até que data isso precisa ser feito?
* Recursos: Quais recursos estão disponíveis para atingir os objetivos do mergulho profundo?
* Segurança: Quem deve ter acesso à informação?

Por exemplo, se você conversar com alguém que trabalha no varejo, pode começar com perguntas como: 
* Específico: você atualmente usa dados para tomar decisões em seu negócio? Em caso afirmativo, que tipo(s) de dados você coleta e como os utiliza?
* Mensurável: você sabe qual é a porcentagem de vendas de seus produtos mais vendidos?
* Orientado para a ação: Existem decisões de negócios ou mudanças que você faria se tivesse as informações corretas? Por exemplo, se você tivesse informações sobre como as vendas de guarda-chuva mudam com o clima, como você as usaria?
* Relevante: Com que frequência você revisa os dados do seu negócio?
* Limite de tempo: você pode descrever como os dados o ajudaram a tomar boas decisões para sua(s) loja(s) no ano passado?

Se você estiver conversando com um professor, poderá fazer perguntas diferentes, como: 
* Específico  Que tipo de dados você usa para construir suas aulas?
* Mensurável: Até que ponto as pontuações dos testes de benchmark dos alunos se correlacionam com suas notas?
* Orientado para a ação: você compartilha seus dados com outros professores para melhorar as aulas?
* Relevante: você compartilhou dados de avaliação com uma turma inteira? Em caso afirmativo, os alunos parecem estar mais ou menos motivados, ou quase o mesmo?
* Limite de tempo: Nos últimos cinco anos, quantas vezes você revisou dados de anos acadêmicos anteriores?

Se você estiver conversando com um pequeno empresário de uma sorveteria, pergunte:
* Específico: quais dados você usa para ajudar nas compras e no estoque?
* Mensurável: você pode ordenar (classificar) esses fatores do mais ao menos influente nas vendas: preço, sabor e época do ano (temporada)?
* Orientado para a ação: Existe um único fator sobre o qual você precisa de mais dados para que possa potencialmente aumentar as vendas?
* Relevante: Como você anuncia ou se comunica com os clientes?
* Limite de tempo: como foi o crescimento de suas vendas ano a ano nos últimos três anos?

<br>

Por exemplo, se as perguntas SMART anteriores levaram o proprietário da sorveteria a propor um projeto para analisar as preferências de sabor do cliente, suas anotações podem ser mais ou menos assim:
* Projeto : coletar dados de preferência de sabor do cliente.
* Objetivo comercial geral : usar dados para oferecer ou criar sabores mais populares.
* Duas fontes de dados : Recibos de caixa registradora e pesquisas de clientes concluídas (e-mail).
* Data de conclusão prevista: Q2 
* O que fazer: Ligue mais tarde e fale com o gerente sobre a localização dos dados da pesquisa.

Há duas maneiras de fazer isso,com tomada de decisão baseada em dados ou inspirada em dados.
Tomada de decisão inspirada em dados explora diferentes fontes de dados para descobrir o que eles têm em comum.
Apenas ter toneladas de dados não é suficiente.Temos que fazer algo significativo com ele. Os dados em si fornecem pouco valor.
Pontos de dados individuais tornam-se maisúteis quando são coletados e estruturados,mas eles ainda são um tanto sem sentido por si mesmos.Precisamos interpretar os dados para transformá-los em informações.

O objetivo de todos os analistas de dados é usar os dados para tirar conclusões precisas e fazer boas recomendações. 

## Dados qualitativos e quantitativos nos negócios <br>
Normalmente, os dados qualitativos podem ajudar os analistas a entender melhor seus dados quantitativos, fornecendo um motivo ou uma explicação mais completa.
Em outras palavras, os dados quantitativos geralmente fornecem o quê e os dados qualitativos geralmente fornecem o porquê.

Faça perguntas SMART sobre fontes de dados da vida real
1. Houve algo desafiador em iniciar a conversa?
2. Houve perguntas que você não conseguiu fazer? 
3. Você administrou seu tempo de forma eficaz?
4. Você fez anotações? Eles são tão detalhados quanto você precisa que sejam?
5. Está faltando alguma informação? Há algo que você ainda acha pouco claro ou vago?  
6. Se você pudesse repetir a conversa, há algo que você mudaria?
7. Quais são as fontes de dados disponíveis para o projeto?
8. Quais fontes de dados eram qualitativas e quais eram quantitativas? Explique sua resposta. 
9. Que decisões você poderia tomar ao considerar cada fonte de dados separadamente? Você poderia tomar decisões diferentes sobre os dados se os combinasse? Em caso afirmativo, dê um exemplo.
10. Existe algum tipo de dado que não está disponível, mas você gostaria de encontrar? Em caso afirmativo, o que é e por que você gostaria de saber mais sobre isso?

### Um relatório é uma coleção estática dedados fornecidos às partes interessadas periodicamente.Por outro lado, um painelmonitora dados recebidos ao vivo. <br>

#### Os relatórios são ótimos para fornecer instantâneos de dados históricos de alto nível para uma organização. Por exemplo, as vendas mensais de uma empresa financeira. Eles podem ser projetados e enviados periodicamente, muitas vezes semanalmente ou mensalmente, informações organizadas e fáceis de consultar. Eles são rápidos de projetar e fáceis de use contanto que você os mantenham continuamente. Finalmente, como os relatórios usam dados estáticos ou dados que não mudam depois de gravados, eles refletem dados que já foram limpos e classificados. Os relatórios precisam de manutenção regular e não são muito atraentes visualmente. Por não serem automáticos ou dinâmicos, os relatórios não mostram dados dinâmicos e em evolução. <br>

#### Os painéis são ótimos por vários motivos, eles dão mais acesso à sua equipe às informações que estão sendo registradas, você pode interagir por meio de dados brincando com filtros, e porque eles são dinâmicos, eles têm valor a longo prazo. Se as partes interessadas precisam acessar informações continuamente, um painel pode ser mais eficiente do que ter que puxar relatórios repetidamente, que é uma grande economia de tempo para você. Por um lado, eles levam muito tempo para projetar e podem ser menos eficientes do que relatórios, se eles não são usados  com muita frequência. Se a tabela base quebrar em qualquer ponto, precisam de muita manutenção para voltar a funcionar novamente. Às vezes, os painéis podem sobrecarregar pessoas com informações também. Se você não está acostumado olhando através de dados em um painel, você pode se perder nele. <br>

### Dados x Métricas: <br>
#### A medição específica de dados é feita usando métricas. As métricas podem ser usadas para transformar dados em informações úteis. Uma métrica é um tipo de dados único e quantificável que pode ser usado para medição. Os dados começam como uma coleção de fatos brutos, até que os organizemos em métricas individuais que representam um único tipo de dados. As métricas também podem ser combinadas em fórmulas que você pode conectar seus dados numéricos. <br>

#### Os dados contêm muitos detalhes brutos sobre o problema que estamos explorando. Mas precisamos das métricas certas para obter as respostas que procuramos. Diferentes setores usarão todos os tipos de métricas para medir as coisas em um conjunto de dados. <br>

#### Diferentes indústrias usam todos os tipos de métricas diferentes.Mas há uma coisa que todos eles têm em comum:todos eles estão tentando atingir um objetivo específico medindo dados. Essa meta métrica é uma meta mensurável definida por uma empresa e avaliada por meio de métricas.E assim como existem muitas métricas possíveis,há muitos objetivos possíveis também. <br>

### Os painéis são ferramentas visuais poderosas que ajudam você a contar sua história de dados. Um painel organiza informações de vários conjuntos de dados em um local central, oferecendo grande economia de tempo. Os analistas de dados usam painéis para rastrear, analisar e visualizar dados para responder a perguntas e resolver problemas. <br>
### A tabela a seguir resume os benefícios de usar um painel para analistas de dados e suas partes interessadas:

### Criando um painel:
1. Identifique as partes interessadas que precisam ver os dados e como eles os usarão. Para começar, você precisa fazer perguntas eficazes.
2. Desenhe o painel (o que deve ser exibido): Use um cabeçalho claro para rotular as informações, adicione descrições curtas de texto a cada visualização e mostrar as informações mais importantes no topo
3. Crie maquetes, se desejar (Isso é opcional, mas muitos analistas de dados gostam de esboçar seus painéis antes de criá-los)
4. Selecione as visualizações que você usará no painel > https://www.tableau.com/data-insights/dashboard-showcase
5. Crie filtros conforme necessário (Os filtros mostram determinados dados enquanto ocultam o restante dos dados em um painel. Isso pode ser uma grande ajuda para identificar padrões, mantendo os dados originais intactos.)

### Tipos de painéis: 
* Estratégico : concentra-se em metas e estratégias de longo prazo no mais alto nível de métricas
* Operacional: rastreamento de desempenho de curto prazo e metas intermediárias
* Analítico: consiste nos conjuntos de dados e na matemática usada nesses conjuntos

<br>

### Painéis estratégicos
Uma ampla gama de empresas usa painéis estratégicos ao avaliar e alinhar seus objetivos estratégicos. Esses painéis fornecem informações no período de tempo mais longo - de um único trimestre financeiro a anos. Eles geralmente contêm informações úteis para a tomada de decisões em toda a empresa. Um exemplo de um painel estratégico que se concentra nos principais indicadores de desempenho (KPIs) ao longo de um ano.

<br>

### Painéis operacionais
Os painéis operacionais são, indiscutivelmente, o tipo mais comum de painel. Como esses painéis contêm informações em uma escala de tempo de dias, semanas ou meses, eles podem fornecer insights de desempenho quase em tempo real. Isso permite que as empresas rastreiem e mantenham seus processos operacionais imediatos à luz de seus objetivos estratégicos. O painel operacional abaixo tem como foco o atendimento ao cliente.

<br>

### Painéis analíticos
Os painéis analíticos contêm uma grande quantidade de dados usados pelos analistas de dados. Esses painéis contêm os detalhes envolvidos no uso, análise e previsões feitas pelos cientistas de dados. Certamente a categoria mais técnica, os painéis analíticos geralmente são criados e mantidos por equipes de ciência de dados e raramente compartilhados com a alta administração, pois podem ser muito difíceis de entender. O painel analítico abaixo se concentra nas métricas do desempenho financeiro de uma empresa.
* Acompanhe o desempenho histórico e atual.
* Estabeleça metas de longo e/ou curto prazo.
* Defina os principais indicadores ou métricas de desempenho.
* Identifique possíveis problemas ou pontos de ineficiência.

### Pensamento matemático <br>
#### O pensamento matemático é uma habilidade poderosa que você pode usar para ajudá-lo a resolver problemas eveja novas soluções.Significa olhar para um problema e decompô-lo logicamente passo a passo,para que você possa ver a relação de padrões em seus dados euse isso para analisar seu problema. Esse tipo de pensamento também pode ajudá-lo a descobrir as melhores ferramentas de análiseporque nos permite ver os diferentes aspectos de um problema eescolha a melhor abordagem lógica. <br>
#### Uma maneira de decidir qual ferramenta usar é pelo tamanho do seu conjunto de dados.Ao trabalhar com dados, você descobrirá que há dados grandes e pequenos. <br>

### Dados pequenos: <br>
* Descreve um conjunto de dados composto de métricas específicas em um período de tempo curto e bem definido
* Geralmente organizado e analisado em planilhas
* Provavelmente será usado por pequenas e médias empresas
* Simples de coletar, armazenar, gerenciar, classificar e representar visualmente
* Geralmente já é um tamanho gerenciável para análise

<br>

### Dados grandes: <br>
* Descreve conjuntos de dados grandes e menos específicos que cobrem um longo período de tempo
* Geralmente mantido em um banco de dados e consultado
* Provavelmente será usado por grandes organizações
* Exige muito esforço para coletar, armazenar, gerenciar, classificar e representar visualmente
* Geralmente precisa ser quebrado em pedaços menores para ser organizado e analisado de forma eficaz para a tomada de decisão

<br>

### Desafios: 
* Muitas organizações lidam com sobrecarga de dados e informações sem importância ou irrelevantes. 
* Dados importantes podem ser escondidos no fundo com todos os dados não importantes, o que torna mais difícil encontrá-los e usá-los. Isso pode levar a prazos de tomada de decisão mais lentos e ineficientes.
* Os dados de que você precisa nem sempre são facilmente acessíveis. 
* As ferramentas e soluções de tecnologia atuais ainda lutam para fornecer dados mensuráveis e relatáveis. Isso pode levar a um viés algorítmico injusto. 
* Existem lacunas em muitas soluções de negócios de big data.

<br>

### Benefícios:
* Quando grandes quantidades de dados podem ser armazenadas e analisadas, isso pode ajudar as empresas a identificar formas mais eficientes de fazer negócios e economizar muito tempo e dinheiro.
* O big data ajuda as organizações a identificar as tendências dos padrões de compra e níveis de satisfação dos clientes, o que pode ajudá-los a criar novos produtos e soluções que deixarão os clientes satisfeitos.
* Ao analisar big data, as empresas obtêm uma compreensão muito melhor das condições atuais do mercado, o que pode ajudá-las a ficar à frente da concorrência.
* Como em nosso exemplo anterior de mídia social, o big data ajuda as empresas a acompanhar sua presença on-line, especialmente o feedback, tanto bom quanto ruim, dos clientes. Isso fornece a eles as informações necessárias para melhorar e proteger sua marca.

### As três (ou quatro) palavras V para big data:
Ao pensar nos benefícios e desafios do big data, é útil pensar nos três Vs: volume, variedade e velocidade. 
* Volume: A quantidade de dados
* Variedade: Os diferentes tipos de dados 
* Velocidade: Quão rápido os dados podem ser processados
* Veracidade: A qualidade e a confiabilidade dos dados

<br>

Todas essas são considerações importantes relacionadas ao processamento de conjuntos de dados enormes e complexos.

### Planilhas e o ciclo de vida dos dados <br>

#### Para entender melhor os benefícios do uso de planilhas na análise de dados, vamos explorar como elas se relacionam com cada fase do ciclo de vida dos dados: planejar (PLAN), capturar (CAPTURE), gerenciar (MANAGE), analisar (ANALYZE), arquivar (ARCHIVE) e destruir (DESTROY).

* Planeje os usuários que trabalharão em uma planilha desenvolvendo padrões organizacionais. Isso pode significar formatar suas células, os títulos que você escolhe destacar, o esquema de cores e a maneira como você ordena seus pontos de dados. Ao dedicar um tempo para definir esses padrões, você melhorará a comunicação, garantirá consistência e ajudará as pessoas a serem mais eficientes com seu tempo.

* Capture dados pela fonte conectando planilhas a outras fontes de dados, como um aplicativo de pesquisa online ou um banco de dados. Esses dados serão atualizados automaticamente na planilha. Dessa forma, as informações são sempre as mais atuais e precisas possíveis.

* Gerencie diferentes tipos de dados com uma planilha. Isso pode envolver armazenamento, organização, filtragem e atualização de informações. As planilhas também permitem que você decida quem pode acessar os dados, como as informações são compartilhadas e como manter seus dados seguros e protegidos. 

* Analise os dados em uma planilha para ajudar a tomar melhores decisões. Algumas das ferramentas de análise de planilhas mais comuns incluem fórmulas para agregar dados ou criar relatórios e tabelas dinâmicas para visuais claros e fáceis de entender. 

* Arquive qualquer planilha que você não usa com frequência, mas pode precisar consultar posteriormente com ferramentas integradas. Isso é especialmente útil se você deseja armazenar dados históricos antes de serem atualizados. 

* Destrua sua planilha quando tiver certeza de que nunca mais precisará dela, se tiver cópias de backup melhores ou por motivos legais ou de segurança. Lembre-se de que muitas empresas são obrigadas a seguir certas regras ou adotar medidas para garantir que os dados sejam destruídos adequadamente.	

<br>

As principais vantagens é poder: A planilha pode ser acessada de qualquer lugar, as informações são armazenadas com segurança, é possível trabalhar em equipe, as alterações são visualizadas em tempo real, a produtividade aumenta, recuperar histórico de revisões, aplicar filtros, bloquear células específicas e criar gráficos. Os comentários possibilita o trabalho em equipe, revisões em grupo e concertar os erros mais rapidamente.

O ponto de partida parapensamento estruturado é o domínio do problema,que você pode ter lembrado de antes.Depois de conhecer a área específica de análise,você pode definir sua base e definirtodos os seus requisitos ehipóteses antes de começar a investigar.

### Outra maneira de praticar o pensamento estruturado eevitar erros é usando um escopo de trabalho. Um escopo de trabalho ou SOW é um acordono esboço do trabalhovocê vai atuar em um projeto. Para muitas empresas, issoinclui coisas como detalhes de trabalho,cronogramas e relatórios que o cliente pode esperar.

<br>

### Neste ponto, tente não confundir declaração de trabalho com escopo de trabalho, ambos abreviados como SOW. Embora ambos sejam usados para definir entregas e um cronograma, eles não são a mesma coisa e não devem ser usados de forma intercambiável. Uma declaração de trabalho é um documento que identifica claramente os produtos e serviços que um fornecedor ou contratado fornecerá a uma organização. Inclui objetivos, diretrizes, entregas, cronograma e custos. Um escopo de trabalho é baseado em projeto e define as expectativas e os limites de um projeto. Um escopo de trabalho pode ser incluído em uma declaração de trabalho para ajudar a definir os resultados do projeto.

### Com um sólido escopo de trabalho,você será capaz de resolver qualquer confusão,contradições ou dúvidas sobre a atualização de dadosfrente e certifique-seesses contratempos sorrateiros não ficam no seu caminho.

<br>

### Por que você precisa de um SOW?
### O objetivo dos projetos de análise de dados é concluir tarefas de negócios que sejam úteis para as partes interessadas. A criação de uma SOW ajuda a garantir que todos os envolvidos, de analistas e engenheiros a gerentes e partes interessadas, compartilhem o entendimento de quais são essas metas de negócios e o plano para alcançá-las.

### Esclarecer requisitos e definir expectativas são duas das partes mais importantes de um projeto. Lembre-se da primeira fase do Processo de Análise de Dados — fazer perguntas . À medida que você faz mais e mais perguntas para esclarecer requisitos, metas, fontes de dados, partes interessadas e qualquer outra informação relevante, uma SOW ajuda você a formalizar tudo registrando todas as respostas e detalhes.  Preparar-se para escrever um SOW é fazer perguntas para obter as informações necessárias sobre o projeto, mas também esclarecer e definir o que você está sendo solicitado a realizar e quais são os limites ou limites do “perguntar”. Afinal, se você não consegue fazer uma distinção entre as questões de negócios pelas quais você é responsável e as que não são responsáveis por responder, fica difícil saber o que significa sucesso!

<br>

### Não há um formato padrão para um SOW. Eles podem diferir significativamente de uma organização para outra, ou de projeto para projeto. No entanto, todos eles têm algumas peças fundamentais de conteúdo em comum. 

* Entregáveis : Que trabalho está sendo feito e que coisas estão sendo criadas como resultado deste projeto? Quando o projeto estiver concluído, o que você espera entregar às partes interessadas? Seja específico aqui. Você coletará dados para este projeto? Quanto ou por quanto tempo?   

Evite afirmações vagas. Por exemplo, “consertar problemas de tráfego” não especifica o escopo. Isso pode significar qualquer coisa, desde tapar alguns buracos até construir um novo viaduto. Seja específico! Use números e aponte para metas e objetivos concretos e mensuráveis. Por exemplo: “Identifique os 10 principais problemas com os padrões de tráfego dentro dos limites da cidade e identifique as 3 principais soluções com melhor custo-benefício para reduzir o congestionamento do tráfego.”

* Marcos : Isso está intimamente relacionado à sua linha do tempo. Quais são os principais marcos para o progresso em seu projeto? Como você sabe quando uma determinada parte do projeto é considerada concluída? 

Os marcos podem ser identificados por você, pelas partes interessadas ou por outros membros da equipe, como o gerente de projeto. Exemplos menores podem incluir etapas incrementais em um projeto maior, como “Coletar e processar 50% dos dados necessários (100 respostas de pesquisa)”, mas também podem ser exemplos maiores, como “relatório inicial completo de análise de dados” ou “entregar visualizações de painel concluídas e relatórios de análise às partes interessadas”. 

* Linha do tempo : sua linha do tempo estará intimamente ligada aos marcos que você criar para o seu projeto. A linha do tempo é uma forma de mapear as expectativas de quanto tempo cada etapa do processo deve levar. O cronograma deve ser específico o suficiente para ajudar todos os envolvidos a decidir se um projeto está dentro do cronograma. Quando as entregas serão concluídas? Quanto tempo você espera que o projeto leve para ser concluído? Se tudo correr como planejado, quanto tempo você espera que cada componente do projeto leve? Quando podemos esperar atingir cada marco?

* Relatórios : Bons SOWs também estabelecem limites para como e quando você fornecerá atualizações de status às partes interessadas. Como você comunicará o progresso com as partes interessadas e patrocinadores e com que frequência? O progresso será relatado semanalmente? Por mês? Quando os marcos são concluídos? Quais informações os relatórios de status conterão?

No mínimo, qualquer SOW deve responder a todas as perguntas relevantes nas áreas acima. Observe que essas áreas podem diferir dependendo do projeto. Mas, em sua essência, o documento SOW deve sempre servir ao mesmo propósito, contendo informações específicas, relevantes e precisas. Se algo mudar no projeto, seu SOW deve refletir essas mudanças.

### O que está dentro e fora do escopo?
#### SOWs também devem conter informações específicas sobre o que é e o que não é considerado parte do projeto. O escopo do seu projeto é tudo o que se espera que você conclua ou realize, definido com um nível de detalhe que não deixe nenhuma ambigüidade ou confusão sobre se uma determinada tarefa ou item faz parte do projeto ou não. <br>

#### Definir requisitos pode ser mais complicado do que parece, por isso é importante ser o mais específico possível nesses documentos e usar declarações quantitativas sempre que possível. O uso de declarações específicas e quantificáveis ajudará a garantir que todos tenham uma compreensão clara do que é esperado.

## A importância do contexto: <br>
### O contexto é importante na análise de dados porque ajuda você a filtrar grandes quantidades de dados desorganizados e transformá-los em algo significativo. O fato é que os dados têm pouco valor se não forem pareados com o contexto. <br>

### Isso significa dar perspectiva aos dados, definindo-os. Para fazer isso, você precisa identificar:
* Quem: A pessoa ou organização que criou, coletou e/ou financiou a coleta de dados
* O quê: as coisas no mundo sobre as quais os dados podem ter impacto
* Onde: A origem dos dados
* Quando: a hora em que os dados foram criados ou coletados
* Por quê: A motivação por trás da criação ou coleção
* Como: O método usado para criá-lo ou coletá-lo

<br>

### Compreender e incluir o contexto é importante durante cada etapa do seu processo de análise, por isso é uma boa ideia se familiarizar com isso no início de sua carreira. Por exemplo, ao coletar dados, você também deve fazer perguntas sobre o contexto para garantir que entendeu o negócio e o processo de negócios. Durante a organização, o contexto é importante para suas convenções de nomenclatura, como você escolhe mostrar relacionamentos entre variáveis e o que você escolhe manter ou deixar de fora. E, finalmente, ao apresentar, é importante incluir informações contextuais para que seus stakeholders entendam sua análise.

### Como analista de dados, parte do seu trabalho é comunicar o processo de análise de dados e suas percepções às partes interessadas. Isso geralmente envolve definir o problema e resumir as principais questões e os dados disponíveis desde o início. Você pode incluir essas informações em um documento formal para as partes interessadas, como um escopo de trabalho (SOW) no início de um projeto. Como lembrete, um SOW é um esboço acordado das tarefas a serem executadas durante um projeto; é importante garantir que as partes interessadas entendam essas informações importantes nesse estágio. 

Aqui estão algumas perguntas para ajudá-lo a começar: 
* Qual é o problema? 
* Pode ser resolvido com dados? Se sim, quais dados?
* Onde estão esses dados? Ele existe ou você precisa coletá-lo?
* Você está usando dados privados aos quais alguém precisará fornecer acesso a você ou dados disponíveis publicamente?
* Quem são os patrocinadores e partes interessadas relevantes para este projeto? Quem está envolvido e como?
* Quais são os limites do seu projeto? O que você considera “dentro do escopo”? O que você considera “fora do escopo”?
* Existe alguma outra informação que você considera relevante para o projeto? 
* Existe alguma informação que você precisa ou perguntas que você precisa responder antes de começar?

## Trabalhando com as partes interessadas <br>

### Seu projeto de análise de dados deve responder à tarefa de negócios e criar oportunidades para a tomada de decisões orientada por dados. É por isso que é tão importante focar nas partes interessadas do projeto. Como analista de dados, é sua responsabilidade entender e gerenciar as expectativas das partes interessadas, mantendo os objetivos do projeto em primeiro plano. Existem três grupos comuns de partes interessadas com os quais você pode trabalhar: a equipe executiva, a equipe voltada para o cliente e a equipe de ciência de dados. <br>

* Time executivo: 
A equipe executiva fornece liderança estratégica e operacional para a empresa. Eles estabelecem metas, desenvolvem estratégias e garantem que a estratégia seja executada com eficácia. A equipe executiva pode incluir vice-presidentes, o diretor de marketing e profissionais de nível sênior que ajudam a planejar e dirigir o trabalho da empresa. Essas partes interessadas pensam sobre as decisões em um nível muito alto e procuram primeiro as manchetes sobre o seu projeto. Eles estão menos interessados nos detalhes. O tempo é muito limitado com eles, então aproveite ao máximo conduzindo suas apresentações com as respostas às suas perguntas. Você pode manter as informações mais detalhadas à mão em seu apêndice de apresentação ou na documentação do projeto para que eles possam se aprofundar quando tiverem mais tempo.

<br>

* Equipe voltada para o cliente:
A equipe voltada para o cliente inclui qualquer pessoa em uma organização que tenha algum nível de interação com clientes e clientes em potencial. Normalmente, eles compilam informações, estabelecem expectativas e comunicam o feedback do cliente a outras partes da organização interna. Essas partes interessadas têm seus próprios objetivos e podem vir até você com perguntas específicas. É importante deixar que os dados contem a história e não ser influenciado por pedidos de seus stakeholders para encontrar certos padrões que podem não existir.

<br>

* Equipe de ciência de dados:
A organização de dados dentro de uma empresa exige trabalho em equipe. Há uma boa chance de você trabalhar com outros analistas de dados, cientistas de dados e engenheiros de dados.

<br>

### Stakeholders são pessoas que investiram tempo, interesse e recursos nos projetos que você estará trabalhando como analista de dados. Há uma boa chance de que eles precisem do trabalho que você faz para realizar suas próprias necessidades. Por isso é tão importante que você certifique-se de que seu trabalho esteja alinhado com as necessidades deles e por que você precisa se comunicar efetivamente com todos as partes interessadas em toda a sua equipe. Seus stakeholders vão querer discutir coisas como o objetivo do projeto, o que você precisa para atingir esse objetivo,e quaisquer desafios ou preocupações que você tenha. <br>

### Trabalhar de forma eficaz com as partes interessadas: <br>

#### 1 - Discuta os objetivos. As solicitações das partes interessadas geralmente estão vinculadas a um projeto ou objetivo maior. Quando te pedirem algo, aproveite para saber mais. Iniciar uma discussão. Pergunte sobre o tipo de resultados que a parte interessada deseja. Às vezes, uma conversa rápida sobre metas pode ajudar a definir expectativas e planejar os próximos passos. <br>

#### 2 - Sinta-se empoderado para dizer “não ”. Digamos que você seja abordado por um diretor de marketing que tem um projeto de “alta prioridade” e precisa de dados para respaldar sua hipótese. Eles pedem que você produza a análise e os gráficos para uma apresentação amanhã de manhã. Talvez você perceba que a hipótese deles não está totalmente formada e tenha ideias úteis sobre uma maneira melhor de abordar a análise. Ou talvez você perceba que levará mais tempo e esforço para realizar a análise do que o estimado. Seja qual for o caso, não tenha medo de recuar quando precisar. 
#### As partes interessadas nem sempre percebem o tempo e o esforço gastos na coleta e análise de dados. Eles também podem não saber o que realmente precisam. Você pode ajudar as partes interessadas perguntando sobre suas metas e determinando se você pode entregar o que elas precisam. Se não puder, tenha confiança para dizer “não” e forneça uma explicação respeitosa. Se houver uma opção que seja mais útil, direcione o interessado para esses recursos. Se você achar que precisa priorizar outros projetos primeiro, discuta o que pode priorizar e quando. Quando suas partes interessadas entenderem o que precisa ser feito e o que pode ser realizado em um determinado cronograma, elas geralmente se sentirão à vontade para redefinir suas expectativas. <br>

#### 3 - Planeje o inesperado. Antes de iniciar um projeto, faça uma lista de possíveis obstáculos. Então, quando você discutir as expectativas e os cronogramas do projeto com as partes interessadas, reserve um tempo extra para a solução de problemas em cada estágio do processo.<br>

#### 4 - Conheça o seu projeto. Acompanhe suas discussões sobre o projeto por e-mail ou relatórios e esteja pronto para responder a perguntas sobre como certos aspectos são importantes para sua organização. Saiba como seu projeto se conecta ao restante da empresa e envolva-se para fornecer o máximo de insights possível. Se você tiver um bom entendimento sobre por que está fazendo uma análise, isso pode ajudá-lo a conectar seu trabalho com outros objetivos e ser mais eficaz na solução de problemas maiores. <br>

#### 5 - Comece com palavras e visuais. É comum que analistas de dados e partes interessadas interpretem as coisas de maneiras diferentes enquanto assumem que o outro está na mesma página. Essa ilusão de acordo* foi historicamente identificada como uma causa de projetos que vão e voltam várias vezes antes que uma direção seja finalmente definida. Quanto mais rápido todos concordarem, mais rápido você poderá realizar a primeira análise para testar a utilidade do projeto, medir o feedback, aprender com os dados e implementar mudanças. <br>

#### 6 - Comunique-se frequentemente. Suas partes interessadas desejarão atualizações regulares sobre seus projetos. Compartilhe notas sobre marcos, contratempos e mudanças do projeto. Em seguida, use suas anotações para criar um relatório compartilhável. Outro grande recurso a ser utilizado é o log de alterações, que é uma ferramenta que será explorada mais adiante ao longo do programa. Por enquanto, apenas saiba que um change-log é um arquivo contendo uma lista ordenada cronologicamente das modificações feitas em um projeto. Dependendo da maneira como você o configura, as partes interessadas podem até aparecer e visualizar as atualizações sempre que quiserem.

### Como analista de dados, você pode estar trabalhando em vários projetos com muitos pessoas diferentes, mas não importa em qual projeto você está trabalhando, há três coisas que você pode concentre-se nisso para ajudá-lo a permanecer na tarefa. 
* Um, quem são as partes interessadas primárias e secundárias? Reserve um tempo no início de cada projeto para identifique seus stakeholders e seus objetivos.veja quem mais está em sua equipe e quais são suas funções.
* Dois que estão gerenciando os dados? Você vai querer perguntar quem está gerenciando os dados.
* Três, onde você pode pedir ajuda? você precisa saber onde você pode ir quando precisar de ajuda. Você gastará menos tempo se preocupando com outros aspectos de o projeto e mais tempo focado no objetivo

<br>

### Você será capaz de atender às necessidades das partes interessadas, sintir-se confiante sobre quem está gerenciando os dados e obtenher ajudar quando você precisar para que você possa ficar de olho no prêmio: o objetivo do projeto.<br>

## Limitações de dados <br>
#### Os dados são poderosos, mas têm suas limitações. A opinião pessoal de alguém entrou nos números? Seus dados estão contando toda a história? Parte de ser um grande analista de dados é conhecer os limites dos dados e planejá-los.
* 1 - Caso de dados incompletos (ou inexistentes): Se você tiver dados incompletos ou inexistentes, poderá perceber durante uma análise que não possui dados suficientes para chegar a uma conclusão. Ou você pode até estar resolvendo um problema completamente diferente! 
* 2 - Não perca dados desalinhados: Se você estiver coletando dados de outras equipes e usando planilhas existentes, é bom ter em mente que as pessoas usam regras de negócios diferentes. Portanto, uma equipe pode definir e medir as coisas de uma maneira completamente diferente da outra. Estabelecer como medir as coisas desde o início padroniza os dados em toda a linha para maior confiabilidade e precisão. Isso garantirá que as comparações entre as equipes sejam significativas e perspicazes.
* 3 - Lidar com dados sujos: Dados sujos referem-se a dados que contêm erros. Dados sujos podem levar à perda de produtividade, gastos desnecessários e tomadas de decisão imprudentes. Um bom esforço de limpeza de dados pode ajudar a evitar isso. Como um lembrete rápido, a limpeza de dados é o processo de correção ou remoção de dados incorretos, corrompidos, formatados incorretamente, duplicados ou incompletos em um conjunto de dados. Quando você encontra e corrige os erros - enquanto rastreia as alterações feitas - você pode evitar um desastre de dados. 
* 4 - Conte uma história clara:
    * Compare os mesmos tipos de dados : os dados podem ser misturados quando você os mapeia para visualização. Certifique-se de comparar os mesmos tipos de dados e verifique se todos os segmentos em seu gráfico exibem métricas diferentes. 
    * Visualize com cuidado : uma queda de 0,01% em uma pontuação pode parecer enorme se você aumentar o zoom o suficiente. Para garantir que seu público veja a história completa com clareza, é uma boa ideia definir seu eixo Y como 0.
    * Deixe de fora gráficos desnecessários: se uma tabela pode mostrar sua história rapidamente, fique com a tabela em vez de um gráfico de pizza ou um gráfico. Seu público ocupado apreciará a clareza.
    * Teste de significância estatística: Às vezes, dois conjuntos de dados parecerão diferentes, mas você precisará de uma maneira de testar se a diferença é real e importante. Portanto, lembre-se de executar testes estatísticos para ver quanta confiança você pode depositar nessa diferença.
    * Preste atenção ao tamanho da amostra : reúna muitos dados. Se o tamanho da amostra for pequeno, algumas respostas incomuns podem distorcer os resultados. Se você achar que tem poucos dados, tenha cuidado ao usá-los para formar julgamentos. Procure oportunidades para coletar mais dados e, em seguida, mapeie essas tendências por períodos mais longos.
    * Seja o juiz: Em qualquer organização, grande parte da função de um analista de dados é fazer julgamentos sólidos. Quando você conhece as limitações de seus dados, pode fazer julgamentos que ajudem as pessoas a tomar decisões melhores com base nos dados. Os dados são uma ferramenta extremamente poderosa para a tomada de decisões, mas se estiverem incompletos, desalinhados ou não tiverem sido limpos, podem ser enganosos. Tome as medidas necessárias para garantir que seus dados estejam completos e consistentes.
