# Módulo 1: Fundamentos - Dados, dados, em todos os lugares

<br>

### Podemos encontrar dados praticamente em todos os lugares. Sempre que se observa e avalia algo no mundo, está coletando e analisando dados. A análise ajuda a encontrar maneiras mais fáceis de fazer as coisas, identificar padrões para economizar tempo e descobrir novas perspectivas surpreendentes que podem mudar completamente a maneira como vivenciamos as coisas. 

<br>

### As seis etapas do processo de análise de dados são: 
* Perguntar = Faça perguntas e defina o problema = “Como defino sucesso para este projeto?”
* Preparar = Prepare os dados coletando e armazenando as informações.
* Processar = Processe os dados limpando e verificando as informações.
* Analisar = Analise os dados para encontrar padrões, relacionamentos e tendências.
* Compartilhar = Compartilhe dados com seu público.
* Agir = Aja sobre os dados e use os resultados da análise
### Essas seis etapas se aplicam a qualquer análise de dados.

<br>

### Exemplo:
### People Analytics é a prática de coletar e analisar dados sobre as pessoas que compõem a força de trabalho de uma empresa, a fim de obter insights para melhorar a forma como a empresa opera. Ser um analista de pessoas envolve o uso da análise de dados para obter insights sobre os funcionários e como eles vivenciam suas vidas profissionais. Os insights são usados para definir e criar um local de trabalho mais produtivo e capacitador. Isso pode liberar o potencial dos funcionários, motivar as pessoas a terem o melhor desempenho e garantir uma cultura empresarial justa e inclusiva. 
### Uma organização estava passando por uma alta taxa de rotatividade entre as novas contratações. Muitos funcionários deixaram a empresa antes do final do primeiro ano de trabalho. Os analistas usaram o processo de análise de dados para responder à seguinte pergunta: como a organização pode melhorar a taxa de retenção de novos funcionários? 
1. Etapa Perguntar: Primeiro, os analistas precisavam definir como seria o projeto e o que se qualificaria como um resultado bem-sucedido. Então, para determinar essas coisas, eles fizeram perguntas eficazes e colaboraram com líderes e gerentes interessados no resultado de sua análise de pessoal. Estes foram os tipos de perguntas que eles fizeram:
    * O que acha que os novos funcionários precisam aprender para serem bem-sucedidos no primeiro ano de trabalho?
    * Já coletou dados de novos funcionários antes? Em caso afirmativo, podemos ter acesso aos dados históricos?
    * Acredita que gerentes com taxas de retenção mais altas oferecem aos novos funcionários algo extra ou exclusivo?
    * O que suspeita ser uma das principais causas de insatisfação entre os novos funcionários?
    * Em que porcentagem gostaria que a retenção de funcionários aumentasse no próximo ano fiscal?
    * Perguntar para definir o problema a ser resolvido e quais são as expectativas das partes interessadas (As partes interessadas têm uma participação no projeto.São pessoas que têm investiu tempo e recursos em um projeto e estão interessados no resultado):
        * 1 - Definir um problema significa que você olha o estado atual e identifica como é diferente do estado ideal. Geralmente há um obstáculo que precisamos nos livrar ou algo errado que precisa ser consertado. (Faça perguntas eficazes > Defina o problema > Use o pensamento estruturado > Comunique-se com outras pessoas)
        * 2 - Entender as expectativas das partes interessadas: Determinar quem são as partes interessadas. Eles têm em comum é que ajudam a tomar decisões, influenciar ações e estratégias,e têm metas específicas que desejam atingir. Eles também se preocupam com o projeto e é por isso é tão importante entender suas expectativas.
2. Etapa Preparar: O grupo construiu um cronograma de três meses e decidiu como desejava retransmitir seu progresso às partes interessadas. Ainda nessa etapa, os analistas identificaram quais dados precisavam para alcançar o resultado de sucesso que identificaram na etapa anterior - nesse caso, os analistas optaram por coletar os dados de uma pesquisa online com novos funcionários. Estas foram as coisas que eles fizeram para se preparar:
    * Desenvolveram perguntas específicas sobre a satisfação dos funcionários com diferentes processos de negócios, como contratação e integração, e sua remuneração geral.
    * Estabeleceram regras para quem teria acesso aos dados coletados - nesse caso, qualquer pessoa de fora do grupo não teria acesso aos dados brutos, mas poderia visualizar os dados resumidos ou agregados. Por exemplo, a remuneração de um indivíduo não estaria disponível, mas as faixas salariais para grupos de indivíduos seriam visíveis.
    * Finalizaram quais informações específicas seriam coletadas e a melhor forma de apresentar os dados visualmente. Os analistas debateram possíveis problemas relacionados a projetos e dados e como evitá-los.
    * É aqui que os analistas de dados coletam e armazenam dadoseles usarão para o próximo processo de análise.Você aprenderá mais sobre os diferentes tiposde dados e como identificarquais tipos de dados são maisútil para resolver um determinado problema.Você também descobrirá por que é tão importante queseus dados e resultados são objetivos e imparciais. (Entenda como os dados são gerados e coletados > Identificar e usar dados diferentes formatos, tipos e estruturas > Certifique-se de que os dados sejam imparciais e confiável > Organizar e proteger dados)
3. Etapa Processar: O grupo enviou a pesquisa. Grandes analistas sabem respeitar tanto seus dados quanto as pessoas que os fornecem. Como os funcionários forneceram os dados, era importante garantir que todos os funcionários dessem seu consentimento para participar. Os analistas de dados também garantiram que os funcionários entendessem como seus dados seriam coletados, armazenados, gerenciados e protegidos. Coletar e usar dados de forma ética é uma das responsabilidades dos analistas de dados. De forma a manter a confidencialidade e proteger e armazenar os dados de forma eficaz, estes foram os passos que tomaram:
    * Restringiram o acesso aos dados a um número limitado de analistas.
    * Limparam os dados para garantir que estavam completos, corretos e relevantes. Certos dados foram agregados e resumidos sem revelar respostas individuais.
    * Carregaram dados brutos em um data warehouse interno para uma camada adicional de segurança.
    * os analistas de dados encontram e eliminamquaisquer erros e imprecisõesque podem atrapalhar os resultados.Isso geralmente significa dados de limpeza,transformando-o em um formato mais útil,combinando dois ou mais conjuntos de dados para fazerinformações mais completas e removendo outliers,quais são quaisquer pontos de dados quepoderia distorcer a informação. Depois disso, você aprenderá como verificar os dados que vocêprepare-se para garantir que esteja completo e correto.Esta fase é sobre acertar os detalhes.Então você também corrigirá erros de digitação, inconsistências ou dados ausentes e imprecisos. Como se não bastasse,você vai ganhar estratégias paraverificando e compartilhandosua limpeza de dados com as partes interessadas. (Criar e transformar dados > Manter a integridade dos dados > Dados de teste > Limpar dados > Verificar e relatar os resultados da limpeza)
4. Etapa Analisar: A partir das pesquisas concluídas, os analistas de dados descobriram que a experiência de um funcionário com determinados processos era um indicador-chave da satisfação geral no trabalho. Estas foram suas descobertas:
    * Funcionários que passaram por um processo de contratação longo e complicado tiveram maior probabilidade de deixar a empresa.
    * Funcionários que passaram por um processo de avaliação e feedback eficiente e transparente tiveram maior probabilidade de permanecer na empresa.
    * Use ferramentas para formatar e transformar dados > Classificar e filtrar dados > Identifique padrões e desenhe conclusões > Faça previsões e recomendações > Tome decisões baseadas em dados
OBS: O grupo sabia que era importante documentar exatamente o que encontraram na análise, independentemente dos resultados. Fazer o contrário diminuiria a confiança no processo de pesquisa e reduziria sua capacidade de coletar dados verdadeiros dos funcionários no futuro.
5. Etapa Compartilhar: Assim como eles garantiram que os dados fossem cuidadosamente protegidos, os analistas também foram cuidadosos ao compartilhar o relatório . Foi assim que eles compartilharam suas descobertas:
    * Compartilharam o relatório com gerentes que atingiram ou excederam o número mínimo de subordinados diretos com respostas enviadas à pesquisa.
    * Apresentaram os resultados aos gerentes para garantir que tivessem uma visão completa.
    * Entenda a visualização > Crie recursos visuais eficazes > Dê vida aos dados > Use narrativa de dados > Comunique-se para ajudar os outros a entender resultados
OBS: Eles pediram aos gerentes que entregassem pessoalmente os resultados para suas equipes.
Esse processo deu aos gerentes a oportunidade de comunicar os resultados com o contexto certo. Como resultado, eles poderiam ter conversas produtivas em equipe sobre as próximas etapas para melhorar o engajamento dos funcionários.
6. Etapa Agir: A última etapa do processo para a equipe de analistas foi trabalhar com os líderes da empresa e decidir a melhor forma de implementar as mudanças e agir com base nas descobertas. Estas foram as suas recomendações:
    * Padronizar o processo de contratação e avaliação de colaboradores com base nas práticas mais eficientes e transparentes.
    * Realizar a mesma pesquisa anualmente e comparar os resultados com os do ano anterior.
    * Aplique seus insights > Resolver problemas > Tomar decisões > Crie algo novo
OBS: Um ano depois, a mesma pesquisa foi distribuída aos funcionários. Os analistas anteciparam que uma comparação entre os dois conjuntos de resultados indicaria que o plano de ação funcionou. Acontece que as mudanças melhoraram a taxa de retenção de novos funcionários e as ações tomadas pelos líderes foram bem-sucedidas!

### Observação: <br>
* A ciência de dados é definida como a criação de novas formas de modelar e entender o desconhecido usando dados brutos. Os cientistas de dados criam novas perguntas usando dados, enquanto os analistas encontram respostas para perguntas existentes criando insights de fontes de dados.
* Data analysis é a coleta, transformação e organização de dados para tirar conclusões, fazer previsões e conduzir a tomada de decisões informadas. Um analista de dados encontra dados, analisa e usa para descobrir tendências, padrões e relacionamentos
* Data analytics nos termos mais simples é a ciência dos dados. É um conceito muito amplo que abrange tudo, desde o trabalho de gerenciamento e uso de dados até as ferramentas e métodos que os trabalhadores de dados usam todos os dias. Portanto, quando pensa em dados, Data analysis e ecossistema de dados, é importante entender que todas essas coisas se encaixam no guarda-chuva da Data analytics. 

### A tomada de decisão baseada em dados: <br> 
É definida como o uso de fatos para orientar a estratégia de negócios.Organizações em muitos setores diferentes são capacitadas para fazer melhor,decisões baseadas em dados por analistas de dados o tempo todo.O primeiro passo na tomada de decisão baseada em dados é descobrir a necessidade do negócio.Normalmente, este é um problema que precisa ser resolvido.
Exemplos:
- Nova empresa que precisa estabelecer melhoresreconhecimento da marca, para que possa competir com concorrentes maiores e mais conhecidos.
- Uma organização queira melhorar um produto e precise descobrir comoobter peças de um fornecedor mais sustentável ou eticamente responsável.
- Uma empresa tentando resolver o problema de funcionários insatisfeitos,baixos níveis de engajamento, satisfação e retenção.

<br>

OBS: As decisões baseadas no instinto, sem nenhum dado para apoiá-las, podem causar erros.

Combinar dados com conhecimento de negócios, mais talvez um toque de instinto, será uma parte comum do seu processo como analista de dados júnior.
A chave é descobrir a combinação exata para cada projeto em particular. Muitas vezes, vai depender dos objetivos da sua análise. É por isso que os analistas costumam perguntar: “Como defino sucesso para este projeto?”
* Que tipo de resultados são necessários?
* Quem será informado?
* Estou respondendo à pergunta que está sendo feita?
* Com que rapidez uma decisão precisa ser tomada?

### Ciclo de vida da análise de dados <br>
O processo que vai dos dados à decisão = Os dados passam por várias fases à medida que são criados, consumidos, testados, processados e reutilizados.
* Pergunte : Desafio de negócios/Objetivo/Pergunta
* Preparar : Geração, coleta, armazenamento e gerenciamento de dados
* Processo : limpeza de dados/integridade de dados
* Analisar : exploração, visualização e análise de dados
* Compartilhar : comunicar e interpretar resultados 
* Agir : Colocar seus insights para trabalhar para resolver o problema

### Habilidades analíticas <br>
São qualidades e características associados à resolução de problemas usando fatos. Há muitos aspectos nas habilidades analíticas, mas, vamos nos concentrar em cinco pontos essenciais. Eles são curiosidade, compreensão do contexto, ter mentalidade técnica, design de dados e estratégia de dados.
* Curiosidade é querer aprender alguma coisa. Pessoas curiosas geralmente procuramnovos desafios e experiências.
* Capacidade de entender o contexto: Contexto é a condição em qual algo existe ou acontece. Pode ser uma estrutura ou um ambiente. Tentar entender o quadro completo é crítico. Tem a ver com como você agrupa as coisas em categorias
* Uma mentalidade técnica envolve a capacidade dedividir as coisas em etapas ou pedaços menorese trabalhar com eles de forma ordenada e lógica.
* O design de dados é como organiza as informações.
* A estratégia de dados é a gestão das pessoas, processos e ferramentas usadas na análise de dados. Você gerencia as pessoas certificando-se de que elas saiba como usar os dados certos para encontrar soluções para o problema em que você está trabalhando. Para processos, trata-se de garantir o caminho para essa solução é claro e acessível. Para ferramentas, certifique-se a tecnologia certa está sendo usada para o trabalho.

### Pensamento analítico <br> 
Envolve identificar e definindo um problema e então resolvê-lo usando dados emuma maneira organizada, passo a passo. Os cinco aspectos-chave para o pensamento analítico.
Eles são visualização, estratégia, orientação para o problema, correlação e, finalmente, pensamento de visão geral e detalhista.

* Na análise de dados, a visualização é a representação gráfica da informação. Alguns exemplos incluem gráficos,mapas ou outros elementos de design. A visualização é importante porque os recursos visuais podem ajudar analistas de dados entendem e explicar as informações de forma mais eficaz.
* Ter uma mentalidade estratégica é chave para se manter focado e no caminho certo.A estratégia ajuda os analistas de dados a ver o que elesquerem alcançar com os dados e como eles podem chegar lá.A estratégia também ajuda a melhorara qualidade e utilidade dos dados que coletamos.Ao criar estratégias, sabemos que todos os nossos dados sãovaliosas e podem nos ajudar a atingir nossos objetivos.
* Os analistas de dados usam um problema-abordagem orientada para identificar, descrever e resolver problemas. É tudo uma questão de manter o problema no topo da mente durante todo o projeto.
* Uma correlação é como um relacionamento.Você pode encontrar todos os tipos de correlações nos dados.
* Isso significa ser capaz de vera imagem grande, bem como os detalhes.Um quebra-cabeça é uma ótima maneira de pensar sobre isso.O pensamento geral é comoolhando para um quebra-cabeça completo.Você pode apreciar toda a imagem semficar preso em cada minúsculopeça que entrou em fazê-lo.Se você se concentrar apenas em peças individuais,você não seria capaz de ver além disso,e é por isso que o pensamento amplo é tão importante. Ele ajuda você a diminuir o zoom ever possibilidades e oportunidades.Isso leva a novas ideias ou inovações empolgantes.
* Por outro lado, o pensamento orientado a detalhes tem tudo a ver descobrindo todos os aspectosque irão ajudá-lo a executar um plano.Em outras palavras, as peças que compõem o seu quebra-cabeça.Existem todos os tipos de problemas emmundo dos negócios que podem se beneficiarfuncionários que têm uma visão gerale uma maneira de pensar detalhista

#### Qual é a causa raiz de um problema?
Uma causa raiz é a razão pela qual um problema ocorre.Se pudermos identificar e nos livrar de uma causa raiz,podemos evitar que esse problema volte a acontecer.Uma maneira simples de envolver sua cabeçacausas profundas é com o processo chamado Cinco Porquês.

#### Onde estão as lacunas em nosso processo?
Para isso, muitas pessoas usarãoalgo chamado análise de lacunas.A análise de lacunas permite examinar e avaliar como um processofunciona atualmente em ordempara chegar onde você quer estar no futuro.As empresas realizam análises de lacunaspara fazer todos os tipos de coisas,como melhorar um produto ou tornar-se mais eficiente.A abordagem geral para a análise de lacunas éentender onde você está agoraem comparação com onde você quer estar.Então você pode identificar as lacunas que existem entreo estado atual e futuroe determinar como superá-los.

#### O que não consideramos antes?
Esta é uma ótima maneira de pensar sobre quais informaçõesou procedimento pode estar faltando em um processo,para que você possa identificar maneiras de fazermelhores decisões e estratégias no futuro

### O ciclo de vida dos dados: <br>
O ciclo de vida dos dados fornece uma estrutura genérica ou comum de como os dados são gerenciados. 
1. Planejamento: Durante o planejamento, uma empresa decide que tipo de dados precisa,como será gerido ao longo do seu ciclo de vida,quem será responsável por isso e os resultados ideais. (Decidir que tipo de dados são necessários, como serão gerenciados e quem será responsável por eles)
2. Captação de dados: É aqui que os dados são coletados de uma variedade de fontes diferentes etrouxe para a organização.Com tantos dados sendo criados todos os dias,as maneiras de coletá-lo são verdadeiramente infinitas. Um método comum é obter dados de recursos externos. (Colete ou traga dados de uma variedade de fontes diferentes)
3. Gerenciamento: Aqui estamos falando sobre como cuidamos de nossos dados,como e onde é armazenado,as ferramentas usadas para mantê-lo seguro e protegido, eas ações tomadas para garantir que ele seja mantido adequadamente.Essa fase é muito importante para a limpeza de dados. (cuidar e manter os dados. Isso inclui determinar como e onde ele é armazenado e as ferramentas usadas para fazê-lo)
4. Analise: Nesta fase, os dados são usados para resolver problemas, tomar grandes decisões e apoiar os objetivos de negócios.
5. Arquivamento: Arquivar significa armazenar dados em um local onde eles ainda estejam disponíveis, masnão pode ser usado novamente.Durante a análise, os analistas lidam com grandes quantidades de dados.Você pode imaginar se tivéssemos que classificar todos os dados disponíveis que estão foralá, mesmo que não fosse mais útil e relevante para o nosso trabalho? Faz muito mais sentido arquivá-lo do que mantê-lo por perto. (mantenha os dados relevantes armazenados para referência futura e de longo prazo)
6. Destruir: Isso é importante para proteger as informações privadas de uma empresa, bemcomo dados privados sobre seus clientes. (remova os dados do armazenamento e exclua todas as cópias compartilhadas dos dados)

### Observação: <br>
#### Precisamos considerar as fases pelas quais os dados passam em seu ciclo de vida e como isso impacta o processo de análise de dados. <br>
Embora o processo de análise de dados conduza seus projetos e ajude você a atingir suas metas de negócios, você deve entender o ciclo de vida de seus dados para usar esse processo. Para analisar bem seus dados, você precisa ter um entendimento completo deles. Da mesma forma, você pode coletar todos os dados que desejar, mas os dados só serão úteis se você tiver um plano para analisá-los. As fases Planejar e Perguntar envolvem planejar e fazer perguntas, mas abordam assuntos diferentes. A fase Ask no processo de análise de dados concentra-se no pensamento estratégico geral sobre as metas de negócios. No entanto, a fase Planejar concentra-se nos fundamentos do projeto, como quais dados você tem acesso, quais dados você precisa e onde irá obtê-los.

### Principais ferramentas do analista de dados: <br>
1 - Planilhas
As planilhas estruturam os dados de maneira significativa, permitindo que: 
* Colete, armazene, organize e classifique informações
* Identifique padrões e reúna os dados de uma maneira que funcione para cada projeto de dados específico
* Crie excelentes visualizações de dados, como gráficos e gráficos.

2 - Bancos de dados e linguagens de consulta SQL
Um banco de dados é uma coleção de dados estruturados armazenados em um sistema de computador. 
* Permitir que os analistas isolem informações específicas de um banco de dados
* Tornar mais fácil para você aprender e entender as solicitações feitas aos bancos de dados
* Permita que os analistas selecionem, criem, adicionem ou baixem dados de um banco de dados para análise

<br>

3 - Ferramentas de visualização
Os analistas de dados usam várias ferramentas de visualização, como gráficos, mapas, tabelas, gráficos e muito mais. 
* Transforme números complexos em uma história que as pessoas possam entender
* Ajudar as partes interessadas a chegar a conclusões que levem a decisões informadas e estratégias de negócios eficazes  
* Tenha vários recursos

### Noções básicas de planilhas <br>
#### Em uma tabela, um atributo é uma característica ou qualidade de dados usada para rotular uma coluna. Uma linha é chamada de observação. Uma observação inclui todos os atributos para o que está contido na linha. <br>
#### Uma planilha ajuda a estruturar dados em linhas e colunas, preparar dados para análise e criar visualizações de dados personalizadas. Para analisar melhor seus dados, limpe seu gráfico para torná-lo mais atraente visualmente e para esclarecer o significado dos dados, tornando seu gráfico mais descritivo. Para fazer isso, é importante adicionar títulos de gráficos e títulos de eixos. ### Noções básicas de banco de dados e consultas <br>
#### O que é uma consulta? <br>
Uma consulta é uma solicitação de dados ou informações de um banco de dados. Ao consultar bancos de dados, você usa SQL para comunicar sua pergunta ou solicitação. 
Cada linguagem de programação, incluindo SQL, segue um conjunto exclusivo de diretrizes conhecidas como sintaxe . A sintaxe é a estrutura predeterminada de um idioma que inclui todas as palavras, símbolos e pontuação necessários, bem como seu posicionamento adequado. Assim que você insere seus critérios de pesquisa usando a sintaxe correta, a consulta começa a trabalhar para extrair os dados solicitados do banco de dados de destino.<br>
#### A sintaxe de cada consulta SQL é a mesma: 
* Use SELECT para escolher as colunas que deseja retornar.
* Use FROM para escolher as tabelas onde as colunas desejadas estão localizadas.
* Use WHERE para filtrar determinadas informações.
