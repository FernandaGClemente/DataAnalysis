# Modelo empresarial
[&#x1F53C;](#Sumário:) Clique para voltar para o sumário

### <u>Evolução do modelo de orientação das empresas:</u><br>
##### - Produção: Demanda > Oferta (Maior volume de produção): Foco no atributo do produto. Exemplo: O cliente pode ter o carro da cor que quiser, contanto que seja preto. (Henry Ford) > Produz e entrego
    A Cor preta era mais barata e secava mais rápida, com isso a montagem em série poderia ser produzida em massa, em menos tempo e a um menor custo. 
<br>

##### - Vendas: Oferta > Demanada (Maior volume de vendas): Foco na persuasão. Exemplo: Ter uma conexão com o ambiente externo para saber com está a concorrência. > Produz e força a venda
    Outras empresas entraram no mercado e com isso era necessário conquistar o cliente, utilizando a persuassão, para vender mais que os concorrentes. 
<br>

##### - Mercado: Ações de fora para dentro (Cliente / Expectativa): Foco nas necessidades e desejos. Exemplo: A capacidade de produzir precisa está relacionada com a capacidade do mercado em absorver > O que o cliente quer que seja produzido
    Precisa entender quais são os hábitos de consumo, os comportamentos de consumo e as frequências de consumo para gerar competitividade dentro do negócio (Entender cada aspecto do negócio).
<br>
    
##### "A meta do marketing é conhecer e entender o consumidor tão bem, que o produto e o serviço se molde a ele e se venda sozinho." Peter Druker <br>
##### "Inteligência é a habilidade de transformar informações em ações e a informação deve ser relevante para mover a empresa." <br>
##### O que faço com a informação? Direciono investimentos, desenvolvo novos produtos, entendo o comportamento dos outros players (Concorrntes), tendências do mercado e dos consumidores, defino planos de expansão e defesa de participação de mercado (Market Share) <br>
##### Modelos mentais das organizações:
* Filme naufrago: Ilhado, se sente isolado, sem nenhuma informação sobre nada, está vulnerável ás mudanças de mercado e não enxerga os movimentos dos concorrentes.
* Filme titanic: Acredita na solidez da empresa para enfrentar o mercado, tem um bom produto/serviço, mas não enxerga nada adiante.
* Filme relatório minoritário (Minority report): Olha para o futuro, compreende o comportamento e as tendências de mercado, antecipa movimentos e aproveita as oportunidades.

# Informação de mercado

### <u>Fatores críticos de sucesso:</u><br>
#### Objetivos estratégicos da companhia: É imprescindível ter visão de negócio (Conhecer a cadeia do negócio de ponta a ponta). <br>
##### O que é marketing e como ele é usado?
* Produtos: Procedimentos para garantir que os produtos estejam prontos para venda
* Preços: Colocá-lo em um nível comparável aos concorrentes
* Promoções: Folhetos, anúncios e informações para gerar interesse (Comunicação)
* Lugar: Como e onde os produtos são distribuídos para venda 

<br>

### <u>Lugar (Acesso das informações do mercado):</u><br>
##### "Informação é tudo o que muda o meu comportamento futuro, o resto é ruído." Stephen Kanitz <br>
##### Saber quais informações são úteis para o negócio (Quais dados responderam as perguntas de negócio) → Como buscar essas informações? <br>
##### Será que já existe algum trabalho na empresa, já feito sobre o tema? (Interno) → Perguntar a equipe comercial para verificar no histórico de trabalhos de marketing. <br>
##### Temos alguma consultoria ou serviço de notícias que podemos consultar? (Externo) → Consultorias: Tendências, Ibope, Ims Health, etc. e Serviços de notícias: Thompson Reuters, Info estado, Bloomberg, etc. <br>
* Dados primários (Interno): São os dados de propriedade da empresa, como resultados de vendas, financeiros, marketing, premissas de planejamento, etc.
* Dados secundários (Externo): Já estes são aqueles que estão disponíveis fora dos domínios da empresa, como dados mercadológicos, setoriais, dos concorrentes e pesquisas realizadas or terceiros a seu respeito.
* Análise dos dados: Relatórios periódicos e AD HOC (Expressão "para esta finalidade" = Geralmente se refere a uma solução especificamente elaborada para um problema ou fim preciso, não generalizável e nem utilizável para outros propósitos), subsídios a tomada de estretégias, táticas e a posterior avaliação dos resultados auferidos (Próprios, concorrentes e/ou mercado)

<br>

### Tipos de consumidores: 
* Equilibritas/vulneráveis = Menos renda média mensal e menos variação de gastos.
* Abaladas = "Quem são tem cão caça com gato". Apesar de sofrerem com a pressão da crise, buscaram alternativas para continuar consumindo. As famílias ficam mais atentas ao carrinho de compra, tornando-se abertas a troca de marcas e priorizando produtos que entreguem uma boa relação de custo-benefício. Conscientes dos preços praticados pelas marcas e lojas, as promoções se tornaram importante gatilho para atrair essas famílias para o consumo. → Menos renda média mensal e Menos variações de gastos
* Conscientes = "Melhro prevenir do que remediar". Assumir riscos não faz parte do perfil, que se planeja para evitar desiquilibrio financeiro. Vivem com metas conservadores, buscam estabilidade, evitam compras parceladas, o que acaba refletindo no modo que consomem. Antes de ir as compras, já tem em mente o que vai comprar e evita muitas ida ao mercado, levando embalagens maiores para abastecer a despesa e economizar. → Menos renda média mensal e Menos variações de gastos 
* Promissoras = "Depois da tempestade, vem a bonança". Abrir mão do padrão conquistado não estava nos planos dessas famílias, mas, com a pressão da crise, tiveram que aprender a se virar. Ainda assim, muitas famílias foram impactadas e não conseguiram equilibrar totalmente seus gastos versus renda. Lazer e entreterimento fora de casa ficaram em segundo plano, abrindo espaço para novas experiencias com familias e amigos em casa. Possuem muito orgulho das suas conquistas e esperam um futuro mais próspero, almejando poupar dinheiro. → Mais renda média mensal e Mais variações de gastos
* Administradoras = "Prepare-se e capitalize o que vier". São as que mais possuem educação financeira e gostam de estar no comando da sua própria vida. Chefes de família e/ou donas de casa geralmente são sócios-proprietários ou profissionais liberais. Desejam manter a qualidade de vida conquistada e por isso, prezam pelo conforto e colocam na mesa bons produtos e preferencialmente marcas habituais, mas alternam os canais de compra para manter equilibrio. Ficam atentos a novidades, gostam de lançamentos e liberam novas tendências. → Mais renda média mensale e Mais variações de gastos

### <u>Produtos:</u> <br>
##### 2/3 dos novos itens não mantem os níveis de venda até o 3° ano por falta de informação e monitoramento. <br>
##### Fatores de sucesso de um produto:
* Proposta distinta: Ofereça um beneficio que torne o produto dificil de substituir
* Chama atenção: Seja notado por meio de elementos execucionais como nome, embalegem, etc.
* Mensagem simples: TRansmita sua mensagem com clareza. Seja simples. 
* Relevância: Seja a resposta para um problema relevante do consumidor
* Vantagens: Traga vantagens claras
* Credibilidade: Ofereça razões suficientes para que acreditem em você
* Facilidade de achar: Esteja nos lugares certos dentro da loja
* Custos aceitaveis: Com a correta equação de valor
* Mensagem clara e concisa: Comunique com foco
* Entrega de produto: Entregue X as promessas feitas
* Lealdade de produto: E seja forte no longo prazo com uma fórmula diferenciada e bom custo benefício. 
##### Foco nos fatorees fundamentais no início: Relevância, Proposta distinta, Vantagens, Credibilidade, Entrega de produto e Lealdade de produto.

### <u>Preços e Promoções:</u> <br>
##### A gestão de preços (Mix de preços) pode ser influênciada por objetivos financeiros, demanda de mercado ou capacidade de produção e atendimento da companhia. <br>
##### 5 níveis de maturidade de preço da companhia: 
* 1 - Modelo reativo
* 2 - Modelo controlado
* 3 - Valor percebido 
* 4 - Otimização
* 5 - Alta maturidade em pricing
##### Os níveis de maturidade serão alterados de acordo a organização dos processos, ferramentas de precificação e inteligência artificiais (Algoritmos de otimização) <br>
##### Que o Preço-mãe estipulado de um produto é diminuido por inúmeras despesas (Descontos escondidos, comerciais, financeiros, devoluções, impostos sobre a venda, vendas contratuais, bonificações, acordos comerciais, custos variaveis e custos fixos), por tanto é preciso saber mensura-lo corretamente.  <br>

##### Os prioncipais modelos de precificação:
1. Precificação em custos
    * Objetivo: Garantir o atingimento da margem operacional alvo.
    * Pontos fortes: Alinhamento com guidance financeiro da empresa.
    * Pontos fracos: Perda da referência de outros influenciadores do preço (valor percebido, concorrência e demanda).
2. Precificação na concorrência
    * Objetivo: Garantir a competividade x players.
    * Pontos fortes: Alinhamento de preços com o mercado.
    * Pontos fracos: Ausência da percepção do valor do produto próprio. Delegação da precificação para terceiros.
3. Precificação em produto
    * Objetivo: Garantir o alinhamento entre o produto e marca.
    * Pontos fortes: Alinhamento entre produto e marca.
    * Pontos fracos: Não atingimento do ponto ótimo (preço e quantidade vendida) de precificação que maximiza a receita.
4. Precificação em oferta e demanda
    * Objetivo: Garantir o atingimento do ponto ótimo de receita (preço x quantidade).
    * Pontos fortes: Atingimento do ponto ótimo de receita.
    * Pontos fracos: Atividade complexa (monitoramento da concorrência, teste de elasticidade, etc).

<br>

##### Tipos de posicionamento da marca (Construção do valor em Marca):
* Marca com valor agregado: Tem diferenciação e customização
* Marca com valor em qualidade (atributo): Tem inovação e pode ser considerada premium
* Marca com valor em preço: Tem massificação, pradronização e pode ser considerada básica

<br>

##### Valor: Uma oferta de marketing só alcançará êxito se proporcionar valor ao seu mercado alvo. Valor traduz-se na taxa dos benefícios em relação ao sacrifício necessário para obter esses benefícios. <br>
##### Preço: Precisa estar em equilíbrio com o valor, mesmo sendo influenciado pela demanda, concorrência, custos e produto. <br>
##### Descontos: Toda ineficiência da organização pode ser transformada em desconto. São difíceis de mensurar os impactos, difíceis de controlar, pode afetar negativamente a rentabilidade do negócio e nem sempre alavanca vendas. <br>
