### <u>Introdução:</u> 
* Tipo de dado: type(objeto)
* Entrada de valores utilizando o input: 
        
        variável = tipo(input( "Texto" ))
* Contar a quantidade de elementos: 
        
        len(objeto)
* Instalação de pacote: 
        
        pythpn -m pip install pacote/biblioteca
* Utilização dos módulos e suas respectivas funcões: 
        
        variável = modulo.função() 
* Acessando o dicionario através da chave e do valor: 
        
        dicionario[ "Chave" ] ou dicionario.keys() e dicionario.values()
* Apagar um elemento do dicionario: 
        
        del dicionario[ "Chave" ]
* Inclusão de um elemento no dicionario: 
        
        dicionario[ "Chave" ] = valor 
* Acessando o valor de um elemento e se não encontrar retorne o segundo parâmetro: 
        
        dicionario.get( "Chave", "Segundo parâmetro" )
* Inclusão de um elemento no sets: 
        
        sets.add( "Valor" ) 
* Acessando um elemento na tupla: 
        
        tupla[ Indice ] 
* Criação de uma matriz unidimensional, definindo o tipo de dado: 
        
        matriz = np.array( [ Lista ], dtype = np.tipo)
* Transformação dos tipos de dados da matriz: 
        
        matriz = matriz.astype(np.tipo) 
* Criação de uma matriz bidimensional: 

        matriz = np.array( [ [Lista], [Lista] ])
* Criação de uma matriz tridimensional: 
        
        matriz = np.array( [ [Lista], [Lista], [Lista] ])
* Acessando um elemento da matriz bidimensional: 
        
        matriz[ Indice ][ Indice ]
* Acessando um elemento da matriz tridimensional: 
        
        matriz[ Indice ][ Indice ][ Indice ] ou
        matriz = np.arange( QtdeElementos ).reshape( ( Linha, Coluna ) )
* Exibição dos valores das dimensões das matrizes: 
        
        matriz.shape
* Criação de uma array tipificado, definindo o tipo de dado e a quantidade de colunas e linhas: 
        
        array = np.empty( [ Linha, Coluna ], dtype = tipo)
* Criação de uma matriz com valores zeros: 
        
        matriz = np.zeros( [ Linha, Coluna ] )
* Criação de uma matriz com valores uns: 
        
        matriz = np.ones( [ Linha, Coluna ] )
* Criação de uma matriz quadrada com a diagonal com valor 1 e o restante, zeros: 
        
        matriz = np.eye( [ QtdeDeLinha/Coluna ] )
* Criação de uma matriz com valores aleatórios entre zero e um: 
        
        matriz = np.random.random( ( QtdeDeLinha/Coluna ) )
* Criação de uma matriz com valores aleatórios com distribuição normal, contendo valores negativos: 
        
        matriz = np.random.randn( ( QtdeDeLinha/Coluna ) )
* Criação de uma matriz com valores aleatórios, definindo a quantidade de colunas e linhas: 
        
        matriz = np.random.random( ( Linha, Coluna ) )
* Criação de uma matriz com valores aleatórios, com uso de semente e com quantidade de elementos: 
        
        array = np.random.default_rng( Nr )
        matriz = array.random( QtdeElementos )
* Criação de uma matriz com valores aleatórios de números inteiros, definindo a quantidade de colunas e linhas e a quantidade de elementos: 
        
        matriz = matriz.integers( QtdeElemento, size = ( Linha, Coluna ))      
* Criação de uma matriz com valores aleatórios, sem números repetidos: 
        
        matriz = np.array( [Lista] )
        matriz.unique()
* Realizando funções matemáticas (max, min, sum, mean, std): 
        
        matriz.funcao()
* Aplicando funções em todos os elementos (sqrt e exp): 
        
        np.funcao(matriz)
* Transposição de matrizes: 
        
        matriz = matriz.T ou matriz = matriz.transpose()
* Criação uma matriz com valores boleanos baseando nos valores do array: 
        
        matriz = (array operador Nr)
* Criação uma matriz com valores -1 e 1 baseando nos valores do array: 
        
        matriz = np.where(array operador Nr, 1, -1)
* Criação de um dataframe:
        
        dataframe = pd.read_extensao( "caminho/arquivo.extensao" )
* Exibição das quantidades das colunas e linhas do dataframe:
        
        dataframe.shape
* Resumo estatístico de colunas numéricas:
        
        dataframe.describe()
* Exibição das 5 primeiras linhas: 
        
        dataframe.head()
* Exibição das 5 ultimas linhas:
        
        dataframe.tail()
* Filtrar os valores de uma coluna:
        
        dataframe[["NomeColuna"]] ou dataframe.NomeColuna
* Filtrar os valores das linhas por indice:
        
        dataframe.loc[ Nr:Nr ]
* Filtrar os valores das linhas utilizando uma lista de valores:        
        
        dataframe.loc[ [ Lista ] ]
* Filtrar os valores das linhas, utilizando uma condição em uma coluna:        
        
        dataframe.loc[ dataframe[ "NomeColuna"] Operador "Valor" ou Nr]
* Filtrar os valores das linhas de uma coluna ou mais colunas, utilizando uma condição em outra coluna:       
        
        dataframe[ "NomeColuna" ].loc[ dataframe[ "OutraNomeColuna"] Operador "Valor" ou Nr] ou
        dataframe[ ["NomeColuna1","NomeColuna2"] ].loc[ dataframe[ "OutraNomeColuna"] Operador "Valor" ou Nr]
        
* Criação de uma série:  
        
        serie = pd.Series( [Lista ou Array ou Coluna de um dataframe] )
* Renomear as colunas:
        
        dataframe.rename(columns = {"NomeAntigo": "NomeNovo",...}, inplace=True)
* Excluir colunas:

        dataframe.drop( "NomeColuna", axis=1, inplace=True )
* Verificar se existe valores nulos:

        dataframe.isnull()
        dataframe.isnull().sum()
* Excluir colunas com valores nulos:
        
        dataframe.dropna()
* Preencher as celulas que tem valores nulos da coluna:

        dataframe[ "NomeColuna" ].fillna(NovoValor, inplace=True)
* Filtrar os valores das linhas e colunas por indice:
        
        dataframe.iloc[ IndiceLinha, IndiceColuna]
* Filtrar os valores das linhas e colunas por indice e por lista:
        
        dataframe.iloc[ [Lista], Indice ]
